## 1️⃣ A mecânica Figure vs Axes

- **Figure**: A janela, a "folha de papel" inteira.
- **Axes**: Os quadros/gráficos individuais desenhados nessa folha.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

# Cria 1 linha, 2 colunas. Retorna a figura (janela) e um array de Eixos [ax_esquerdo, ax_direito]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# Acessamos o eixo esquerdo via indexação de array!
axes[0].plot(x, np.sin(x), color="blue")
axes[0].set_title("Onda Senoidal")

# Eixo direito
axes[1].plot(x, np.cos(x), color="red")
axes[1].set_title("Onda Cossenoidal")

plt.tight_layout()  # Ajusta os espaçamentos para nada se sobrepor
plt.show()
```
