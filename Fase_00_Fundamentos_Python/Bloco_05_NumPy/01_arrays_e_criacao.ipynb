## 1️⃣ NumPy Arrays vs Listas Python

Por que não usar Listas normais? Porque o NumPy Array possui um **tamanho fixo na memória** e um **tipo de dado fixo** (todos os itens são do mesmo tipo, usualmente float32 ou int64), rodando operações em C puro (extremamente rápido).

```python
import numpy as np
import time

# Comparativo de Tempo
tamanho = 5_000_000

lista_python = list(range(tamanho))
array_numpy = np.arange(tamanho)

# Soma tradicional em Python
inicio = time.time()
soma_lista = [x + 5 for x in lista_python]
print(f"Tempo Lista Python: {time.time()-inicio:.4f}s")

# Soma Vetorizada do NumPy
inicio = time.time()
soma_numpy = array_numpy + 5
print(f"Tempo Numpy:        {time.time()-inicio:.4f}s")
```

## 2️⃣ Criando a partir de listas

```python
vetor_1d = np.array([1, 2, 3, 4])
print("1D:", vetor_1d, f"| Shape: {vetor_1d.shape} | Tipo: {vetor_1d.dtype}")

matriz_2d = np.array([
    [1.5, 2.0],
    [3.1, 4.2]
])
print("\n2D:\n", matriz_2d, f"\nShape: {matriz_2d.shape} | Tipo: {matriz_2d.dtype}")
```

## 3️⃣ Geradores Inteligentes (Factory Methods)

Em vez de digitar números, o NumPy possui atalhos para gerar tensores inteiros, vitais para inicializar pesos de Redes Neurais.

```python
print("Zeros (2x3):")
print(np.zeros((2, 3)))

print("\nOnes (3x3) - e forçando que sejam int8 para poupar memória:")
print(np.ones((3, 3), dtype=np.int8))

print("\nIdentidade:")
print(np.eye(3))

print("\nLinspace (5 valores bem divididos entre 0 e 1):")
print(np.linspace(0, 1, 5))

print("\nAleatório Uniforme (0 a 1):")
print(np.random.rand(2, 2))
```
