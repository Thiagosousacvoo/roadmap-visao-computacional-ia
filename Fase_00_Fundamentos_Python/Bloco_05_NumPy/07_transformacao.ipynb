## 🎯 Objetivos deste notebook

Mudar a forma (Shape) de um array sem mudar seus dados é uma rotina diária em redes neurais. Se um modelo de IA exige uma entrada `(Batch, Canais, Altura, Largura)` e você tem `(Altura, Largura, Canais)`, a transformação de shape é necessária.
- `reshape()`, `flatten()` e `ravel()`.
- Transposição.
- Concatenação de arrays.

---

## 1️⃣ Reshape

Reorganiza os elementos em um novo layout, desde que a **quantidade total de elementos** seja exatamente a mesma (ex: 4x3 pode virar 6x2 ou 12x1).

```python
import numpy as np

array_1d = np.arange(12)
print("Original 1D:", array_1d)

matriz_3x4 = array_1d.reshape((3, 4))
print("\nMatriz 3x4:\n", matriz_3x4)

# O poder do '-1':
# Se você colocar '-1' em uma das dimensões do reshape, o Numpy calcula matematicamente qual deve ser o valor daquela dimensão.
# Ex: Eu sei que quero 2 colunas, mas não sei quantas linhas. Deixo linhas como -1.
matriz_nx2 = matriz_3x4.reshape((-1, 2))
print("\nReshape automático com -1 (Nx2):\n", matriz_nx2)
```

## 2️⃣ Flatten (Achatamento)

O oposto do reshape. Pega uma matriz de múltiplas dimensões (ex: imagem 2D) e a transforma em um vetor 1D. Essencial antes da camada final (Dense/Linear) de uma rede neural clássica.

```python
imagem = np.array([[255, 0], [128, 64]])

# flatten() cria uma COPIA na RAM.
# ravel() cria apenas uma VIEW (atalho) na RAM, muito mais rápido, mas alterar o ravel altera a original.

vetorizado = imagem.flatten()
print("Imagem Flattened:", vetorizado)
```

## 3️⃣ Transposição e Swap Axes

Muito comum em processamento de imagens: Modelos PyTorch usam `(C, H, W)` e OpenCV/Numpy carregam imagens como `(H, W, C)`. Precisamos girar ou trocar esses eixos.

```python
img_hwc = np.ones((1080, 1920, 3))  # Altura, Largura, Canais

# .transpose permite reordenar os eixos informando a nova ordem dos índices (0, 1, 2)
# Para ir de (H, W, C) para (C, H, W), o eixo 2 vem pro lugar 0.
img_chw = img_hwc.transpose((2, 0, 1))

print("Original shape:", img_hwc.shape)
print("Novo shape:    ", img_chw.shape)
```

## 4️⃣ Concatenando Arrays

Juntando blocos como peças de Lego. Você pode concatenar verticalmente (`vstack`) ou horizontalmente (`hstack`).

```python
topo = np.zeros((2, 3))
base = np.ones((2, 3))

empilhado_vertical = np.vstack([topo, base])
print("VStack (Shape:", empilhado_vertical.shape, "):\n", empilhado_vertical)

# np.concatenate é mais genérico e permite escolher o axis.
junto_axis1 = np.concatenate([topo, base], axis=1)
print("\nConcatenando no axis=1 (Shape:", junto_axis1.shape, "):\n", junto_axis1)
```
