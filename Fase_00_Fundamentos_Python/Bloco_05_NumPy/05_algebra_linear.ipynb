## 1️⃣ Produto Escalar (Dot Product) de Vetores

O produto escalar multiplica elementos correspondentes e *soma* tudo. É a operação base das Redes Neurais (Peso * Entrada).

```python
import numpy as np

pesos = np.array([0.2, 0.8, -0.5])
entradas = np.array([1.0, 2.0, -1.0])

# Três formas de fazer a mesma coisa:
res_1 = np.sum(pesos * entradas) # Element-wise + sum
res_2 = np.dot(pesos, entradas)  # Numpy Dot
res_3 = pesos @ entradas         # Operador @ do Python (Mais moderno e recomendado)

print(f"Saída do Neurônio: {res_3:.2f}")
```

## 2️⃣ Multiplicação de Matrizes

Quando multiplicamos duas matrizes `A (m x n)` e `B (n x p)`, o número de colunas da primeira (n) DEVE ser igual ao número de linhas da segunda (n). A matriz resultante terá shape `(m x p)`.

```python
A = np.array([
    [1, 2],
    [3, 4]
]) # Shape (2,2)

B = np.array([
    [1, 2, 3],
    [4, 5, 6]
]) # Shape (2,3)

# Multiplicação Matricial com o operador @
resultado = A @ B

print(f"Shape de A: {A.shape} | Shape de B: {B.shape} -> Resultante: {resultado.shape}")
print("\nMatriz Resultante:\n", resultado)
```
