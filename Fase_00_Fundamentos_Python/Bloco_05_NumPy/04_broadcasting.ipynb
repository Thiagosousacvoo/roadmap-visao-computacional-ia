## 🎯 Objetivos deste notebook

O Broadcasting é o mecanismo mais poderoso do NumPy. Ele define as regras de como operar entre arrays com formatos (shapes) DIFERENTES.

---

## 1️⃣ O que é Broadcasting?

Broadcasting significa "transmissão" ou "espalhamento". Quando tentamos somar um array 3x3 com um array 1x3, o Numpy "estica" o menor array virtualmente para que eles fiquem compatíveis, sem copiar dados na RAM.

```python
import numpy as np

matriz = np.ones((3, 3))        # Shape: (3, 3)
vetor = np.array([10, 20, 30])  # Shape: (3,)

# Como o NumPy fará isso?
# Ele estica o vetor para que cada linha da matriz some 10, 20, 30.
print("Resultado do Broadcasting:\n", matriz + vetor)
```

## 2️⃣ As Regras do Broadcasting

O Numpy compara as dimensões (shapes) da direita para a esquerda. Duas dimensões são compatíveis se:
1. Elas forem iguais, OU
2. Uma delas for 1.

Se o array tiver menos dimensões, o Numpy adiciona dimensões com "1" à esquerda.

```python
# Exemplo: Matriz de Imagens (Batch_size=10, Altura=64, Largura=64, Canais=3)
# Shape: (10, 64, 64, 3)
# Quero subtrair a "Média RGB" (um vetor com 3 valores: [0.4, 0.5, 0.4]) -> Shape: (3,)

# O NumPy compara (direita p/ esquerda):
# Imagens: 10 x 64 x 64 x 3
# Media  :                  3
# 
# Compatível! O 3 bate. O resto vira "1": 1x1x1x3. Ele estica!

imagens = np.random.rand(10, 64, 64, 3)
media_rgb = np.array([0.485, 0.456, 0.406])

imagens_centralizadas = imagens - media_rgb
print(f"Subtração concluída com sucesso! Shape final: {imagens_centralizadas.shape}")
```
