## 1️⃣ Argumentos Posicionais Variáveis (`*args`)

Em Ciência de Dados, muitas vezes criamos funções que podem receber um número variável de tensores, listas ou parâmetros. O `*args` agrupa todos os argumentos passados em uma **tupla**.

> 💡 **Nota:** A palavra `args` é uma convenção. O que realmente importa é o asterisco `*`.

```python
# Exemplo prático de *args
def calcular_media(*valores):
    if not valores:
        return 0.0
    # valores é uma tupla internamente
    return sum(valores) / len(valores)

print(f"Média de 2 valores: {calcular_media(10, 20)}")
print(f"Média de 5 valores: {calcular_media(1, 2, 3, 4, 5)}")
```

```python
# O poder do "Unpacking" (Desempacotamento)
# O asterisco também serve para "abrir" uma lista e passá-la como argumentos separados.

notas_turma = [8.5, 9.0, 7.5, 10.0, 6.5]
# print(calcular_media(notas_turma))  # Isso daria erro! Pois passaria uma lista como um único argumento.

print(f"Média da turma: {calcular_media(*notas_turma):.2f}")
```

## 2️⃣ Argumentos Nomeados Variáveis (`**kwargs`)

O `**kwargs` (Keyword Arguments) captura qualquer argumento nomeado extra e o armazena em um **dicionário**. Muito usado ao passar hyperparâmetros para modelos de IA.

```python
def configurar_modelo(nome_modelo, **hiperparametros):
    print(f"--- Configurando {nome_modelo} ---")
    for chave, valor in hiperparametros.items():
        print(f" > {chave}: {valor}")
    print("-" * 30)

configurar_modelo("ResNet50", learning_rate=0.001, batch_size=32, otimizador="Adam")
```

```python
# Desempacotando dicionários com **
config_dict = {
    "dropout": 0.5,
    "epochs": 100,
    "momentum": 0.9
}

configurar_modelo("VGG16", **config_dict)
```

## 3️⃣ Funções como Cidadãs de Primeira Classe

Em Python, funções podem ser guardadas em variáveis, passadas como argumento para outras funções e até retornadas. Isso é essencial em pipelines de processamento de dados.

```python
def quadrado(x):
    return x ** 2

def cubo(x):
    return x ** 3

# Função que recebe outra função!
def aplicar_operacao(valores, operacao):
    resultados = []
    for v in valores:
        resultados.append(operacao(v))
    return resultados

dados = [1, 2, 3, 4, 5]
print("Quadrados:", aplicar_operacao(dados, quadrado))
print("Cubos:    ", aplicar_operacao(dados, cubo))
```

## 4️⃣ Funções Anônimas (`lambda`)

Quando precisamos de uma função rápida, simples e de uso único (geralmente para passar como argumento), usamos o `lambda`.

Sintaxe: `lambda argumentos: retorno`

```python
# Exemplo lambda básico
dobrar = lambda x: x * 2
print(f"Dobro de 5 é {dobrar(5)}")
```

```python
# Onde lambda brilha de verdade: ordenação!
modelos = [
    {"nome": "ResNet", "acuracia": 0.95},
    {"nome": "VGG", "acuracia": 0.91},
    {"nome": "EfficientNet", "acuracia": 0.97}
]

# Como ordenar essa lista pelo valor da "acuracia"?
modelos_ordenados = sorted(modelos, key=lambda modelo: modelo["acuracia"], reverse=True)

for m in modelos_ordenados:
    print(f"{m['nome']}: {m['acuracia']}")
```
