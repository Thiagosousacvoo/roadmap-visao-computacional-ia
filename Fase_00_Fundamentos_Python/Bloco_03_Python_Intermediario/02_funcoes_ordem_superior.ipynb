## 🎯 Objetivos deste notebook

Ao finalizar este notebook você vai dominar ferramentas poderosas para iterar em coleções sem usar o clássico `for` loop:
- `enumerate()` e `zip()`
- Paradigma funcional com `map()` e `filter()`

---

## 1️⃣ `enumerate()`

Frequentemente precisamos iterar sobre uma lista e também saber o **índice** do elemento. Evite usar `for i in range(len(lista))`. O `enumerate` é mais elegante (Pythonic).

```python
labels_imagens = ["cachorro.jpg", "gato.png", "passaro.jpg"]

# Jeito "errado" (Não Pythonic)
print("--- Jeito antigo ---")
for i in range(len(labels_imagens)):
    print(f"ID {i}: {labels_imagens[i]}")

# Jeito Pythonic
print("\n--- Com enumerate ---")
for i, imagem in enumerate(labels_imagens, start=1):
    print(f"Amostra {i} -> {imagem}")
```

## 2️⃣ `zip()`

O `zip` permite iterar sobre múltiplas listas ao mesmo tempo, "zippando" os valores em tuplas. É perfeito para alinhar X (dados) e Y (labels).

```python
imagens = ["img_001.jpg", "img_002.jpg", "img_003.jpg"]
preds = [0.98, 0.12, 0.85]
labels = ["Gato", "Cachorro", "Gato"]

for img, pred, label in zip(imagens, preds, labels):
    status = "✅ Acerto" if (pred > 0.5 and label == "Gato") or (pred <= 0.5 and label == "Cachorro") else "❌ Erro"
    print(f"{img} | Real: {label:8} | Predição: {pred:.2f} | {status}")
```

## 3️⃣ O Poder do `map()`

O `map` aplica uma função a **todos** os elementos de um iterável e retorna um objeto map (que pode ser convertido em lista).

```python
precos_dolar = [10.0, 25.5, 100.0, 5.0]
taxa_cambio = 5.20

# Usando map + lambda para converter tudo para reais
precos_reais = list(map(lambda p: round(p * taxa_cambio, 2), precos_dolar))
print(f"Em Reais: {precos_reais}")
```

## 4️⃣ Filtrando ruído com `filter()`

O `filter` retorna apenas os elementos em que a função retornar `True`. Ótimo para limpeza de datasets.

```python
confiancas = [0.99, 0.45, 0.88, 0.12, 0.95, 0.33]

# Queremos manter apenas predições com confiança > 0.8
preds_seguras = list(filter(lambda x: x > 0.8, confiancas))
print(f"Predições confiáveis: {preds_seguras}")
```

## 5️⃣ Bônus: List Comprehension

Em Python moderno, `map` e `filter` muitas vezes são substituídos por List Comprehensions, que costumam ser mais fáceis de ler.

Sintaxe: `[expressao for item in iteravel if condicao]`

```python
# O mesmo filtro e map de cima, mas com List Comprehension
preds_seguras_lc = [round(c * 100, 1) for c in confiancas if c > 0.8]
print(f"Predições confiáveis em %: {preds_seguras_lc}")
```
