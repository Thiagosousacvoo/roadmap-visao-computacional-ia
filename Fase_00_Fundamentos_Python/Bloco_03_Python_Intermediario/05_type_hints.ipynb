## 1️⃣ O Básico de Tipagem

Você informa o tipo dos argumentos com `:` e o tipo de retorno com `->`. A IDE (como o VSCode) vai te alertar se você passar os dados errados e, principalmente, vai te dar **autocomplete** excelente!

```python
def calcular_acuracia(acertos: int, total: int) -> float:
    """Retorna a acurácia como uma porcentagem."""
    return (acertos / total) * 100

acc = calcular_acuracia(90, 100)
print(f"Acurácia: {acc:.1f}%")

# calcular_acuracia("Muitos", "Todos") # O Python ainda deixa executar e gerará erro na divisão, mas a sua IDE vai pintar isso de amarelo alertando que os tipos estão errados.
```

## 2️⃣ Listas e Dicionários

A partir do Python 3.9, você pode usar os tipos normais de coleções com colchetes para especificar o que vai dentro deles.

```python
def extrair_nomes(dados: list[dict]) -> list[str]:
    """Recebe uma lista de dicionários e retorna uma lista apenas com as strings dos nomes."""
    return [item["nome"] for item in dados]

# Autocomplete funciona!
usuarios = [
    {"id": 1, "nome": "Alan"},
    {"id": 2, "nome": "Ada"}
]
print(extrair_nomes(usuarios))
```

## 3️⃣ O Operador de União (`|`)

Muitas vezes, uma variável pode aceitar Múltiplos Tipos, ou aceitar um valor `None`. O operador pipe `|` (Python 3.10+) facilita isso.

```python
def processar_imagem(caminho_imagem: str | None = None) -> str:
    if caminho_imagem is None:
        return "Imagem padrão do sistema carregada."
    
    return f"Lendo imagem do caminho: {caminho_imagem}"

print(processar_imagem())
print(processar_imagem("gato.jpg"))
```

## 4️⃣ TypeAlias: Deixando o código semântico

Em IA, muitas vezes lidamos com matrizes (tensores) complexos. Podemos criar "apelidos" de tipos para tornar as funções extremamente claras.

```python
from typing import TypeAlias

# BoundingBox de uma detecção no OpenCV (x, y, largura, altura)
BoundingBox: TypeAlias = tuple[int, int, int, int]
Score: TypeAlias = float
Classe: TypeAlias = str

Deteccao: TypeAlias = tuple[Classe, Score, BoundingBox]

def detectar_objetos(imagem: list[list[int]]) -> list[Deteccao]:
    """Simulação de um modelo Yolo."""
    return [
        ("Pessoa", 0.98, (10, 20, 100, 200)),
        ("Carro", 0.75, (250, 150, 300, 120))
    ]

resultados = detectar_objetos([])
for obj, precisao, bbox in resultados:
    print(f"Objeto: {obj:6} | Precisão: {precisao:.2f} | Coordenadas: {bbox}")
```
