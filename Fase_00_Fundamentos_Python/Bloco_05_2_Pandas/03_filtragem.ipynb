## 🎯 Objetivos deste notebook

Filtragem é a arte de fazer perguntas aos dados: "Quem são os clientes com mais de 30 anos que não pagaram a fatura?"
- Indexação Booleana (a mesma do NumPy!).
- Operadores `&` (E) e `|` (OU).
- Uso do `.loc`.

---

## 1️⃣ O Retorno da Máscara Booleana

Como o Pandas é feito em cima do Numpy, a mesma lógica se aplica: criamos um Series de True/False e o injetamos nos colchetes do DataFrame.

```python
import pandas as pd

dados = {
    "Nome": ["Ana", "Bruno", "Caio", "Diana", "Enzo"],
    "Idade": [25, 42, 19, 31, 22],
    "Status": ["Ativo", "Inativo", "Ativo", "Ativo", "Inativo"]
}
df = pd.DataFrame(dados)

# Passo 1: Criar a condição (Máscara)
mascara_jovens = df["Idade"] < 30
print("Série Booleana:\n", mascara_jovens)

# Passo 2: Filtrar!
print("\nDataFrame Filtrado:")
df_jovens = df[mascara_jovens]
display(df_jovens)
```

## 2️⃣ Múltiplas Condições

No Python puro usamos `and` e `or`. No Pandas, usamos `&` e `|`. 
> ⚠️ **MUITO IMPORTANTE:** Você é obrigado a colocar cada condição entre parênteses, senão as regras de precedência quebram o código.

```python
# Errado: df[df["Idade"] < 30 & df["Status"] == "Ativo"]

# Certo:
ativos_e_jovens = df[(df["Idade"] < 30) & (df["Status"] == "Ativo")]
display(ativos_e_jovens)
```

## 3️⃣ O poderoso `.loc` e `.iloc`

- `df.loc[linhas, colunas]`: Localiza por NOME DO RÓTULO ou CONDIÇÃO.
- `df.iloc[linhas, colunas]`: Localiza por ÍNDICE NUMÉRICO (0, 1, 2...).

```python
# Usando .loc para alterar um dado específico mediante uma condição
# Se o Status for "Inativo", vou mudar o nome da pessoa para "[BLOQUEADO]"

# Note o .copy() para evitar warnings!
df_modificado = df.copy()

df_modificado.loc[df_modificado["Status"] == "Inativo", "Nome"] = "BLOQUEADO"
display(df_modificado)
```
