## 1️⃣ O Conceito do GroupBy

O GroupBy pega o DataFrame, divide em "mini-tabelas" baseadas em alguma chave categórica, aplica uma função (soma, média) em cada tabela, e junta tudo de volta num DataFrame resumo.

```python
import pandas as pd

vendas = pd.DataFrame({
    "Regiao": ["Norte", "Sul", "Norte", "Sul", "Norte"],
    "Vendedor": ["João", "Ana", "Carlos", "Beatriz", "Diego"],
    "Total_Vendas": [1500, 2000, 3000, 1000, 1200]
})
display(vendas)

# Agrupando por Regiao e calculando a SOMA Total
resumo_soma = vendas.groupby("Regiao")["Total_Vendas"].sum()
print("\nSoma por Região:")
display(resumo_soma)
```

## 2️⃣ O Poderoso `.agg()`

O `.agg()` (Aggregate) permite passarmos um dicionário dizendo exatamente o que queremos calcular para cada coluna numérica.

```python
resumo_completo = vendas.groupby("Regiao").agg({
    "Total_Vendas": ["sum", "mean", "max"],
    "Vendedor": "count"  # Conta quantos vendedores temos
})

display(resumo_completo)
```
