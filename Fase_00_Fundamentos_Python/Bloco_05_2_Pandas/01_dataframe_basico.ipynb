## 🎯 Objetivos deste notebook

Pandas é o "Excel com esteroides" para Python. Construído sobre o NumPy, ele traz o famoso `DataFrame`.
- Criar Séries e DataFrames do zero.
- Acessar colunas (Features).
- Obter visões rápidas dos dados (`.head()`, `.info()`).

---

## 1️⃣ Series vs DataFrame

- **Series**: É uma coluna única, um vetor unidimensional com rótulos (índices).
- **DataFrame**: É uma tabela inteira. Um dicionário de Series.

```python
import pandas as pd

# Criando uma Series (Coluna)
idades = pd.Series([25, 30, 35, 40], name="Idade")
print("Série de Idades:\n", idades)

# Criando um DataFrame através de um Dicionário
dados_cadastrais = {
    "Nome": ["Ana", "Carlos", "Beatriz", "João"],
    "Idade": [25, 30, 35, 40],
    "Salario": [4500.0, 5200.0, 8000.0, 3100.0]
}

df = pd.DataFrame(dados_cadastrais)
print("\nDataFrame Completo:")
display(df) # Em notebooks, use display() para uma formatação HTML bonita em vez de print()
```

## 2️⃣ Analisando a Estrutura (Primeira coisa a se fazer!)

Quando carregamos um dataset com milhões de linhas, não podemos imprimi-lo inteiro.

```python
print("As primeiras 2 linhas:\n")
display(df.head(2))

print("\nInformações das Colunas e Tipos de Dados:")
df.info()
```

## 3️⃣ Seleção Básica

Se queremos tratar apenas uma coluna do DataFrame (ex: normalizar os salários para a IA), acessamos a coluna como uma chave de dicionário.

```python
coluna_salario = df["Salario"]
print(coluna_salario)

# Selecionando múltiplas colunas de uma vez (passamos uma LISTA de colunas)
subset = df[["Nome", "Salario"]]
display(subset)
```
