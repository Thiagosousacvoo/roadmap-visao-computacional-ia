## 1️⃣ Inteiros (`int`)

Números sem parte decimal. Em Python, `int` não tem limite de tamanho fixo.

```python
# --- Operações aritméticas ---
a = 17
b = 5

print(f"Adição:          {a} + {b}  = {a + b}")
print(f"Subtração:       {a} - {b}  = {a - b}")
print(f"Multiplicação:   {a} * {b}  = {a * b}")
print(f"Divisão float:   {a} / {b}  = {a / b}")   # Sempre retorna float!
print(f"Divisão inteira: {a} // {b} = {a // b}")  # Floor division
print(f"Módulo (resto):  {a} % {b}  = {a % b}")
print(f"Potência:        {a} ** {b} = {a ** b}")
```

```python
# --- Comparações: retornam bool ---
x = 10
print(f"{x} == 10 → {x == 10}")
print(f"{x} != 10 → {x != 10}")
print(f"{x} >  5  → {x > 5}")
print(f"{x} <  5  → {x < 5}")
print(f"{x} >= 10 → {x >= 10}")
print(f"{x} <= 9  → {x <= 9}")
```

## 2️⃣ Ponto Flutuante (`float`)

> ⚠️ **Armadilha clássica:** `0.1 + 0.2 ≠ 0.3` por causa da representação binária de ponto flutuante.

```python
# Representação interna de float
print(0.1 + 0.2)           # Não é 0.3!
print(0.1 + 0.2 == 0.3)    # False!

# Comparação correta com tolerância
import math
print(math.isclose(0.1 + 0.2, 0.3))  # True ✅
```

```python
# Arredondamento
pi = 3.141592653589793
print(f"Original:     {pi}")
print(f"round(pi, 2): {round(pi, 2)}")
print(f"round(pi, 4): {round(pi, 4)}")

# Notação científica
pequeno = 1.5e-10
grande  = 2.3e8
print(f"\nPequeno: {pequeno}")
print(f"Grande:  {grande}")
```

## 3️⃣ Strings (`str`)

Sequência imutável de caracteres. Em Python, strings são extremamente poderosas.

```python
# Criação e concatenação
nome     = "Visão"
sobrenome = 'Computacional'
completo = nome + " " + sobrenome

# f-string (preferência em Python moderno)
print(f"Completo: {completo}")
print(f"Tamanho: {len(completo)} caracteres")
print(f"Maiúsculo: {completo.upper()}")
print(f"Minúsculo: {completo.lower()}")
```

```python
# Indexação e slicing — CRÍTICO para numpy/imagens depois
texto = "Python para IA"
#         01234567890123

print(f"Primeiro char:    {texto[0]}")
print(f"Último char:      {texto[-1]}")
print(f"Primeiros 6:      {texto[:6]}")
print(f"Do índice 11:     {texto[11:]}")
print(f"Do 7 ao 10:       {texto[7:10]}")
print(f"Invertido:        {texto[::-1]}")
print(f"Cada 2 chars:     {texto[::2]}")
```

```python
# Métodos essenciais
frase = "   Visão Computacional e IA   "

print(repr(frase.strip()))              # Remove espaços das bordas
print(frase.strip().replace("IA", "ML"))
print(frase.strip().split(" "))         # Divide em lista

palavras = ["NumPy", "Pandas", "OpenCV"]
print(" | ".join(palavras))             # Junta lista em string
```

## 4️⃣ Booleanos (`bool`)

> 💡 Em Python, `bool` é subclasse de `int`. `True == 1` e `False == 0`.

```python
# Operadores lógicos
tem_gpu = True
tem_dados = False

print(f"and: {tem_gpu and tem_dados}")   # Ambos verdadeiros?
print(f"or:  {tem_gpu or tem_dados}")    # Pelo menos um?
print(f"not: {not tem_gpu}")             # Inverte

# Valores "falsy" em Python
print("\n--- Valores falsy ---")
for val in [0, 0.0, "", [], {}, None, False]:
    print(f"bool({repr(val):10}) = {bool(val)}")
```

```python
# bool é int — útil para contar verdadeiros
predicoes = [True, False, True, True, False, True]
acertos = sum(predicoes)  # Conta os True
acuracia = acertos / len(predicoes)
print(f"Acertos: {acertos}/{len(predicoes)}")
print(f"Acurácia: {acuracia:.1%}")
```

## 5️⃣ Conversão de Tipos (`type casting`)

```python
# Conversões comuns
print(int("42"))        # str → int
print(float("3.14"))    # str → float
print(str(100))         # int → str
print(int(True))        # bool → int: 1
print(int(False))       # bool → int: 0
print(int(3.9))         # float → int: trunca (não arredonda!)

# Verificar tipo
valor = 42
print(f"\nTipo de {valor}: {type(valor)}")
print(f"É int?  {isinstance(valor, int)}")
print(f"É str?  {isinstance(valor, str)}")
```

```python
# ⚠️ Conversões que geram erro — sempre proteger com try/except
entradas_usuario = ["42", "3.14", "abc", ""]

for entrada in entradas_usuario:
    try:
        numero = float(entrada)
        print(f"'{entrada}' → {numero} ✅")
    except ValueError as e:
        print(f"'{entrada}' → Erro: {e} ❌")
```
