## 1️⃣ Listas (`list`)

```python
# Criação e acesso
modelos = ["ResNet", "VGG", "EfficientNet", "YOLO", "U-Net"]

print(f"Lista:          {modelos}")
print(f"Tamanho:        {len(modelos)}")
print(f"Primeiro:       {modelos[0]}")
print(f"Último:         {modelos[-1]}")
print(f"Slicing [1:3]:  {modelos[1:3]}")
print(f"Invertida:      {modelos[::-1]}")
```

```python
# Métodos de modificação
scores = [0.82, 0.91, 0.75, 0.88]
print(f"Original:  {scores}")

scores.append(0.95)         # Adiciona no final
print(f"append:    {scores}")

scores.insert(0, 0.70)      # Insere na posição 0
print(f"insert(0): {scores}")

scores.remove(0.75)         # Remove a primeira ocorrência
print(f"remove:    {scores}")

ultimo = scores.pop()       # Remove e retorna o último
print(f"pop():     {scores} (removido: {ultimo})")

scores.sort(reverse=True)   # Ordena in-place (modifica a lista)
print(f"sort(rev): {scores}")
```

```python
# Listas aninhadas — base para matrizes (antes do NumPy)
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(f"Linha 0:          {matriz[0]}")
print(f"Elemento [1][2]:  {matriz[1][2]}")

# Iterar com índice
for i, linha in enumerate(matriz):
    for j, val in enumerate(linha):
        print(f"  [{i}][{j}] = {val}", end="")
    print()
```

## 2️⃣ Tuplas (`tuple`)

> Use tupla quando os dados **não devem ser alterados**: coordenadas, shapes de tensores, configurações.

```python
# Criação
shape_imagem = (224, 224, 3)   # Altura, Largura, Canais
ponto = (10, 20)               # x, y

print(f"Shape: {shape_imagem}")
print(f"Altura: {shape_imagem[0]}, Largura: {shape_imagem[1]}, Canais: {shape_imagem[2]}")

# Unpacking — extrair valores em variáveis
altura, largura, canais = shape_imagem
print(f"\nUnpacking → altura={altura}, largura={largura}, canais={canais}")

x, y = ponto
print(f"Ponto → x={x}, y={y}")
```

```python
# Funções retornando múltiplos valores são tuplas
def calcular_estatisticas(dados: list) -> tuple:
    media = sum(dados) / len(dados)
    minimo = min(dados)
    maximo = max(dados)
    return media, minimo, maximo   # Retorna tupla

valores = [0.82, 0.91, 0.75, 0.88, 0.96]
media, minimo, maximo = calcular_estatisticas(valores)
print(f"Média: {media:.4f} | Mín: {minimo} | Máx: {maximo}")
```

## 3️⃣ Sets (`set`)

> Use set quando precisar de **unicidade** ou **operações de conjunto** (união, interseção).

```python
# Remover duplicatas de uma lista
labels_com_dup = ["gato", "cão", "gato", "pássaro", "cão", "cão"]
classes_unicas = set(labels_com_dup)
print(f"Com duplicatas:  {labels_com_dup}")
print(f"Únicas (set):    {classes_unicas}")
print(f"Total único:     {len(classes_unicas)}")
```

```python
# Operações de conjunto — úteis para comparar datasets
classes_treino = {"gato", "cão", "pássaro", "peixe"}
classes_teste  = {"gato", "cão", "cobra", "hamster"}

print(f"União (|):              {classes_treino | classes_teste}")
print(f"Interseção (&):         {classes_treino & classes_teste}")
print(f"Diferença (treino-test):{classes_treino - classes_teste}")
print(f"Dif. simétrica (^):     {classes_treino ^ classes_teste}")

# 'in' em set é O(1) — muito mais rápido que list!
print(f"\n'gato' no treino?  {'gato' in classes_treino}")
print(f"'cobra' no treino? {'cobra' in classes_treino}")
```

## 4️⃣ Dicionários (`dict`)

> Use dict para mapeamentos `chave → valor`: metadados, configurações, contagens.

```python
# Criação e acesso
config_modelo = {
    "arquitetura": "ResNet50",
    "num_classes": 10,
    "learning_rate": 0.001,
    "batch_size": 32,
    "use_gpu": True
}

print(f"Arquitetura: {config_modelo['arquitetura']}")
print(f"LR:          {config_modelo['learning_rate']}")

# .get() com valor padrão — evita KeyError
print(f"Dropout:     {config_modelo.get('dropout', 0.0)}")
```

```python
# Iterar sobre dicionário
print("--- Configurações do Modelo ---")
for chave, valor in config_modelo.items():
    print(f"  {chave:20} : {valor}")

print(f"\nChaves: {list(config_modelo.keys())}")
print(f"Valores: {list(config_modelo.values())}")
```

```python
# Padrão clássico: contar ocorrências
predicoes = ["gato", "cão", "gato", "pássaro", "gato", "cão"]
contagem = {}

for classe in predicoes:
    contagem[classe] = contagem.get(classe, 0) + 1

print("Contagem de predições:")
for classe, count in sorted(contagem.items(), key=lambda x: x[1], reverse=True):
    barra = "█" * count
    print(f"  {classe:10} {barra} ({count})")
```
