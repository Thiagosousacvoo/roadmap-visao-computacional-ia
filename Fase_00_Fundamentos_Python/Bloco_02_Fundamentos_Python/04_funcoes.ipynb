## 1️⃣ Definição Básica

```python
# Anatomia de uma função
def calcular_iou(box1: tuple, box2: tuple) -> float:
    """
    Calcula Intersection over Union entre dois bounding boxes.

    Args:
        box1: (x1, y1, x2, y2) — canto superior-esquerdo e inferior-direito
        box2: (x1, y1, x2, y2)

    Returns:
        IoU entre 0.0 e 1.0
    """
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    intersecao = max(0, x2 - x1) * max(0, y2 - y1)
    area1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area2 = (box2[2] - box2[0]) * (box2[3] - box2[1])
    uniao = area1 + area2 - intersecao

    return intersecao / (uniao + 1e-6)

# Teste
b1 = (0, 0, 10, 10)
b2 = (5, 5, 15, 15)
print(f"IoU: {calcular_iou(b1, b2):.4f}")
```

## 2️⃣ Parâmetros Padrão (Default)

```python
def normalizar(valor: float, minimo: float = 0.0, maximo: float = 1.0) -> float:
    """
    Normaliza um valor para o intervalo [minimo, maximo].
    Por padrão retorna valor entre 0 e 1.
    """
    return (valor - minimo) / (maximo - minimo + 1e-8)

print(normalizar(50, 0, 255))    # Normaliza pixel
print(normalizar(0.75))          # Usa padrão [0, 1]

# ⚠️ ARMADILHA CLÁSSICA: nunca use lista mutável como default!
def errado(lista=[]):    # Bug!
    lista.append(1)
    return lista

def correto(lista=None): # Correto
    if lista is None:
        lista = []
    lista.append(1)
    return lista

print(errado())   # [1]
print(errado())   # [1, 1]  ← Bug! Mesma lista sendo reusada
print(correto())  # [1]
print(correto())  # [1]  ← Correto
```

## 3️⃣ Múltiplos Retornos

```python
def analisar_dataset(dados: list) -> tuple:
    """
    Retorna estatísticas descritivas de uma lista de valores.

    Returns:
        Tuple de (media, desvio_padrao, minimo, maximo, tamanho)
    """
    n = len(dados)
    media = sum(dados) / n
    variancia = sum((x - media) ** 2 for x in dados) / n
    std = variancia ** 0.5
    return media, std, min(dados), max(dados), n

scores = [0.82, 0.91, 0.75, 0.88, 0.96, 0.71, 0.84]

# Unpacking do retorno
media, std, minimo, maximo, total = analisar_dataset(scores)

print(f"N:     {total}")
print(f"Média: {media:.4f}")
print(f"Std:   {std:.4f}")
print(f"Min:   {minimo}  | Max: {maximo}")
```

## 4️⃣ Escopo: Local vs Global

```python
# Variável global — existe fora da função
learning_rate = 0.001

def ajustar_lr(fator: float) -> float:
    # Lê a global, cria local com mesmo nome
    novo_lr = learning_rate * fator   # Referencia global
    return novo_lr

print(f"LR ajustado: {ajustar_lr(0.1)}")  # 0.0001
print(f"LR original: {learning_rate}")     # 0.001 — não foi alterado
```

```python
# Funções dentro de funções (closures — prévia)
def criar_multiplicador(fator: float):
    """Retorna uma função que multiplica por fator."""
    def multiplicar(x: float) -> float:
        return x * fator   # 'fator' vem do escopo externo
    return multiplicar

dobrar   = criar_multiplicador(2.0)
triplicar = criar_multiplicador(3.0)

print(f"Dobrar 5:    {dobrar(5)}")
print(f"Triplicar 5: {triplicar(5)}")
```
