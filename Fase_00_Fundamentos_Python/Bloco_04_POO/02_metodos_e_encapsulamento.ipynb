## 1️⃣ Métodos de Instância

Métodos definem o que um objeto "sabe fazer". Diferente de funções soltas, eles sempre recebem o `self` como primeiro parâmetro e podem alterar o estado (atributos) do próprio objeto.

```python
class ModeloML:
    def __init__(self, nome: str):
        self.nome = nome
        self.treinado = False
        self.acuracia = 0.0
        
    def treinar(self, acc_alcancada: float):
        print(f"[{self.nome}] Treinamento iniciado...")
        self.treinado = True
        self.acuracia = acc_alcancada
        print(f"[{self.nome}] Treinamento concluído. Acc: {self.acuracia:.2f}")

    def prever(self, dado):
        if not self.treinado:
            return "❌ Erro: O modelo precisa ser treinado primeiro!"
        return f"Predição para '{dado}': Sucesso (Confiança: {self.acuracia})"

modelo = ModeloML("RandomForest")
print(modelo.prever("imagem.jpg"))
modelo.treinar(0.92)
print(modelo.prever("imagem.jpg"))
```

## 2️⃣ Métodos Mágicos (Dunders)

Dunder methods (Double Underscore) são métodos do Python que permitem que nossos objetos interajam com funções nativas, como `print()` (via `__str__`).

```python
class Pipeline:
    def __init__(self, versao: float, etapas: list[str]):
        self.versao = versao
        self.etapas = etapas
        
    def __str__(self):
        # O que será exibido se fizermos print(pipeline)?
        return f"<Pipeline v{self.versao} | Etapas: {len(self.etapas)}>"

p = Pipeline(1.2, ["Leitura", "Limpeza", "Treino"])
print(p)
```

## 3️⃣ Encapsulamento

Por convenção, em Python, se uma variável ou método começa com um underline `_`, ela é considerada "privada" ou "protegida". Significa que só deve ser alterada de dentro da classe, e não de fora.

```python
class ServidorGPU:
    def __init__(self, id_servidor: int):
        self.id = id_servidor
        # Atributo "privado": desenvolvedores externos não devem alterar isso!
        self._temperatura = 40.0 
        
    def processar_carga(self):
        self._temperatura += 15.5
        if self._temperatura > 80:
            print("⚠️ ALERTA: Superaquecimento!")

gpu = ServidorGPU(101)
gpu.processar_carga()

# Python NÃO proíbe acessar (diferente do Java/C++), mas é uma quebra de contrato.
# print(gpu._temperatura)  # Funciona, mas você não deveria fazer isso.
```
