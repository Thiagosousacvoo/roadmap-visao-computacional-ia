## 1️⃣ Herança ("É Um")

Criamos uma Classe "Filha" que herda características de uma Classe "Mãe/Pai". Ideal quando você tem vários modelos que fazem essencialmente as mesmas coisas básicas.

```python
class ArquivoDados:  # Classe Pai
    def __init__(self, caminho: str):
        self.caminho = caminho
        
    def info(self):
        print(f"Arquivo localizado em: {self.caminho}")

# Classe Filha: CSV "É UM" ArquivoDados
class ArquivoCSV(ArquivoDados):
    def ler_separador(self):
        print(f"Lendo {self.caminho} com separador vírgula.")

# Classe Filha: Imagem "É UM" ArquivoDados
class ImagemJPG(ArquivoDados):
    def mostrar_resolucao(self):
        print(f"Resolucao de {self.caminho}: 1920x1080")

csv = ArquivoCSV("/data/treino.csv")
csv.info()           # Herdou da classe Pai!
csv.ler_separador()  # Específico da classe Filha
```

## 2️⃣ Sobrescrita e o `super()`

O que acontece se a filha precisar do `__init__` do pai, mas quiser adicionar mais coisas? Usamos a função `super()`.

```python
class ModeloBase:
    def __init__(self, versao: str):
        self.versao = versao
        self.framework = "Desconhecido"

class ModeloPyTorch(ModeloBase):
    def __init__(self, versao: str, uso_gpu: bool):
        # Chama o __init__ do ModeloBase
        super().__init__(versao)
        self.framework = "PyTorch"  # Sobrescreve/define
        self.uso_gpu = uso_gpu
        
m = ModeloPyTorch("v1.5", True)
print(f"Framework: {m.framework} | GPU: {m.uso_gpu} | Versão: {m.versao}")
```

## 3️⃣ Composição ("Tem Um")

Composição é o ato de colocar um objeto *dentro* do outro. Em vez de herdar, um "Treinador" *TEM* um "Modelo" e um "Dataset". A comunidade Python prefere muito mais a Composição do que a Herança profunda.

```python
class LogManager:
    def gravar(self, msg: str):
        print(f"[LOG DISCO] {msg}")

class TreinadorDeRede:
    def __init__(self, nome_rede: str):
        self.nome_rede = nome_rede
        # Composição: instanciamos um outro objeto AQUI DENTRO
        self.logger = LogManager()
        
    def executar_treinamento(self):
        self.logger.gravar(f"Iniciando rede {self.nome_rede}...")
        # logica pesada aqui...
        self.logger.gravar("Treino concluído com sucesso.")

t = TreinadorDeRede("YoloV8")
t.executar_treinamento()
```
