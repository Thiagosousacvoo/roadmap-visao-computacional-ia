## 🎯 Objetivos deste notebook

---

## 1️⃣ Múltiplos Arquivos (`import`)

Se você tiver uma classe complexa, não deve deixá-la no seu `main.py`. Deve criar módulos separados.

````python
"""
Exemplo de estrutura de diretório em Python:

meu_projeto_ia/
│
├── main.py                <-- Arquivo que você executa (Entry Point)
├── requirements.txt       <-- Bibliotecas (pip install -r)
│
├── data/
│   ├── raw/               <-- CSVs ou imagens originais
│   └── processed/         <-- Dados limpos
│
└── src/                   <-- Código Fonte Base
    ├── __init__.py        <-- Torna a pasta 'src' um pacote importável
    ├── dataset.py         <-- Contém a classe Dataset
    ├── model.py           <-- Contém a classe Modelo
    └── utils.py           <-- Funções soltas (ex: plot_grafico)
"""
print("Entendido a separação modular de código.")
````

## 3️⃣ O bloco `if __name__ == "__main__":`

Essa é a trava de segurança do Python. Ela garante que, se alguém "importar" o `main.py` por engano, ele não comece a rodar o pipeline imediatamente. O código abaixo disso só executa se o script for executado DIRETAMENTE pelo terminal (`python main.py`).

````python
def main():
    print("Pipeline de dados em execução...")

if __name__ == "__main__":
    main()
````
