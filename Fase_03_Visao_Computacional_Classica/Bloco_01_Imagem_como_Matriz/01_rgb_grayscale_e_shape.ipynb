## 1️⃣ Criando Imagens do Zero

Vamos criar imagens pretas, brancas e com ruído usando apenas NumPy. Isso desmistifica o que é um arquivo JPG/PNG.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Imagem preta 100x100 (valores 0)
img_preta = np.zeros((100, 100), dtype=np.uint8)

# 2. Imagem branca 100x100 (valores 255)
img_branca = np.ones((100, 100), dtype=np.uint8) * 255

# 3. Imagem com ruído aleatório
img_ruido = np.random.randint(0, 256, (100, 100), dtype=np.uint8)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(img_preta, cmap='gray', vmin=0, vmax=255); axes[0].set_title('Preta (Zeros)')
axes[1].imshow(img_branca, cmap='gray', vmin=0, vmax=255); axes[1].set_title('Branca (255)')
axes[2].imshow(img_ruido, cmap='gray'); axes[2].set_title('Ruído (Aleatório)')
for ax in axes: ax.axis('off')
plt.tight_layout()
plt.show()
```

## 2️⃣ Explorando Imagens Coloridas (RGB)

Uma imagem colorida tem **3 canais** (Red, Green, Blue). É um array 3D: `(altura, largura, canais)`.

```python
# Imagem colorida (RGB)
img_color = np.zeros((100, 100, 3), dtype=np.uint8)

# Pintar o canal Vermelho (R) inteiro de 255
img_color[:, :, 0] = 255  # Canal 0: R

print(f'Shape da imagem: {img_color.shape} → Altura=100, Largura=100, Canais=3')
print(f'Tipo de dado:    {img_color.dtype}')
print(f'Maior valor:     {img_color.max()}')

plt.imshow(img_color)
plt.title('Canal Red (R) = 255', fontweight='bold')
plt.axis('off')
plt.show()
```

## 3️⃣ Combinando Canais

Cores na tela (padrão RGB) são formadas pela **mistura de luz** (sistema aditivo).

```python
img_cores = np.zeros((200, 300, 3), dtype=np.uint8)

# Retângulo Vermelho (R)
img_cores[20:100, 20:100] = [255, 0, 0]
# Retângulo Verde (G)
img_cores[20:100, 120:200] = [0, 255, 0]
# Retângulo Azul (B)
img_cores[20:100, 220:300] = [0, 0, 255]

# Misturas!
img_cores[120:180, 20:100] = [255, 255, 0]    # Amarelo (R+G)
img_cores[120:180, 120:200] = [255, 0, 255]   # Magenta (R+B)
img_cores[120:180, 220:300] = [0, 255, 255]   # Ciano (G+B)

plt.figure(figsize=(8, 5))
plt.imshow(img_cores)
plt.title('Combinação de Canais RGB', fontweight='bold')
plt.axis('off')
plt.show()
```
