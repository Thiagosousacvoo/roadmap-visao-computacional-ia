## 🎯 Mini-Projeto: Ajuste de Brilho e Contraste Manual

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Criar imagem sintética com gradiente
img = np.zeros((200, 300, 3), dtype=np.uint8)
for i in range(300):
    img[:, i, :] = int(255 * i / 300)

# BRILHO: somar constante a todos os pixels
brilho = 80
img_clara = np.clip(img.astype(np.int16) + brilho, 0, 255).astype(np.uint8)
img_escura = np.clip(img.astype(np.int16) - brilho, 0, 255).astype(np.uint8)

# CONTRASTE: multiplicar por fator
contraste = 1.8
img_alto_contraste = np.clip(img.astype(np.float32) * contraste, 0, 255).astype(np.uint8)
img_baixo_contraste = np.clip(img.astype(np.float32) * 0.4, 0, 255).astype(np.uint8)

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
imagens = [img_escura, img, img_clara, img_baixo_contraste, img, img_alto_contraste]
titulos = [f'Brilho -{brilho}', 'Original', f'Brilho +{brilho}',
           'Contraste 0.4x', 'Original', f'Contraste {contraste}x']
for ax, im, t in zip(axes.flat, imagens, titulos):
    ax.imshow(im)
    ax.set_title(t, fontsize=11, fontweight='bold')
    ax.axis('off')
plt.suptitle('Brilho = Somar | Contraste = Multiplicar', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print('⚠️ np.clip é ESSENCIAL: sem ele, valores >255 fazem overflow e corrompem a imagem!')
```

## Histograma: Antes e Depois

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, im, t in zip(axes, [img_escura, img, img_clara], ['Escura', 'Original', 'Clara']):
    ax.hist(im.ravel(), bins=256, range=(0, 255), color='steelblue', alpha=0.7)
    ax.set_title(f'Histograma — {t}', fontsize=11, fontweight='bold')
    ax.set_xlim(0, 255)
plt.tight_layout()
plt.show()
```
