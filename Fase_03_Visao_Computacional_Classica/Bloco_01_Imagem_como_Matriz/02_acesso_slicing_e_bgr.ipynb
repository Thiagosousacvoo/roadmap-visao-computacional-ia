## 1️⃣ Acesso a Pixels e Slicing

Se você quer um pedaço da imagem (Crop), basta fatiar a matriz: `img[y_start:y_end, x_start:x_end]`.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Usar uma imagem embutida de teste, se não houver arquivo local
from skimage import data
img_astro = data.astronaut()  # Retorna RGB

# 1. Slicing: Recortar o rosto do astronauta
# Coordenadas: y de 50 a 170, x de 150 a 270
rosto = img_astro[50:170, 150:270]

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_astro)
axes[0].set_title(f'Original {img_astro.shape}', fontweight='bold')
axes[1].imshow(rosto)
axes[1].set_title(f'Recorte {rosto.shape}', fontweight='bold')
for ax in axes: ax.axis('off')
plt.show()
```

## 2️⃣ Desmontando os Canais de Cor

Uma imagem RGB pode ser separada em 3 matrizes 2D. Podemos visualizá-las em grayscale para ver onde há mais ou menos de cada cor.

```python
R = img_astro[:, :, 0]
G = img_astro[:, :, 1]
B = img_astro[:, :, 2]

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(img_astro)
axes[0].set_title('RGB Completa', fontweight='bold')
axes[1].imshow(R, cmap='gray')
axes[1].set_title('Canal R (Vermelho)', fontweight='bold')
axes[2].imshow(G, cmap='gray')
axes[2].set_title('Canal G (Verde)', fontweight='bold')
axes[3].imshow(B, cmap='gray')
axes[3].set_title('Canal B (Azul)', fontweight='bold')
for ax in axes: ax.axis('off')
plt.tight_layout()
plt.show()

print('Dica: Nas imagens dos canais, áreas mais brancas indicam maior intensidade daquela cor.')
```

## 3️⃣ O Problema BGR do OpenCV

Por razões históricas (padrões de câmeras antigas), a função `cv2.imread` carrega os canais na ordem **Azul, Verde, Vermelho** (BGR).
O `matplotlib` espera RGB. Se não converter, o Azul troca de lugar com o Vermelho!

```python
# Vamos simular o carregamento do OpenCV salvando e lendo a imagem
# IMPORTANTE: skimage usa RGB. cv2 usa BGR.
cv2.imwrite('astro_teste.jpg', cv2.cvtColor(img_astro, cv2.COLOR_RGB2BGR))

# Agora lemos com cv2
img_lida_cv2 = cv2.imread('astro_teste.jpg')

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img_lida_cv2)
axes[0].set_title('Lida com cv2 (BGR)\n"Por que estou azul?"', fontweight='bold')

# A solução correta:
img_corrigida = cv2.cvtColor(img_lida_cv2, cv2.COLOR_BGR2RGB)
axes[1].imshow(img_corrigida)
axes[1].set_title('Convertida BGR → RGB\n"Ah, melhorou!"', fontweight='bold')

for ax in axes: ax.axis('off')
plt.show()
```
