## 1️⃣ Leitura e Escrita de Imagens

OpenCV lê em **BGR** (não RGB!). Sempre converter para exibir com matplotlib.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Criar imagem sintética (se não tiver arquivo)
img_bgr = np.zeros((200, 300, 3), dtype=np.uint8)
img_bgr[50:150, 50:250] = [255, 128, 0]  # Retângulo BGR (azul + verde = ciano)
cv2.rectangle(img_bgr, (80, 80), (220, 120), (0, 0, 255), 2)  # Borda vermelha em BGR

# Converter BGR → RGB para matplotlib
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].imshow(img_bgr)  # ERRADO — cores invertidas!
axes[0].set_title('Sem conversão (BGR) — ERRADO', fontsize=12, fontweight='bold')
axes[1].imshow(img_rgb)  # CERTO
axes[1].set_title('Com conversão (RGB) — CERTO', fontsize=12, fontweight='bold')
for ax in axes: ax.axis('off')
plt.tight_layout()
plt.show()
print('⚠️ Armadilha clássica: OpenCV = BGR, Matplotlib = RGB')
```

## 2️⃣ Resize — Redimensionamento

```python
# Criar imagem gradiente para ver o efeito
img = np.zeros((400, 600, 3), dtype=np.uint8)
for i in range(400):
    img[i, :, 0] = int(255 * i / 400)  # Gradiente vermelho
    img[i, :, 2] = int(255 * (1 - i / 400))  # Gradiente azul
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize para tamanho fixo
resize_fixo = cv2.resize(img, (224, 224))  # (largura, altura) ← OpenCV inverte!

# Resize com escala
resize_escala = cv2.resize(img, None, fx=0.3, fy=0.3)

fig, axes = plt.subplots(1, 3, figsize=(16, 4))
for ax, im, titulo in zip(axes, [img, resize_fixo, resize_escala],
                           [f'Original {img.shape[:2]}', f'Fixo {resize_fixo.shape[:2]}', f'Escala {resize_escala.shape[:2]}']):
    ax.imshow(im)
    ax.set_title(titulo, fontsize=12, fontweight='bold')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

## 3️⃣ Normalização para Redes Neurais

Redes neurais esperam valores entre 0 e 1 (float32), não 0-255 (uint8).

```python
# Normalização simples
img_uint8 = np.random.randint(0, 256, (224, 224, 3), dtype=np.uint8)
img_float = img_uint8.astype(np.float32) / 255.0

print(f'uint8:   dtype={img_uint8.dtype}, min={img_uint8.min()}, max={img_uint8.max()}')
print(f'float32: dtype={img_float.dtype}, min={img_float.min():.4f}, max={img_float.max():.4f}')

# Normalização ImageNet (padrão para transfer learning)
mean = np.array([0.485, 0.456, 0.406])  # Médias RGB do ImageNet
std = np.array([0.229, 0.224, 0.225])    # Desvios padrão
img_normalizado = (img_float - mean) / std
print(f'\nImageNet: min={img_normalizado.min():.4f}, max={img_normalizado.max():.4f}')
print('→ Valores podem ser negativos! Isso é normal e esperado.')
```
