## 🎯 Projeto Final: Sistema de Segurança (Detecção de Movimento)

---

A diferença entre frames consecutivos é a forma clássica de detectar movimento! Mas precisamos processar os deltas corretamente para não detectar pequenas tremidas.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Simular 3 frames de vídeo de segurança
f1 = np.ones((200, 300), dtype=np.uint8) * 100
f2 = f1.copy(); cv2.rectangle(f2, (50, 50), (100, 100), 50, -1)
f3 = f1.copy(); cv2.rectangle(f3, (70, 50), (120, 100), 50, -1)

diff1 = cv2.absdiff(f1, f2)
diff2 = cv2.absdiff(f2, f3)

_, t1 = cv2.threshold(diff1, 20, 255, cv2.THRESH_BINARY)
_, t2 = cv2.threshold(diff2, 20, 255, cv2.THRESH_BINARY)

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes[0,0].imshow(f1, cmap='gray'); axes[0,0].set_title('Frame 1')
axes[0,1].imshow(f2, cmap='gray'); axes[0,1].set_title('Frame 2')
axes[0,2].imshow(f3, cmap='gray'); axes[0,2].set_title('Frame 3')
axes[1,0].imshow(diff1, cmap='gray'); axes[1,0].set_title('Diff: F1 e F2')
axes[1,1].imshow(diff2, cmap='gray'); axes[1,1].set_title('Diff: F2 e F3')
axes[1,2].imshow(t2, cmap='gray'); axes[1,2].set_title('Thresh Final')
for ax in axes.flat: ax.axis('off')
plt.show()
```
