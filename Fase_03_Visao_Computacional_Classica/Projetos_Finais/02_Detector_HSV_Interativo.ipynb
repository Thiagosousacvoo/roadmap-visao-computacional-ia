## 🎯 Projeto Final: Chroma Key com OpenCV (Detecção HSV)

---

Vamos usar o espaço HSV para isolar as cores e criar máscaras, igual ao famoso 'Fundo Verde' do cinema.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = np.zeros((300, 400, 3), dtype=np.uint8)
cv2.circle(img, (100, 150), 60, (0, 0, 255), -1)    # Vermelho
cv2.circle(img, (200, 100), 50, (0, 255, 0), -1)    # Verde
cv2.circle(img, (300, 200), 45, (255, 128, 0), -1)  # Azul-ciano

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Intervalo do VERDE no OpenCV HSV
lower_green = np.array([35, 100, 100])
upper_green = np.array([85, 255, 255])

mask = cv2.inRange(hsv, lower_green, upper_green)
resultado = cv2.bitwise_and(img, img, mask=mask)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); axes[0].set_title('Original')
axes[1].imshow(mask, cmap='gray'); axes[1].set_title('Máscara Verde')
axes[2].imshow(cv2.cvtColor(resultado, cv2.COLOR_BGR2RGB)); axes[2].set_title('Isolado')
for ax in axes: ax.axis('off')
plt.show()
```

## 🛠️ Melhorando a Máscara

Muitas vezes a máscara fica com buracos ou ruídos. Nós podemos usar **Erosão e Dilatação** (`cv2.erode` e `cv2.dilate`) para fechá-los.

```python
kernel = np.ones((5,5), np.uint8)
mask_limpa = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
mask_limpa = cv2.morphologyEx(mask_limpa, cv2.MORPH_CLOSE, kernel)
print('Máscara limpa com Morfologia Matemática!')
```
