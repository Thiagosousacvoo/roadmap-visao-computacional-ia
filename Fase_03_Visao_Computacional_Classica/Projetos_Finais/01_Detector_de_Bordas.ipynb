## 🎯 Projeto Final: Detector de Bordas Avançado

---

O Canny Edge Detector é poderoso, mas sensível ao ruído e aos thresholds escolhidos. Neste projeto, você vai aplicar um pipeline completo de suavização e extração.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 1. Imagem Base
img = np.zeros((300, 400, 3), dtype=np.uint8)
cv2.rectangle(img, (30, 30), (180, 120), (200, 200, 200), -1)
cv2.circle(img, (300, 100), 60, (150, 150, 150), -1)
cv2.line(img, (50, 200), (350, 280), (220, 220, 220), 3)

# 2. Grayscale + Blur
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# 3. Teste de múltiplos Thresholds
thresholds = [(30, 80), (50, 150), (100, 250)]
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); axes[0].set_title('Original')

for ax, (t1, t2) in zip(axes[1:], thresholds):
    edges = cv2.Canny(blurred, t1, t2)
    overlay = img.copy()
    overlay[edges > 0] = [0, 255, 0]  # Pinta as bordas de verde
    ax.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
    ax.set_title(f'Threshold: {t1} - {t2}', fontweight='bold')
for ax in axes: ax.axis('off')
plt.tight_layout()
plt.show()
```
