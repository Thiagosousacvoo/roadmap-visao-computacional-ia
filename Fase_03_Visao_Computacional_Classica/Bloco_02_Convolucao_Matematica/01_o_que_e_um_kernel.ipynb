## 1️⃣ O Kernel da Média (Blur)

Kernel 3×3 cheio de 1/9 → calcula a **média** dos 9 vizinhos → imagem fica embaçada.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Kernels clássicos
kernel_blur = np.ones((3, 3)) / 9
kernel_sharpen = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
kernel_sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
kernel_sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
kernel_laplaciano = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])

print('Kernel Blur (3x3):')
print(kernel_blur.round(3))
print('\nKernel Sharpen:')
print(kernel_sharpen)
print('\nKernel Sobel X (borda vertical):')
print(kernel_sobel_x)
```

## 2️⃣ Aplicando Kernels com OpenCV

```python
# Criar imagem de teste com formas geométricas
img = np.zeros((300, 300), dtype=np.uint8)
cv2.rectangle(img, (50, 50), (150, 150), 255, -1)
cv2.circle(img, (220, 120), 50, 200, -1)
cv2.line(img, (30, 250), (270, 200), 180, 3)

# Aplicar filtros
blur = cv2.filter2D(img, -1, kernel_blur)
sharp = cv2.filter2D(img, -1, kernel_sharpen)
borda_x = cv2.filter2D(img, cv2.CV_64F, kernel_sobel_x)
borda_y = cv2.filter2D(img, cv2.CV_64F, kernel_sobel_y)
magnitude = np.sqrt(borda_x**2 + borda_y**2).astype(np.uint8)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
imagens = [img, blur, sharp, np.abs(borda_x).astype(np.uint8), np.abs(borda_y).astype(np.uint8), magnitude]
titulos = ['Original', 'Blur (média)', 'Sharpen', 'Sobel X (bordas vert.)', 'Sobel Y (bordas horiz.)', 'Magnitude']
for ax, im, t in zip(axes.flat, imagens, titulos):
    ax.imshow(im, cmap='gray')
    ax.set_title(t, fontsize=11, fontweight='bold')
    ax.axis('off')
plt.suptitle('Efeito de Diferentes Kernels', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 3️⃣ Por que isso importa para CNNs?

Uma CNN **aprende os kernels automaticamente** durante o treino!
- Primeiras camadas: kernels simples (bordas, texturas)
- Camadas profundas: kernels complexos (olhos, rodas, rostos)
