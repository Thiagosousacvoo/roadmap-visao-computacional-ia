## 1️⃣ Binarizando a Imagem e Encontrando Contornos

O `findContours` exige que a imagem de entrada seja **binária**, onde os objetos que queremos encontrar devem ser **brancos** e o fundo **preto**.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = np.zeros((300, 400, 3), dtype=np.uint8)
cv2.rectangle(img, (30, 30), (120, 100), (255, 255, 255), -1)
cv2.circle(img, (250, 80), 50, (255, 255, 255), -1)
cv2.ellipse(img, (200, 220), (80, 40), 30, 0, 360, (255, 255, 255), -1)
for _ in range(30):
    x, y = np.random.randint(0, 400), np.random.randint(0, 300)
    cv2.circle(img, (x, y), 3, (255, 255, 255), -1)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# RETR_EXTERNAL pega apenas o contorno de fora dos objetos.
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(f'O algoritmo encontrou {len(contours)} contornos (incluindo o ruído!)')
```

## 2️⃣ Filtrando e Extraindo Bounding Boxes

Podemos iterar sobre a lista de contornos, checar a área com `cv2.contourArea()` e ignorar os menores.

```python
img_result = img.copy()
area_min = 500
objetos = 0

for c in contours:
    area = cv2.contourArea(c)
    if area < area_min:
        continue  # Pula os pequenos
    objetos += 1
    
    # Extrair caixa delimitadora reta
    x, y, w, h = cv2.boundingRect(c)
    cv2.rectangle(img_result, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cv2.putText(img_result, f'Obj {objetos}: Area {int(area)}', (x, y-10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].imshow(img); axes[0].set_title('Original com ruído')
axes[1].imshow(img_result); axes[1].set_title(f'Filtrado e Detectado: {objetos} Objetos reais')
for ax in axes: ax.axis('off')
plt.show()
```
