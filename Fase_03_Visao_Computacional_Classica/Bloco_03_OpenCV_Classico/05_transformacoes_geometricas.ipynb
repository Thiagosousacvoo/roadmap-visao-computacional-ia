## 1️⃣ Flips (Espelhamento)

A operação mais rápida e simples, extremamente usada para duplicar dados de treino (ex: virar rosto horizontalmente).

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = np.zeros((200, 300, 3), dtype=np.uint8)
cv2.rectangle(img, (20, 20), (100, 80), (0, 255, 0), -1)
cv2.circle(img, (220, 150), 40, (0, 0, 255), -1)
cv2.putText(img, 'F', (100, 120), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 3)

flip_h = cv2.flip(img, 1)
flip_v = cv2.flip(img, 0)
flip_both = cv2.flip(img, -1)

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, im, t in zip(axes, [img, flip_h, flip_v, flip_both],
                     ['Original', 'Horizontal (1)', 'Vertical (0)', 'Ambos (-1)']):
    ax.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
    ax.set_title(t, fontweight='bold')
    ax.axis('off')
plt.show()
```

## 2️⃣ Translação e Rotação usando Matrizes Afins

A translação e a rotação exigem uma Matriz de Transformação `M` que diz ao OpenCV para onde cada pixel vai.

```python
h, w = img.shape[:2]

# M_trans: [[1, 0, tx], [0, 1, ty]]
M_trans = np.float32([[1, 0, 50], [0, 1, -30]])  # Dir 50, Cima 30
translated = cv2.warpAffine(img, M_trans, (w, h))

# M_rot: gerado pelo OpenCV em volta do centro
centro = (w // 2, h // 2)
M_rot = cv2.getRotationMatrix2D(centro, angle=45, scale=0.8)
rotated = cv2.warpAffine(img, M_rot, (w, h))

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(cv2.cvtColor(translated, cv2.COLOR_BGR2RGB)); axes[0].set_title('Translação')
axes[1].imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)); axes[1].set_title('Rotação 45° + Zoom Out')
for ax in axes: ax.axis('off')
plt.show()
```
