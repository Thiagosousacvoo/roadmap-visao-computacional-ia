## 🎯 Objetivos deste notebook

Se você tirar uma foto com um iPhone e outra com uma GoPro olho de peixe, elas serão diferentes. Como a IA sabe disso?

---

### Intrínsecas (A Lente) `K`
É uma matriz 3x3 que guarda o `Focal Length` e o `Centro Óptico`. Ela diz como o raio de luz entorta ao passar pelo vidro do seu celular.

### Extrínsecas (A Posição) `[R|t]`
É uma matriz de Rotação (`R`) e Translação (`t`) que diz EXATAMENTE onde o celular estava no ar (Latitude, Longitude e para onde estava apontando) no momento do "Click".
