## 🎯 Objetivos deste notebook

Como você diz para o computador que uma cadeira andou 2 metros pra frente? Em Visão 3D, adicionamos um "1" falso no final do vetor 3D `[x, y, z, 1]`. Isso permite que a **Translação** seja resolvida apenas com multiplicação de matrizes.

---

```python
import numpy as np

print("Imagine um ponto em (X=5, Y=0, Z=0):")
ponto_mundo = np.array([5, 0, 0, 1])  # Coordenada homogênea

print("\nA Matriz de Translação de +3 metros no eixo Y:")
T = np.array([
    [1, 0, 0, 0],
    [0, 1, 0, 3], # Translação em Y
    [0, 0, 1, 0],
    [0, 0, 0, 1]
])

print("Novo Ponto = T @ ponto_mundo")
print(T @ ponto_mundo)
```
