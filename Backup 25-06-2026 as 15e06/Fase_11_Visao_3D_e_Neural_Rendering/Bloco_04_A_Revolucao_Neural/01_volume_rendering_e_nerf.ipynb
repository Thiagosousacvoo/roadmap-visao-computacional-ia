## 🎯 Objetivos deste notebook

Em 2020 a Universidade de Berkeley parou o mundo com o NeRF. Eles provaram que você pode salvar uma CENA INTEIRA dentro dos Pesos e Vieses de uma simples Rede Neural Densa (MLP).

---

### Como a U-Net não entra aqui?
O NeRF não usa CNNs. Você passa como Input o `[X, Y, Z]` do raio da câmera e ela cospe a Cor `[R, G, B]` e a Densidade (O quão sólido o ar é naquele ponto). Se você atirar mil raios por pixel, você desenha a cena completa de qualquer ângulo não visto!
