## 1️⃣ Exceções — Por que e Como

```python
# Sem tratamento: programa quebra
# resultado = 10 / 0   # ← ZeroDivisionError

# Com tratamento: controle
def dividir(a: float, b: float) -> float:
    """Divide a por b com tratamento de erro."""
    try:
        resultado = a / b
    except ZeroDivisionError:
        print(f"  ❌ Divisão por zero! Retornando 0.0")
        return 0.0
    else:
        # Executa APENAS se não houve exceção
        print(f"  ✅ {a} / {b} = {resultado:.4f}")
        return resultado
    finally:
        # Executa SEMPRE (com ou sem exceção)
        print(f"  [divisão executada]")

dividir(10, 3)
print()
dividir(10, 0)
```

```python
# Capturar múltiplas exceções
def converter_para_float(texto: str) -> float | None:
    """Converte string para float com tratamento robusto."""
    try:
        return float(texto.strip())
    except ValueError:
        print(f"  ❌ ValueError: '{texto}' não é um número")
        return None
    except AttributeError:
        print(f"  ❌ AttributeError: entrada não é string (tipo: {type(texto)})")
        return None

entradas = ["3.14", "  42  ", "abc", None, "1e-3"]
resultados = [converter_para_float(e) for e in entradas]
print(f"\nResultados: {resultados}")
```

```python
# Levantar exceções (raise)
def validar_imagem(altura: int, largura: int, canais: int):
    """Valida dimensões de imagem."""
    if altura <= 0 or largura <= 0:
        raise ValueError(f"Dimensões inválidas: ({altura}, {largura})")
    if canais not in (1, 3, 4):
        raise ValueError(f"Número de canais inválido: {canais}. Use 1 (gray), 3 (RGB) ou 4 (RGBA)")
    print(f"  ✅ Imagem válida: ({altura}, {largura}, {canais})")

try:
    validar_imagem(224, 224, 3)  # OK
    validar_imagem(224, -1, 3)   # Vai falhar
except ValueError as e:
    print(f"  ❌ Erro de validação: {e}")
```

## 2️⃣ Leitura e Escrita de Arquivos

```python
import os

# Criar arquivo de exemplo para os exercícios
conteudo_exemplo = """modelo,acuracia,parametros_M,velocidade_fps
ResNet50,0.932,25.6,45
EfficientNetB0,0.951,5.3,120
VGG16,0.914,138.4,22
MobileNetV3,0.901,5.4,200
ViT-B16,0.961,86.0,30
"""

with open("modelos.csv", 'w', encoding='utf-8') as f:
    f.write(conteudo_exemplo)

print("✅ Arquivo 'modelos.csv' criado")
```

```python
# Leitura: read() — todo conteúdo de uma vez
with open("modelos.csv", 'r', encoding='utf-8') as f:
    conteudo = f.read()

print(f"Tipo: {type(conteudo)}")
print(f"Tamanho: {len(conteudo)} chars")
print("\nPrimeiros 100 chars:")
print(conteudo[:100])
```

```python
# Leitura: readlines() — lista de linhas
with open("modelos.csv", 'r', encoding='utf-8') as f:
    linhas = f.readlines()

print(f"Total de linhas: {len(linhas)}")
for i, linha in enumerate(linhas):
    print(f"  [{i}] {repr(linha)}")
```

```python
# Leitura linha a linha (eficiente para arquivos grandes)
with open("modelos.csv", 'r', encoding='utf-8') as f:
    cabecalho = f.readline().strip().split(',')
    print(f"Colunas: {cabecalho}")
    
    print("\n--- Dados ---")
    for linha in f:               # Itera o resto
        campos = linha.strip().split(',')
        if campos:                 # Evita linha em branco
            nome = campos[0]
            acc  = float(campos[1])
            print(f"  {nome:20} acc={acc:.3f}")
```

```python
# Escrita e append
novo_modelo = "YOLOv8n,0.887,3.2,150"

with open("modelos.csv", 'a', encoding='utf-8') as f:  # 'a' = append
    f.write(f"\n{novo_modelo}")

print("✅ YOLOv8n adicionado")

# Verificar
with open("modelos.csv", 'r') as f:
    print(f"Total de linhas agora: {len(f.readlines())}")
```
