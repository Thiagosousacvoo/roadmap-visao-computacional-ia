### 📜 O Desafio

Imagine que o seu sistema gera eventos de treinamento em formato string.
Cada evento tem um nível, um valor de loss e uma precisão (ex: `INFO: Loss 0.45, Acc 0.88`).

Sua tarefa é finalizar o pipeline base abaixo.

```python
import time
import functools
from typing import Callable, Any

# PASSO 1: Crie o Decorator de Tempo de Execução
def timer(func: Callable) -> Callable:
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        inicio = time.time()
        res = func(*args, **kwargs)
        print(f"[Timer] '{func.__name__}' rodou em {time.time()-inicio:.6f}s")
        return res
    return wrapper
```

```python
# PASSO 2: Crie uma função que processa eventos variados usando *args e Type Hints

Event: str # Um TypeAlias para eventos (apenas por semântica)

@timer
def processar_logs(*logs: str, threshold_acc: float = 0.90) -> list[str]:
    """
    Recebe qualquer número de strings de log no formato: "INFO: Loss X, Acc Y"
    Retorna apenas os logs que possuem "Acc" MAIOR que o threshold_acc.
    
    Dica: Você precisará dar um .split() na string para encontrar os valores numéricos.
    """
    logs_filtrados = []
    
    for log in logs:
        # TODO: Implementar a lógica de split e verificação numérico
        # Para extrair a Acc: log.split("Acc ")[1] 
        
        # Resolvendo para o aluno como exemplo do nível esperado:
        try:
            acc_str = log.split("Acc ")[1]
            acc_float = float(acc_str)
            if acc_float > threshold_acc:
                logs_filtrados.append(log)
        except (IndexError, ValueError):
            pass
            
    return logs_filtrados
```

```python
# Testando o pipeline!
log1 = "INFO: Loss 0.50, Acc 0.88"
log2 = "WARNING: Loss 0.20, Acc 0.95"
log3 = "INFO: Loss 0.10, Acc 0.98"
log4 = "ERROR: NaN detectado"

print("Filtrando logs com ACC > 0.90...")
logs_bons = processar_logs(log1, log2, log3, log4, threshold_acc=0.90)

print("\nLogs Aprovados:")
for lb in logs_bons:
    print(" -", lb)
```
