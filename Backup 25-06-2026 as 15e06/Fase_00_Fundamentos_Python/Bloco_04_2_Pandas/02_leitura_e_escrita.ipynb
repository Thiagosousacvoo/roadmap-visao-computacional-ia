## 1️⃣ Simulando um Arquivo CSV

Primeiro, vamos gerar um arquivo falso na sua máquina para podermos lê-lo.

```python
import pandas as pd
import os

csv_text = """id;produto;preco_brl;qtd_estoque
101;Teclado Mecânico;350.0;45
102;Mouse Gamer;220.0;120
103;Monitor 24";1200.5;15
104;Mousepad;45.9;200"""

with open("produtos_mock.csv", "w", encoding="utf-8") as f:
    f.write(csv_text)
    
print("Arquivo 'produtos_mock.csv' gerado com sucesso!")
```

## 2️⃣ Lendo Arquivos (`pd.read_csv`)

Se apenas chamarmos `pd.read_csv()`, ele pode quebrar. Repare que nosso arquivo usa PONTO E VÍRGULA (`;`) como separador, e o Pandas espera VÍRGULA (`,`) por padrão.

```python
# df_quebrado = pd.read_csv("produtos_mock.csv") # Isso faria tudo ficar numa coluna só!

df = pd.read_csv("produtos_mock.csv", sep=";")
display(df)
```

## 3️⃣ Salvando Arquivos (`df.to_csv`)

Após todo seu trabalho limpando os dados, você precisa salvar a base "Processed".

> ⚠️ **Atenção:** O Pandas por padrão salva a coluna de índice (0, 1, 2, 3...) no arquivo. Normalmente isso polui o CSV. Use `index=False`.

```python
# Vamos aumentar o preço do Mousepad em 10%
df.loc[df["produto"] == "Mousepad", "preco_brl"] *= 1.10

# Salvando sem o índice, e forçando o separador oficial (,)
df.to_csv("produtos_processado.csv", sep=",", index=False)
print("Salvo em 'produtos_processado.csv'. Abra o arquivo na sua IDE para ver!")
```
