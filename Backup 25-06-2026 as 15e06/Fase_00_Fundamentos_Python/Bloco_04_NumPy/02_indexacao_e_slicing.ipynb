## 1️⃣ Acesso Básico e Slicing

No Python padrão fazíamos `matriz[0][1]`. No NumPy, a sintaxe profissional e muito mais eficiente é `matriz[0, 1]`.

```python
import numpy as np

# Criando uma matriz 3x3
matriz = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print("Elemento central:", matriz[1, 1])

# Slicing: Quero todas as linhas da coluna 2
# O ':' sozinho significa "pegue tudo neste eixo"
print("Última coluna:", matriz[:, 2])

# Slicing: Pegar o bloco 2x2 do canto inferior direito
print("\nCanto 2x2:")
print(matriz[1:, 1:])
```

## 2️⃣ Indexação Booleana (Máscaras)

Essa é uma das ferramentas MAIS USADAS em Ciência de Dados. Podemos aplicar uma condição sobre a matriz inteira, gerando uma matriz de `True`/`False`. Ao passar essa máscara para a matriz, recebemos apenas os valores que passaram na condição.

```python
dados = np.random.randint(-5, 10, size=10)
print("Dados originais:", dados)

# 1. Cria a máscara (Vetor de Booleanos)
mascara_positivos = dados > 0
print("Máscara (> 0)  :", mascara_positivos)

# 2. Aplica a máscara para filtrar
dados_filtrados = dados[mascara_positivos]
print("Filtrados      :", dados_filtrados)

# 3. Dá para fazer direto!
print("Pares          :", dados[dados % 2 == 0])
```

```python
# Podemos usar a máscara para MODIFICAR valores diretamente.
# Exemplo: Transformando todos os números negativos em 0 (Isso é essencialmente a função ReLU de redes neurais!)

dados[dados < 0] = 0
print("Após a ReLU    :", dados)
```
