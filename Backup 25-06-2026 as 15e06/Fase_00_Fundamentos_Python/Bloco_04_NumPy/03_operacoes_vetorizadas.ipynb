## 1️⃣ Matemática Element-Wise (Elemento a Elemento)

Todas as operações aritméticas normais do Python (`+`, `-`, `*`, `/`, `**`) funcionam magicamente sobre matrizes inteiras.

```python
import numpy as np

A = np.array([1, 2, 3, 4])
B = np.array([10, 20, 30, 40])

print("Soma:       ", A + B)
print("Subtração:  ", B - A)
print("Mult (*):   ", A * B)    # ATENÇÃO: ISSO NÃO É MULTIPLICAÇÃO DE MATRIZES! É elemento a elemento.
print("Divisão:    ", B / A)
print("Potência:   ", A ** 2)
```

## 2️⃣ Funções Universais (UFuncs)

O Numpy já provê funções extremamente otimizadas para matemática, chamadas de ufuncs. Elas agem em todo o vetor de uma vez.

```python
angulo = np.array([0, np.pi/2, np.pi])

print("Seno:       ", np.sin(angulo))
print("Log:        ", np.log(np.array([1, 10, 100])))
print("Exponencial:", np.exp(np.array([0, 1, 2])))
```
