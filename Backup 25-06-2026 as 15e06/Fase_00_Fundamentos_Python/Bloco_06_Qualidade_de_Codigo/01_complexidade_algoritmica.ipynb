## 🎯 Objetivos deste notebook

Entender o "Big O Notation" (Notação Grande O). Como a performance do seu código escala quando o dataset passa de 1.000 para 1.000.000 de linhas?
- Tempo constante `O(1)`.
- Tempo linear `O(N)`.
- Tempo quadrático `O(N²)` (O grande inimigo!).

---

## 1️⃣ O(1) - Constante

Não importa o tamanho da lista, a operação demora sempre o mesmo tempo. Exemplo: acessar um índice de array ou chave de dicionário.

```python
import time

def acessar_primeiro(lista):
    return lista[0]

# Rodando com lista pequena e gigante
print("Acesso constante, o tempo não cresce junto com os dados.")
```

## 2️⃣ O(N) - Linear

Se a lista dobra de tamanho, o tempo de execução dobra. O `for` loop clássico sobre os dados é `O(N)`.

```python
def buscar_valor_n(lista, alvo):
    for item in lista:
        if item == alvo:
            return True
    return False

print("Para cada item a mais, damos um passo a mais no loop.")
```

## 3️⃣ O(N²) - Quadrático (Cuidado!)

Um loop dentro de um loop (`for` aninhado). Se os dados aumentam 10x, o tempo aumenta 100x! Redes Neurais costumam exigir multiplicações matriciais gigantes, por isso usamos GPUs (elas paralelizam os loops internos).

```python
def encontrar_pares_duplicados(lista):
    """Função ingênua e lenta O(N^2)"""
    for i in range(len(lista)):
        for j in range(len(lista)):
            if i != j and lista[i] == lista[j]:
                return True
    return False

print("Evite 'for dentro de for' sempre que possível usando NumPy/Pandas ou Sets(O(1)).")
```
