## 🎯 Objetivos deste notebook

Bem-vindo à Matemática de Ian Goodfellow. A Generative Adversarial Network não tenta aprender a imagem. Ela bota duas redes para brigarem até a morte.

---

## A Dinâmica Minimax
1. O **Generator (G)** pega ruído aleatório e cria uma imagem.
2. O **Discriminator (D)** olha imagens reais e as do G, e tenta adivinhar qual é falsa.
3. O **G** atualiza seus pesos apenas para tentar "enganar" o **D**.
4. O **D** atualiza seus pesos apenas para não ser feito de bobo.
