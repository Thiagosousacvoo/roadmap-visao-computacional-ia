## 🎯 Objetivos deste notebook

Antes de gerarmos do nada, precisamos saber "comprimir" e "descomprimir" uma imagem existente. Se você pegar uma foto 256x256 e passar por um Encoder até virar um vetor de 64 números (Espaço Latente), e depois passar por um Decoder, a foto sai igual? Quase.

---

### O Problema do Latente Discreto
Se você tentar inventar um vetor aleatório de 64 números e jogar no Decoder de um Autoencoder clássico, ele vai cuspir lixo. O modelo clássico **decora pontos esparsos** no espaço latente. Ele não serve para inventar imagens novas.
