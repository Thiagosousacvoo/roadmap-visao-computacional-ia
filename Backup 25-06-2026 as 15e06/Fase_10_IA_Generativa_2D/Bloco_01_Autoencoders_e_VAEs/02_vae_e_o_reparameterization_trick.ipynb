### Kullback-Leibler Divergence (L_KL)
O VAE é punido pela `L_KL` se o espaço latente dele se afastar demais de uma Curva de Sino (Normal Padrão). Isso obriga todas as fotos a ficarem aglomeradas no centro do espaço, o que permite **interpolação** sem gerar "lixo" no meio do caminho.
