## 🎯 Objetivos deste notebook

Entender $z = Wx + b$. O que são Pesos (W) e o que é o Bias (b)?

---

## O Conceito
O Perceptron recebe $X$ (Ex: Tamanho da casa e número de quartos). Ele multiplica pelo seu próprio vetor interno $W$ (O peso de importância de cada variável), soma um empurrão base $b$ e gera um resultado $z$.

- Se o Bias não existisse, a matemática obrigaria que todas as "linhas de decisão" cruzassem o número Zero. O Bias é o que permite que a rede "deslize" a reta pelo plano!
