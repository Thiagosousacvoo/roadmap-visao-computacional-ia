## 🎯 Objetivos deste notebook

Nós NÃO usaremos PyTorch agora. Você precisa entender que o Neurônio é apenas um "For Loop" estúpido que tenta corrigir o seu próprio vetor $W$ a cada vez que erra o palpite.

---

```python
import numpy as np

class PerceptronManual:
    def __init__(self, n_features, lr=0.01):
        self.w = np.random.randn(n_features) * 0.01  # Começa chutando pesos pequenos
        self.b = 0.0                               # Começa sem Bias
        self.lr = lr                               # O quão rápido ele muda de ideia

    def predict(self, X):
        z = X @ self.w + self.b
        return (z >= 0).astype(int)  # Se for positivo bate 1, se não 0.

    def fit(self, X, y, epochs=100):
        for epoch in range(epochs):
            for xi, yi in zip(X, y):
                # 1. Faz o palpite
                pred = self.predict(xi)
                # 2. Vê o quanto errou (Real - Palpite)
                erro = yi - pred
                # 3. Altera a mente (Pesos e Bias)
                self.w += self.lr * erro * xi
                self.b += self.lr * erro

print("Motor Neuronal criado!")
```
