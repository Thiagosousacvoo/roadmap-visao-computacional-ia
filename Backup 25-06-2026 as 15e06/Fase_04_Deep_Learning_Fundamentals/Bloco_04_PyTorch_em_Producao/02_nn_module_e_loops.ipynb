## 🎯 Objetivos deste notebook

A criação do Modelo Orientado a Objetos e o Loop de Treinamento Padrão de Indústria.

---

```python
import torch.nn as nn
import torch
import torch.optim as optim

# O Paradigma Profissional de criar Redes Neurais
class MinhaRede(nn.Module):
    def __init__(self):
        super().__init__()
        self.camadas = nn.Sequential(
            nn.Linear(784, 128), # Camada 1
            nn.ReLU(),           # Ativação
            nn.Dropout(0.2),     # Previne vício
            nn.Linear(128, 10)   # Saída de Classificação (10 números)
        )

    def forward(self, x):
        return self.camadas(x)

modelo = MinhaRede()
criterio_juiz = nn.CrossEntropyLoss()
otimizador = optim.Adam(modelo.parameters(), lr=0.001)

# ESSE É O LOOP QUE VOCÊ VAI FAZER PELOS PRÓXIMOS 10 ANOS NA SUA VIDA:
# 1. otimizador.zero_grad()
# 2. previsoes = modelo(x)
# 3. erro = juiz(previsoes, y_real)
# 4. erro.backward()
# 5. otimizador.step()
```
