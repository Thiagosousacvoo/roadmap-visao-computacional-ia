## 🎯 Objetivos deste notebook

Nenhuma rede treina carregando todas as imagens na RAM de uma vez. O `DataLoader` é o garçom que traz as imagens do HD direto para a GPU em pequenos "Batches" (lotes).

---

```python
from torch.utils.data import Dataset, DataLoader
import torch

# Classe Obrigatória do Pytorch
class MeuDatasetMock(Dataset):
    def __init__(self):
        # 100 imagens fake de 28x28
        self.X = torch.randn(100, 28 * 28)
        self.y = torch.randint(0, 10, (100,)) # Labels de 0 a 9

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

dataset = MeuDatasetMock()

# O Garçom!
garcom = DataLoader(dataset, batch_size=16, shuffle=True)

for batch_x, batch_y in garcom:
    print(f"O Garçom trouxe um Batch de Imagens de tamanho: {batch_x.shape}")
    break # Mostra só o primeiro!
```
