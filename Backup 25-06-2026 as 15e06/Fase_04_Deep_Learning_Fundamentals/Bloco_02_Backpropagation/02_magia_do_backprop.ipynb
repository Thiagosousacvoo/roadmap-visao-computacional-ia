## 🎯 Objetivos deste notebook

Aqui se separa os profissionais dos copiadores de código. Como a camada $W1$ sabe que errou, se ela nunca viu o resultado Final? 
Através da Regra da Cadeia (Chain Rule) do Cálculo!

---

## O Princípio
A Camada 2 (que gera o output final) sabe exatamente quanto errou. Ela diz: "Eu errei 10 pontos. Ei, Camada 1, eu só usei a sua resposta $A1$ multiplicada por 2. Então a culpa também é sua! Conserte seus pesos".

Os gradientes são passados de Trás pra Frente (Backwards). O PyTorch faz isso sozinho em UMA linha de código (`loss.backward()`), mas o que ele tá fazendo por baixo é rodando as derivadas parciais do seu grafo.
