## 🎯 Objetivos deste notebook

Na "Ida", a rede multiplica $X$ por $W1$, gera $A1$, multiplica por $W2$ e gera $y\_pred$.
E então o Juiz entra na jogada: A Função de Custo.

---

## O Juiz
- Regressão: Ele usa o MSE (`(y_real - y_pred)^2`).
- Classificação: Ele usa a assustadora Cross-Entropy (`-y * log(y_pred)`). Ela castiga a rede severamente se a rede tiver muita "certeza" de um chute errado.
