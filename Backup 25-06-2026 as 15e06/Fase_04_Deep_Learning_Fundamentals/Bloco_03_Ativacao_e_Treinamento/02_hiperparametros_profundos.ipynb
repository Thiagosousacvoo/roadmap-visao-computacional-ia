## 🎯 Objetivos deste notebook

- **Learning Rate (lr):** O tamanho do passo que a rede dá ladeira abaixo. Se for alto demais, ela erra o buraco. Se for baixo demais, ela leva a eternidade.
- **Epochs:** Quantas vezes ela vai ver a "Apostila Inteira" (Dataset inteiro) antes de fazer a prova?
- **Batch Size:** A rede estuda página por página (Batch=1)? Ou de 32 em 32 páginas (Batch=32)?
