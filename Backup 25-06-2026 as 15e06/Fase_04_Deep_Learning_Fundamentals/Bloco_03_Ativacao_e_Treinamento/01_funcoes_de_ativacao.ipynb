## 🎯 Objetivos deste notebook

Por que não podemos apenas somar $W * X$? Porque matemática linear empilhada continua sendo linear! Para que a IA crie linhas curvas no espaço, ela precisa passar o valor num filtro Não-Linear (Ativação).

---

```python
import torch
import matplotlib.pyplot as plt

# Criando dados de -10 a 10
z = torch.linspace(-10, 10, 100)

# 1. Sigmoid (Esmaga entre 0 e 1, boa para Probabilidades, mas morre nas pontas)
ativacao_sigmoid = torch.sigmoid(z)

# 2. ReLU (O grande Rei moderno. Se for negativo zera, se for positivo deixa passar!)
ativacao_relu = torch.relu(z)

# 3. Tanh (Esmaga entre -1 e 1, bom para centralizar dados negativos)
ativacao_tanh = torch.tanh(z)

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1).set_title("Sigmoid")
plt.plot(z.numpy(), ativacao_sigmoid.numpy())

plt.subplot(1, 3, 2).set_title("ReLU")
plt.plot(z.numpy(), ativacao_relu.numpy())

plt.subplot(1, 3, 3).set_title("Tanh")
plt.plot(z.numpy(), ativacao_tanh.numpy())
plt.show()
```
