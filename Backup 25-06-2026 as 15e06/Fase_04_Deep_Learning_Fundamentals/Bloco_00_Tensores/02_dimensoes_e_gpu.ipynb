## 🎯 Objetivos deste notebook

Dominar os Shapes de uma vez por todas. Em Deep Learning, se você souber o `shape` de entrada e o `shape` de saída, você entende 80% do que o modelo faz.

---

## 1️⃣ Tabela de Dimensões
- Escalar: ()
- Vetor: (N)
- Matriz: (N, M)
- Tensor 3D (Imagem): (Canais, Altura, Largura)
- Batch de Imagens: (Batch_Size, Canais, Altura, Largura)

*Nota: O PyTorch usa Canais primeiro (C, H, W). O OpenCV usa (H, W, C). Cuidado!*

```python
import torch

# Simulando um Lote de 32 Imagens, RGB, de 224x224 pixels!
batch_de_imagens = torch.randn(32, 3, 224, 224)
print(f"O Shape é: {batch_de_imagens.shape}")

# 2️⃣ Mandando para a Placa de Vídeo (GPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Dispositivo Atual: {device}")

# Move o tensor da Memoria RAM (CPU) para a VRAM (GPU)
batch_na_gpu = batch_de_imagens.to(device)
```
