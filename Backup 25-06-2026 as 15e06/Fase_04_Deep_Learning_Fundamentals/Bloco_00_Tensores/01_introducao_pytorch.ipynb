## 🎯 Objetivos deste notebook

Nesta fase nós transicionamos para o PyTorch. O PyTorch é essencialmente o "NumPy que roda na Placa de Vídeo" e que calcula Derivadas automaticamente (Autograd). 

---

```python
import torch
import numpy as np

# Criando um Array Numpy
arr_np = np.array([1, 2, 3])
print("NumPy:", type(arr_np))

# Criando um Tensor PyTorch
tensor_pt = torch.tensor([1.0, 2.0, 3.0])
print("PyTorch:", type(tensor_pt))

# A ponte mágica
tensor_convertido = torch.from_numpy(arr_np)
print("Convertido rápido de NumPy para PyTorch:", type(tensor_convertido))
```
