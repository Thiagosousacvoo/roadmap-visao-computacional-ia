## 🎯 A Vitrine do Código

Seu código pode ser genial, mas se o README for apenas um bloco de texto branco... o Recrutador fecha a aba em 5 segundos.

---

### Elementos Obrigatórios em um README Nível Sênior:
- Um GIF ou Vídeo de 10 segundos mostrando o programa em ação (Ex: YOLO pintando caixas).
- O Problema real que a ferramenta resolve.
- Uma tabela dura de **Métricas Técnicas** (mAP, F1-Score, FPS, Tempo de Inferência).
- Instruções passo a passo (Copiar e Colar) de como rodar no Docker.
