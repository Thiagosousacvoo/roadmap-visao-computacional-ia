## 🎯 Objetivos deste notebook

Como impedir que a Árvore de Decisão memorize tudo e sofra overfitting?
- Hiperparâmetros vs Parâmetros aprendidos.
- Lasso (L1) e Ridge (L2).

---

## 1️⃣ Controlando a Árvore

Ao usar `DecisionTreeClassifier(max_depth=3)`, você está passando um **Hiperparâmetro**. É um botão que VOCÊ regula antes do treino. Os "Parâmetros" (o valor da quebra dentro do nó) a IA descobre sozinha.

## 2️⃣ Regularização (Colocando pesos no modelo)

Se damos 100 features para uma Regressão Linear, ela vai tentar usar as 100. 
- **Lasso (L1):** Força o modelo a ZERAR os pesos das colunas irrelevantes. (Age como Feature Selection).
- **Ridge (L2):** Diminui os pesos de colunas instáveis, mas não as zera.
