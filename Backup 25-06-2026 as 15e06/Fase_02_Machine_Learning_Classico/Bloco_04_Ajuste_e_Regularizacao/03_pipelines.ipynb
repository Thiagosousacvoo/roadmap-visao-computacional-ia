## 🎯 Objetivos deste notebook

Se você normalizar (`StandardScaler`) todo o X antes do `train_test_split`, VOCÊ CONTAMINOU SEU TESTE COM A MÉDIA GERAL DOS DADOS. Seu modelo vai roubar e ir bem. 
O Pipeline garante que a normalização só aconteça DENTRO do fluxo de treino, rodada por rodada.

---

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

# Monta a esteira de produção: Primeiro normaliza, depois treina.
pipeline = Pipeline([
    ('scaler', StandardScaler()), 
    ('svm', SVC(kernel='rbf'))
])

# Quando fazemos .fit() no pipeline, ele treina o Scaler APENAS no X_train,
# Transforma o X_train, e joga pro SVM!
pipeline.fit(X_train, y_train)

# Quando fazemos .predict(), ele usa o Scaler (já travado) pra transformar 
# o X_test, impedindo vazamento de dados!
previsoes = pipeline.predict(X_test)
```
