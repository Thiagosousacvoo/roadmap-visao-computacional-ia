## 🎯 Objetivos deste notebook

Dividir em apenas Treino/Teste significa que sua pontuação depende da SORTE de quais dados caíram no Teste. O Cross Validation resolve isso rodando 5 treinamentos diferentes!

---

## 1️⃣ K-Fold

Corta os dados em 5 fatias. Treina em 4, testa em 1. Repete isso 5 vezes trocando a fatia de teste. A sua nota real é a MÉDIA das 5 tentativas!

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
modelo = LogisticRegression(max_iter=200)

# 5 rodadas de testes rigorosos!
scores = cross_val_score(modelo, X, y, cv=5)
print(f"Notas de cada Rodada: {scores}")
print(f"Nota Final Consolidada: {scores.mean():.3f}")
```
