Como o computador sabe que está indo no caminho certo? Através de uma Função de Custo (Loss Function).

## 1️⃣ MSE (Mean Squared Error)

Usado para **Regressão** (Prever números contínuos, como preços, temperaturas, alturas).
$$ MSE = \frac{1}{N} \sum_{i=1}^{N} (y_{real} - y_{predito})^2 $$

```python
import numpy as np

y_real = np.array([100, 200, 300])
y_pred = np.array([110, 190, 250])

erro_absoluto = y_real - y_pred
print("Erros Brutos:", erro_absoluto)

erros_quadrados = erro_absoluto ** 2
print("Erros Quadrados:", erros_quadrados)

mse = np.mean(erros_quadrados)
print(f"MSE Final: {mse}\n")
print("Por que ao quadrado? 1. Remove números negativos. 2. Punições quadráticas forçam a IA a detestar erros grandes (errar 50 custa 2500, errar 10 custa 100).")
```
