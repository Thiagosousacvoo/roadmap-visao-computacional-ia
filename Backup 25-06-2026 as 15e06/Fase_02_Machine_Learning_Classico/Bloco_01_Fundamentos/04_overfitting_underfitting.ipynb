## 🎯 Objetivos deste notebook

O modelo precisa generalizar, e não decorar os dados.
- **Underfitting (Alto Viés):** O modelo é "burro" demais. Ele erra muito no treino e no teste.
- **Overfitting (Alta Variância):** O modelo decorou os ruídos. Ele não erra NADA no treino (100% de precisão), mas vai muito mal no teste.

---

## 1️⃣ O Diagnóstico Visual

Sempre que treinamos, monitoramos o Erro no Treino e o Erro na Validação.
- Se ambos estão caindo: Modelo está aprendendo de forma saudável.
- Se o erro no Treino continua caindo, mas o erro de Validação começa a **SUBIR**: Aconteceu o **OVERFITTING**. Pare de treinar imediatamente! (Early Stopping).
