## 🎯 Objetivos deste notebook

Entender o paradigma do Aprendizado de Máquina Supervisionado.
- Features (`X`) vs Target (`y`).
- Criação de um dataset sintético usando NumPy e Pandas.

---

## 1️⃣ A Diferença de Paradigmas

Na **Programação Clássica**, você entrega as `Regras` e os `Dados`, e o computador devolve as `Respostas`.
Em **Machine Learning**, você entrega os `Dados` e as `Respostas`, e o computador devolve as **`Regras`** (A função matemática que os mapeia)!

## 2️⃣ Matriz X e Vetor y

Todo modelo precisa que os dados entrem num formato matemático restrito.

```python
import numpy as np
import pandas as pd

# Simulando 5 casas (Amostras)
# Colunas (Features): Quartos, Metros Quadrados, Distância_Centro_km
X = np.array([
    [2, 50, 10.5],
    [3, 80, 5.0],
    [4, 120, 2.0],
    [1, 35, 15.0],
    [3, 90, 8.5]
])

# Vetor y (Target): Preço da casa em milhares
y = np.array([200, 350, 500, 120, 300])

print("Matriz X (Features):\n", X)
print("\nVetor y (Alvo):\n", y)
print("\nShape de X:", X.shape) # 5 amostras, 3 features
print("Shape de y:", y.shape) # 5 respostas correspondentes
```
