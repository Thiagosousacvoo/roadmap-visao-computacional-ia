## 🎯 Objetivos deste notebook

A Regressão Linear Simples define a reta ótima `y = mx + b`.
Nós vamos treinar ela na unha através de Gradient Descent para você ver de onde o "mágico" `model.fit()` do Sklearn tira os números.

---

## 1️⃣ Regressão Linear "No Braço" (Sem Bibliotecas de IA)

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados Fictícios de Tamanho de Casa(X) vs Preço(y)
np.random.seed(42)
X = np.random.rand(50, 1) * 10
y = 2.5 * X + np.random.randn(50, 1) * 2 + 5  # y = 2.5X + 5 + ruido

# Inicializando Pesos Aleatórios
w = np.random.randn(1, 1) # Slope (m)
b = np.random.randn(1)    # Intercept (b)

lr = 0.01  # Learning Rate
epochs = 50

plt.scatter(X, y, color="blue", label="Dados Reais")

# Treinamento via Descida do Gradiente!
for epoch in range(epochs):
    # Forward Pass (A Previsão)
    y_pred = X @ w + b
    
    # Erro (Apenas pra monitorar)
    erro = y_pred - y
    
    # Backward Pass (Cálculo do Gradiente usando a Regra da Cadeia)
    dw = (2/len(X)) * (X.T @ erro)
    db = (2/len(X)) * np.sum(erro)
    
    # Atualização de Pesos
    w = w - lr * dw
    b = b - lr * db
    
    # Plota algumas retas pelo caminho para visualizarmos
    if epoch % 10 == 0:
        plt.plot(X, X@w+b, alpha=0.3, color="red")

# Plota a reta final (perfeita!)
plt.plot(X, X@w+b, color="green", linewidth=2, label="Reta Final do Treino")
plt.legend()
plt.title("A IA Aprendendo sozinha a Matemática")
plt.show()

print(f"O Modelo adivinhou a formula! y = {w[0][0]:.2f}X + {b[0]:.2f}")
```

## 2️⃣ Como a Indústria faz (Scikit-Learn)

Em produção, nunca reimplementamos a roda. Usamos o Sklearn que resolve isso em 2 linhas através do `np.linalg.lstsq` otimizado em C++.

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X, y) # A Mágica de Treinar acontece aqui!

print(f"Pesos do Sklearn: w={modelo.coef_[0][0]:.2f}, b={modelo.intercept_[0]:.2f}")
print("Batem perfeitamente com nossa implementação manual!")
```
