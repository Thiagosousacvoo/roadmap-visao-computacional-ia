## 🎯 Objetivos deste notebook

Algoritmos baseados em árvores são os Reis absolutos em Dados Tabulares (Planilhas), derrotando Redes Neurais 90% das vezes.

---

## 1️⃣ Árvore de Decisão Simples

O algoritmo constrói uma cadeia de `if` e `else`. O nó só quebra se isso diminuir o "Caos" daquele grupo (Chamado matematicamente de redução de Gini Impurity ou Entropia).

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris

# O dataset de flores
iris = load_iris()
X, y = iris.data, iris.target

# max_depth protege de Overfitting
arvore = DecisionTreeClassifier(max_depth=3, random_state=42)
arvore.fit(X, y)
print("Árvore treinada e pronta para dividir!")
```

## 2️⃣ A Filosofia do Random Forest (Ensemble)

Uma única árvore vicia muito rápido nos dados (Overfitting extremo).
A Random Forest cria **Centenas** de árvores. Cada árvore treina apenas com amostras e colunas aleatórias.
No final, todas votam democraticamente na resposta. A genialidade da estatística diz que os ruídos (erros individuais) se anulam!
