## 🎯 Objetivos deste notebook

Como um modelo preve se um email é SPAM ou NÃO SPAM? Usando uma Regressão Linear + Função Sigmoide.
- Transição de prever um "Número infinito" para prever uma "Probabilidade (0 a 1)".

---

## 1️⃣ A Curva Sigmoide

A Sigmoide pega o output da regressão linear (que pode ir de `-Infinito` a `+Infinito`) e esmaga o número para que ele caia exatamente entre `0` e `1`.

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

valores_z = np.linspace(-10, 10, 100)
probs = sigmoid(valores_z)

plt.plot(valores_z, probs)
plt.axvline(0, color='red', linestyle='--')
plt.axhline(0.5, color='red', linestyle='--')
plt.title("A Curva Sigmoide")
plt.grid()
plt.show()
```

## 2️⃣ A Mágica da Fronteira

Se a probabilidade bater > 0.5, a IA crava que a classe é "1" (Verdadeiro, Doente, Cachorro).
Se bater < 0.5, a IA crava que é "0" (Falso, Saudável, Gato).
