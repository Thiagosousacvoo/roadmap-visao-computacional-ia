## 1️⃣ K-Nearest Neighbors (K Vizinhos mais Próximos)

Ele não possui um ciclo de treinamento! Ele literalmente guarda todos os dados na RAM.
Quando chega um dado novo, ele calcula a distância (L2 Euclidiana) para TODOS OS PONTOS do banco. Pega os K pontos mais próximos, e faz uma votação.
- K Pequeno (K=1): Overfitting fatal.
- K Grande (K=100): Underfitting.

## 2️⃣ Support Vector Machine (SVM)

Encontra a "Rodovia mais larga" possível (Margem de Supote) para separar duas classes.
Se as classes são como um "ovo cozido" (uma bola dentro da outra, impossível de cortar com reta), o SVM eleva os dados para a Terceira Dimensão (Z), onde eles de repente viram separáveis! Isso é a genialidade do **Kernel Trick**.
