## 🎯 Objetivos deste notebook

Na Fase 01 você construiu a álgebra do PCA (Autovetores da Matriz de Covariância). Agora vamos usar do jeito certo em 2 linhas de código.

---

```python
from sklearn.decomposition import PCA
import numpy as np

# Dataset inventado de 100 amostras com 50 Colunas inúteis
X_altadimensao = np.random.rand(100, 50)

# Queremos compactar a essência de 50 colunas em apenas 2 para plotar em 2D!
pca = PCA(n_components=2)
X_compactado = pca.fit_transform(X_altadimensao)

print("Dados Originais  :", X_altadimensao.shape)
print("Dados Compactados:", X_compactado.shape)
```
