## 🎯 Objetivos deste notebook

Se o seu modelo prevê "Preço de Casa", ele nunca vai acertar os centavos exatos. Não dá para usar Accuracy. Usamos medidores de Distância.

---

## 1️⃣ As Ferramentas

- **MAE (Mean Absolute Error):** Em média, o modelo errou X reais para mais ou para menos.
- **MSE (Mean Squared Error):** Pune severamente erros grotescos (eleva a diferença ao quadrado).
- **RMSE (Root MSE):** Mesma coisa do MSE, mas com raiz quadrada para voltar para a escala de Reais.
- **R2 Score:** De 0 a 1. O quanto o seu modelo é melhor do que simplesmente "chutar a média" de todos os preços?
