## A Solução: Monitoramento de Confiança
A API deve obrigatoriamente registrar um Log JSON com a `Confiança Média` a cada requisição. Se a confiança cair para menos de 60%, a API **salva aquela imagem automaticamente** na pasta `para_revisao/`.

Na sexta-feira, o engenheiro revisa essas imagens difíceis, anota elas no Roboflow e treina a versão V2 do modelo. O nome disso é **Active Learning**.
