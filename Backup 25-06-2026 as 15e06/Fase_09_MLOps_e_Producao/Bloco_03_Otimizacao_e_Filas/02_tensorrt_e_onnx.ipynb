## ONNX Runtime vs TensorRT
- **ONNX (.onnx):** A linguagem universal. Você pode pegar um modelo em PyTorch, converter pra ONNX, e pedir para seu time rodar ele usando C# ou NodeJS. É excelente para CPU pura.
- **TensorRT (.engine):** Mágica proprietária da NVIDIA. Ele re-compila a arquitetura da sua rede neural fisicamente fundida com a sua Placa de Vídeo específica. Ele atinge o FPS (Frames Per Second) máximo absoluto que a humanidade tem a oferecer atualmente em Deep Learning.
