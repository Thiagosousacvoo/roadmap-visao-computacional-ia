## 🎯 Objetivos deste notebook

Se o processamento do seu modelo (ex: Um Vídeo) durar 1 minuto, o celular do usuário vai dar Time-Out. O modelo HTTP síncrono padrão não suporta isso.

---

## A Solução Assíncrona
1. O Usuário faz POST do vídeo.
2. A API apenas gera um Número de Ticket: "job_id = 99", joga o vídeo num Banco de Fila (Ex: Redis ou RabbitMQ) e responde `202 ACCEPTED`.
3. O usuário fica batendo no endpoint `GET /status/99` de 5 em 5 segundos.
4. Quando o Worker secundário terminar a GPU pesada e salvar no Redis, o status muda para `DONE` e o usuário recebe os dados.
