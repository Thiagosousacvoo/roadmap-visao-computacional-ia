## 🎯 Objetivos deste notebook

"Professor, o YOLO não roda porque faltou um DLL do Visual C++ no Windows". Isso não existe no mundo corporativo. Nós embalamos nosso código, nosso Python e todo o nosso Linux num bloco isolado: O **Container Docker**.

---

Em vez de dar para o seu colega o código fonte, você dá para ele uma **Imagem Docker**.
E ele só precisa digitar:

`docker run -p 8000:8000 minha-api-visao`
