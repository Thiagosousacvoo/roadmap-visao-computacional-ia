## YOLO (You Only Look Once)
O YOLO pega a imagem, divide num Grid gigantesco (ex: 20x20). Cada quadradinho do grid é encarregado de "cuspir" uma ou duas caixas preditivas. O YOLO olha a imagem inteira de uma vez só. O processamento que demorava 2 segundos caiu para **20 milissegundos**.
