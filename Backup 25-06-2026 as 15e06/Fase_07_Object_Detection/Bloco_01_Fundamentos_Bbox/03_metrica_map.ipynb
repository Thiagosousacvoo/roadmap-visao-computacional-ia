## 🎯 Objetivos deste notebook

Acurácia não existe em Detecção de Objetos. Imagine que não tem nenhuma placa de Pare na foto. A IA diz que não achou nada. A acurácia dela seria 100%! 
Nós usamos a **Mean Average Precision (mAP)**.

---

## Precision vs Recall (A eterna balança)
- **Precision (Precisão):** Quando a IA grita "GATO!", quão certa ela está? (Evita falsos positivos).
- **Recall (Revocação):** De todos os gatos que realmente existem na foto, quantos a IA achou? (Evita que objetos fiquem invisíveis).

O `mAP@50` é a média da performance de todas as classes, exigindo um IoU mínimo de 50% para considerar um acerto.
