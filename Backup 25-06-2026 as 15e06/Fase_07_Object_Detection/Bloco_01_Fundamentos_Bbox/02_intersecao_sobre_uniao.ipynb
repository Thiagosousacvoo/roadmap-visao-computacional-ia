## 🎯 Objetivos deste notebook

Como sabemos se a IA acertou? Se o humano desenhou a caixa da Placa de Pare na rua, e a IA desenhou a caixa 5 pixels pro lado, ela errou?
Para isso usamos o **IoU**.

---

## A Fórmula do Juiz
IoU = (Área de Interseção) / (Área Total Juntando as Duas Caixas)
- Se for **1.0**, as caixas estão perfeitamente coladas. Acerto de Ouro.
- Se for **0.5**, metade da caixa bate. A indústria considera isso como um Acerto Padrão.
- Se for **0.0**, a IA caçou borboletas e não viu o objeto.
