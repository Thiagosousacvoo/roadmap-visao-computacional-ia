## 🎯 Objetivos deste notebook

Pegaremos o mesmo Dataset da Fase 04 (Os números escritos à mão), só que desta vez não vamos esmagá-los em 784 pixels. Vamos passar os blocos numéricos pelo espremedor convolucional!

---

```python
import torch.nn as nn

class CNN_MNIST(nn.Module):
    def __init__(self):
        super().__init__()
        # BLOCO EXTRATOR DE FEATURES
        self.extrator = nn.Sequential(
            nn.Conv2d(1, 16, 3, padding=1), # Entra 1 canal, sai 16
            nn.ReLU(),
            nn.MaxPool2d(2, 2),             # Cai de 28x28 para 14x14
            
            nn.Conv2d(16, 32, 3, padding=1),# Entra 16, sai 32
            nn.ReLU(),
            nn.MaxPool2d(2, 2)              # Cai de 14x14 para 7x7
        )
        
        # BLOCO CLASSIFICADOR
        self.classificador = nn.Sequential(
            nn.Flatten(),                   # 32 canais * 7 * 7 = 1568 
            nn.Linear(1568, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 10)              # 10 números
        )

    def forward(self, x):
        features = self.extrator(x)
        return self.classificador(features)

print("A CNN foi instanciada!")
```
