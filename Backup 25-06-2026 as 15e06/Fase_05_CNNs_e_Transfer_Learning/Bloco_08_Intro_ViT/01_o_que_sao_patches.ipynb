## 🎯 Objetivos deste notebook

As CNNs dominaram por quase uma década. Mas agora os **Vision Transformers (ViT)** chegaram.
Em vez de escanear a imagem com bloquinhos convolucionais, o ViT corta a imagem em pedaços (Patches) como num quebra cabeça, e fofoca sobre todos eles ao mesmo tempo!
