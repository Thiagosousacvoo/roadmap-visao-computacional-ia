## 🎯 Objetivos deste notebook

Chega de `nn.Linear` puxando a imagem como se fosse uma "tripa longa". Na imagem, vizinhos importam! A Convolução analisa quadrantes mantendo a informação de `Cima-Baixo-Esquerda-Direita`.

---

```python
import torch
import torch.nn as nn

# 1 Imagem, 3 Canais (RGB), 224 de Altura, 224 de Largura
img_fake = torch.randn(1, 3, 224, 224)

# O nosso Scanner Inteligente
camada_conv = nn.Conv2d(
    in_channels=3,     # Recebe RGB
    out_channels=32,   # Cospe 32 "Filtros" Diferentes (32 Feature Maps)
    kernel_size=3,     # O scanner lê matrizes de 3x3
    stride=1,          # O scanner pula de 1 em 1 pixel
    padding=1          # Adiciona uma "borda falsa" preta ao redor da imagem
)

saida = camada_conv(img_fake)
print(f"Entrada: {img_fake.shape}")
print(f"Saída: {saida.shape}")
```
