## 🎯 Objetivos deste notebook

Se tivermos 5 camadas de convolução gerando 256 canais de imagem 224x224, a Placa de Vídeo explode! Precisamos encolher a imagem na marra. O Pooling pega blocos de pixels e escolhe apenas o maior valor (`MaxPool2d`).

---

```python
import torch\nimport torch.nn as nn\n\nimg_feature = torch.randn(1, 32, 224, 224)\n\n# Reduz a Altura e Largura pela Metade! Esmagando janelas 2x2.\npooling = nn.MaxPool2d(kernel_size=2, stride=2)\n\nsaida = pooling(img_feature)\nprint(f"Depois do Max Pooling: {saida.shape}")\n\n# Para jogar no classificador final, a imagem precisa virar uma tripa.\n# (Batch, Canais * Altura * Largura)\n# Mas os prós não usam mais Flatten. Eles usam AdaptiveAvgPool2d!\nprint("Flatten ou AdaptiveAvgPool2d esmagam as features para a camada Linear final!")
```
