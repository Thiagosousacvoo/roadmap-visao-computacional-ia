## 🎯 Objetivos deste notebook

Nossa CNN do CIFAR-10 provavelmente decorou as imagens (Overfitting). Para impedir isso, usaremos o `torchvision.transforms` para girar, torcer, inverter e escurecer as imagens aleatoriamente toda vez que a rede for treiná-las!

---

```python
from torchvision import transforms

# O pipeline Mágico de Mutações
# NOTA: ISSO SÓ SE APLICA AO TREINO. Você NUNCA muta imagens de Teste!
pipeline_treino = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5), # 50% de chance de espelhar
    transforms.RandomRotation(degrees=15),  # Gira até 15 graus
    transforms.ColorJitter(brightness=0.2), # Altera o brilho
    transforms.ToTensor(),                  # Vira matriz do PyTorch
    # Essa normalização usa a Média e o Desvio Padrão Oficial do Banco de Dados ImageNet
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
print("Pipeline de Augmentation criado!")
```
