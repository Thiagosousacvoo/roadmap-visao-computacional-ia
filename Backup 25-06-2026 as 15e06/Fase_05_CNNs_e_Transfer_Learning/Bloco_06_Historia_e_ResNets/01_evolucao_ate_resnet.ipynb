## 🎯 Objetivos deste notebook

Por anos as redes empacaram. Se você colocasse 100 camadas convolucionais, a rede piorava. O gradiente morria. Até que em 2015 a arquitetura **ResNet** inventou as *Skip Connections*.

---

## Skip Connections (O Atalho da ResNet)
A ideia é brutalmente simples: em vez da camada 3 tentar aprender a imagem toda do zero pela camada 2, a imagem da Camada 1 pega um atalho por fora (Skip) e se SOMAR com a saída da camada 3.

`Saída = Camada3(Camada2(x)) + x`

Isso permite construir redes de 152 camadas sem o gradiente desaparecer!
