## 🎯 Objetivos deste notebook

A ResNet original tem 1000 neurônios na última camada (pois o ImageNet tem 1000 classes). Se nós só queremos classificar "Cachorro" ou "Gato", nós só precisamos de 2 neurônios.

Nós vamos arrancar a última camada fora, e colocar uma nova!

---

```python
import torch.nn as nn
import torchvision.models as models

resnet = models.resnet18(pretrained=True)

# 1. Congelar os Olhos (Nós não queremos desaprender as texturas de focinhos!)
for param in resnet.parameters():
    param.requires_grad = False

# 2. Cortar a Cabeça e colocar uma nova
# A última camada linear da resnet chama-se "fc" (Fully Connected)
numero_de_features_originais = resnet.fc.in_features

# Substituímos a camada de 1000 saídas por uma camada de 2 saídas!
resnet.fc = nn.Linear(numero_de_features_originais, 2)

# OBSERVAÇÃO MÁGICA: A nossa nova camada "fc" foi recém criada,
# logo o requires_grad dela é True por padrão.
# O Otimizador vai treinar APENAS a cabeça!
print("Cabeça trocada com sucesso. Pronta para Gatos vs Cachorros!")
```
