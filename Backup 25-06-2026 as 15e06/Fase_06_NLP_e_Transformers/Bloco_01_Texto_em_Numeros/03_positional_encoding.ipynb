## 🎯 Objetivos deste notebook

Diferente das Redes Recorrentes antigas que liam a frase palavra por palavra, os Transformers leem TUDO ao mesmo tempo. Mas se a rede lê "O homem mordeu o cachorro" tudo ao mesmo tempo... como ela sabe quem mordeu quem?

---

O Transformer precisa que você "tatue" a posição na palavra. Para isso adicionamos uma onda matemática (Seno e Cosseno) por cima do Embedding da palavra. Isso informa para a rede: "Essa palavra é Cachorro, e a onda diz que ela está na Posição 4".
