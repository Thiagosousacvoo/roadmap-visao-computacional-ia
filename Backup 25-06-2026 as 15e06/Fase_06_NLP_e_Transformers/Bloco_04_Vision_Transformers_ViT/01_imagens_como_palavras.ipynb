## Patch Embedding
1. O ViT pega uma imagem 224x224.
2. Corta ela em 196 quadradinhos de 16x16 pixels (Os *Patches*).
3. Trata cada Quadradinho como se fosse uma PALAVRA de uma frase!
4. Joga os 196 quadradinhos num Transformer Encoder padrão.

```python
import torch
import torch.nn as nn

# Simulando o fatiador mágico do ViT
class PatchEmbedding(nn.Module):
    def __init__(self):
        super().__init__()
        # Truque sujo: Usar uma Convolução com Stride IGUAL ao Kernel (16x16)
        # Isso garante que a rede nunca varre, ela apenas corta quadrados perfeitos!
        self.cortador = nn.Conv2d(in_channels=3, out_channels=768, kernel_size=16, stride=16)

    def forward(self, img):
        # img: [1, 3, 224, 224]
        patches = self.cortador(img) # Sai: [1, 768, 14, 14]
        return patches.flatten(2).transpose(1, 2) # Achata para [1, 196, 768] (196 "Palavras" de 768 dimensões)

print("Fatiador de Patches (ViT) criado!")
```
