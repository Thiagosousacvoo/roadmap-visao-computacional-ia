## 🎯 Objetivos deste notebook

Por que ter apenas 1 mecanismo de Atenção julgando a frase? E se nós cortarmos os vetores e criarmos 8 "Cabeças" (Heads)?

---

## Multi-Head
A Cabeça 1 pode prestar atenção no "Sujeito" da frase. A Cabeça 2 pode prestar atenção em quem praticou a "Ação". A Cabeça 3 pode focar na "Rima" sonora.
Isso permite à IA interpretar o mundo com dezenas de perspectivas em paralelo!
