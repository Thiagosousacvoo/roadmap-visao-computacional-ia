## A Equação de Ouro

$$ Attention(Q, K, V) = Softmax(\frac{QK^T}{\sqrt{d_k}})V $$

- **Q (Query):** "O que eu estou procurando?"
- **K (Key):** "O que cada palavra pode oferecer?"
- **V (Value):** "Qual a informação real dessa palavra?"

Quando multiplicamos Q e K, obtemos uma Matriz de Scores (Pesos) que diz para a palavra "banco" prestar 90% de atenção na palavra "dinheiro" e não na palavra "praça" na frase atual!

```python
import torch
import torch.nn.functional as F

def attention_bruta(Q, K, V):
    dk = Q.shape[-1]
    # 1. Multiplica o "Que estou procurando" pelo "Que as palavras oferecem"
    scores = torch.matmul(Q, K.transpose(-2, -1)) / (dk ** 0.5)
    
    # 2. Converte em Porcentagem de 0 a 1 (Ex: Preste 80% de atenção ali)
    pesos = F.softmax(scores, dim=-1)
    
    # 3. Mistura os valores reais com os pesos gerados
    return torch.matmul(pesos, V)

print("Motor de Atenção Ativado!")
```
