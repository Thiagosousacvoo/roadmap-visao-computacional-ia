## 🎯 Objetivos deste notebook

O Mecanismo de Atenção não vive sozinho. Ele é apenas o "filtro". Ele precisa estar dentro de um bloco completo chamado *Transformer Encoder Block*.

---

## Anatomia do Bloco
1. O texto entra no `Multi-Head Attention`.
2. A saída ganha um atalho de segurança (`Residual Connection` - A mesma mágica da ResNet!).
3. Passa pela `Layer Normalization`.
4. Entra numa Rede Neural normal (o famoso MLP `Feed-Forward`).
5. Ganha outra `Residual Connection`.

## E as Imagens?
O **Self-Attention** é quando o texto julga o próprio texto. Mas quando chegarmos no mundo Visual, veremos o **Cross-Attention**, onde o TEXTO pergunta e a IMAGEM responde! Isso cria a Geração de Imagem por IA.
