## 🎯 Objetivos deste notebook

Se o modelo de Texto cospe Vetores... E o modelo de Imagem (ViT) cospe Vetores... Será que não podemos ensinar os dois modelos a cuspir o **MESMO** vetor para coisas iguais? É isso que o CLIP (Contrastive Language-Image Pretraining) faz.

---

## Contrastive Learning
A OpenAI pegou 400 Milhões de Imagens da internet junto com os textos Alt. 
Ela treinou 2 Redes ao mesmo tempo. Se a imagem é de um Gato, e o texto é "Um gato", as duas redes são forçadas a criar vetores que têm 99% de similaridade.

Resultado? O CLIP pode classificar QUALQUER imagem do mundo em qualquer classe de texto **sem precisar de treinamento prévio** (Zero-Shot). E ele é o motor invisível por trás do Midjourney e Stable Diffusion!
