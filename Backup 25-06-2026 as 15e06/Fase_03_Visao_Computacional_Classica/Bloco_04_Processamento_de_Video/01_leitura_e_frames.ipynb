## 🎯 Objetivos deste notebook

Um vídeo é apenas um Loop `while True` disparando 30 matrizes (frames) na tela por segundo.

---

```python
import cv2

# Como o código funciona (NÃO RODE AQUI, Notebooks travam com video loops)
# Salve num script .py para testar!

texto_codigo = """
import cv2

cap = cv2.VideoCapture(0) # 0 é a sua Webcam

while True:
    ret, frame = cap.read()
    if not ret: break
    
    # Transformar a sua webcam em cinza em TEMPO REAL!
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow("Minha Webcam", gray)
    
    # Se apertar 'q', sai do loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
"""
print(texto_codigo)
```
