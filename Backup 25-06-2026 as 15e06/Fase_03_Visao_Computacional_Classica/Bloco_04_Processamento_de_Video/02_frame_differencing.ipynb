## 🎯 Objetivos deste notebook

Como uma câmera de segurança barata sabe que tem um ladrão se movendo? Ela não usa Inteligência Artificial. Ela usa matemática básica: Subtração de Matrizes!

---

## A Lógica:
1. Salve o Frame 1 (A sala vazia).
2. No Frame 2 (O ladrão entrou), faça `cv2.absdiff(Frame1, Frame2)`.
3. Todos os pixels que são a "parede" vão dar `Parede - Parede = Zero (Preto)`.
4. Onde o ladrão está, a matemática dará `Parede - Ladrão = Diferença (Branco)`.
5. O ladrão literalmente acende na tela!
