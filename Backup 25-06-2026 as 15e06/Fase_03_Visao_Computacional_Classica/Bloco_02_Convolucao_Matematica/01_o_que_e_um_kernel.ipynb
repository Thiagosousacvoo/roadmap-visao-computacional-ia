## 🎯 Objetivos deste notebook

Um Kernel (ou Filtro) é uma janelinha matriz (ex: 3x3) que desliza em cima de uma imagem gigante, multiplicando os pixels e somando tudo para cuspir UM ÚNICO NOVO PIXEL.

---

## 1️⃣ O Kernel da Média (Blur)

Se o seu Kernel for uma matriz 3x3 cheia de "1", dividida por 9. O que ela faz?
Ela calcula a média aritmética de você e dos seus 8 vizinhos ao redor. E adivinha o que acontece quando todos os pixels viram a "média" dos seus vizinhos? A imagem fica EMBAÇADA!

```python
import numpy as np

kernel_blur = np.ones((3, 3)) / 9
print("Filtro de Desfoque (Blur):")
print(kernel_blur)
```
