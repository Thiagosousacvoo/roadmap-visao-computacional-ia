Agora que o nosso motor está pronto, vamos testá-lo com matrizes famosas da engenharia!

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d  # Usaremos a versão rápida do SciPy no lugar dos nossos For Loops demorados

# Imagem de uma borda vertical simples
img = np.zeros((20, 20), dtype=np.uint8)
img[:, 10:] = 255

# O FILTRO SOBEL X (Detecta Pulos Verticais)
# Perceba: O Lado esquerdo é negativo, o direito é positivo. 
# Se a imagem for igual dos dois lados, a soma vai dar Zero (Preto).
# Se houver um "Degrau" de cor, a soma vai disparar!
sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

resultado = convolve2d(img, sobel_x, mode="same")

plt.subplot(1, 2, 1)
plt.title("Imagem Original")
plt.imshow(img, cmap="gray")

plt.subplot(1, 2, 2)
plt.title("Bordas (Sobel X)")
plt.imshow(np.abs(resultado), cmap="gray")
plt.show()
```
