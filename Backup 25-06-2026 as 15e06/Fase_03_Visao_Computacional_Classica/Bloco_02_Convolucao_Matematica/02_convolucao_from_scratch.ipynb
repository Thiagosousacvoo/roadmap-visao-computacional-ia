## 🎯 Objetivos deste notebook

Esta é a etapa em que o menino vira homem na Visão Computacional. Nós escreveremos o motor de varredura (Convolução) puro em Python. Se você entende esse loop, você entende CNNs (YOLO, ResNet, etc).

---

```python
import numpy as np
import matplotlib.pyplot as plt

def convolucao_manual(imagem, kernel):
    h_img, w_img = imagem.shape
    h_k, w_k = kernel.shape
    
    # Offset para centralizar a janelinha (Padding)
    pad_h = h_k // 2
    pad_w = w_k // 2
    
    # Imagem vazia para os resultados
    resultado = np.zeros_like(imagem, dtype=np.float32)
    
    # Adicionando uma borda preta na imagem original para o filtro não vazar pra fora da tela
    img_padded = np.pad(imagem, ((pad_h, pad_h), (pad_w, pad_w)), mode="constant")
    
    # O GRANDE LOOP DE VARREDURA
    for i in range(h_img):
        for j in range(w_img):
            # Recorta a Vizinhança Exata do tamanho do Kernel
            vizinhanca = img_padded[i : i+h_k, j : j+w_k]
            # Multiplica matrizes elemento a elemento e SOMA TUDO num escalar.
            novo_pixel = np.sum(vizinhanca * kernel)
            resultado[i, j] = novo_pixel
            
    return np.clip(resultado, 0, 255).astype(np.uint8)

print("Engine Convolucional compilada!")
```
