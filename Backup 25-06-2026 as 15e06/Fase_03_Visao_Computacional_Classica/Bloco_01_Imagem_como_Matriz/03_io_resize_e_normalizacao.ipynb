## 🎯 Objetivos deste notebook

Redes Neurais odeiam números grandes (como 255). Elas preferem números entre `0.0` e `1.0`. Aprender a normalizar corretamente a imagem é vital para o Deep Learning.

---

## 1️⃣ Escala e Resize

Você NUNCA passará uma imagem 4K (3840x2160) para uma IA. Vai explodir sua placa de vídeo. Reduzimos para tamanhos de 224x224 ou 640x640.

```python
import cv2
import numpy as np

img = np.zeros((1080, 1920, 3), dtype=np.uint8)

# Forçando para um tamanho fixo
img_pequena = cv2.resize(img, (224, 224)) # Note que no OpenCV, passamos (Largura, Altura) e NÃO (Linha, Coluna)!
print(f"Reduzida: {img_pequena.shape}")

# 2️⃣ Normalização
img_float = img_pequena.astype(np.float32) / 255.0
print(f"O formato agora é Float32, indo de 0 a 1! Max: {img_float.max()}")
```
