## 1️⃣ A História do BGR

O OpenCV foi criado nos anos 90. Naquela época, os monitores liam as cores na ordem **Blue-Green-Red (BGR)**. Hoje, o padrão de internet é **RGB**.
Mas o OpenCV NUNCA alterou sua API. Então, quando ele carrega uma foto, ele a carrega de trás pra frente!

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Criando uma imagem puramente VERMELHA no padrão Matemático (RGB)
img_vermelha = np.zeros((100, 100, 3), dtype=np.uint8)
img_vermelha[:, :, 0] = 255 # R=255, G=0, B=0

plt.subplot(1, 2, 1)
plt.title("Como o Matplotlib Lê (RGB)")
plt.imshow(img_vermelha)

# Se salvarmos essa imagem com o OpenCV, ele vai achar que o primeiro canal é o AZUL!
# Para consertar, precisamos Inverter!
img_corrigida = cv2.cvtColor(img_vermelha, cv2.COLOR_RGB2BGR)

plt.subplot(1, 2, 2)
plt.title("O que o OpenCV acha que é")
plt.imshow(img_corrigida)
plt.show()
```

## 2️⃣ Slicing (Cortando a foto com Matemática)

Para recortar a imagem, basta dizer: Quais linhas eu quero? Quais colunas eu quero? (E mantenha todos os 3 canais).

```python
# Crop: [Linha_Inicio : Linha_Fim , Coluna_Inicio : Coluna_Fim]
crop = img_vermelha[20:80, 20:80]
print(f"Original: {img_vermelha.shape} | Recortada: {crop.shape}")
```
