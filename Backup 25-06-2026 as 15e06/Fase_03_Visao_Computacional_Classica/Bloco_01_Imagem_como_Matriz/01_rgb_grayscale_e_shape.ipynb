## 1️⃣ Gerando nossa Própria Imagem

Em vez de carregar um `.jpg`, vamos criar nossa imagem puramente preenchendo números no NumPy!

```python
import numpy as np
import matplotlib.pyplot as plt

# Criando uma matriz de Zeros com formato: Altura 200, Largura 300, Canais 3 (RGB)
# Tipo uint8 (Unsigned Integer de 8 bits) = Só aceita números inteiros entre 0 e 255.
img_fake = np.zeros((200, 300, 3), dtype=np.uint8)

# Pintando um retângulo Vermelho!
# O canal 0 é o RED no mundo matplotlib.
img_fake[50:150, 50:150, 0] = 255  

# Pintando um retângulo Verde!
img_fake[80:180, 120:250, 1] = 255  

plt.imshow(img_fake)
plt.title(f"A Imagem é uma matriz: {img_fake.shape}")
plt.axis("off")
plt.show()

print("O computador não viu cores, ele apenas montou esse array gigante:", img_fake.shape)
```
