## 🎯 Objetivos deste notebook

Tentar detectar uma cor no formato RGB é um pesadelo. Se o sol bater na maçã vermelha, o vermelho (`R=255`) acaba ficando branco (`R=255, G=255, B=255`)!
O espaço **HSV** (Hue, Saturation, Value) separa a **Luz** da **Cor**.

---

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Imagine que você tirou uma foto e tem uma bolinha Laranja.
# H=15 (Laranja), S=255 (Muito viva), V=255 (Muito brilhante)
bolinha = np.full((100, 100, 3), [15, 255, 255], dtype=np.uint8)

# Se a iluminação escurecer muito, o "V" cai pra 100. Mas a matriz "H" AINDA CONTINUA SENDO 15 (Laranja)!
# É por isso que no espaço HSV é muito fácil criar um filtro Mágico: "Pegue tudo que o H está entre 10 e 20" e a luz não importa!

lower_orange = np.array([10, 50, 50])
upper_orange = np.array([25, 255, 255])

mascara = cv2.inRange(bolinha, lower_orange, upper_orange)
print("Mascara criada com sucesso, detectando pixels na faixa laranja do HSV!")
```
