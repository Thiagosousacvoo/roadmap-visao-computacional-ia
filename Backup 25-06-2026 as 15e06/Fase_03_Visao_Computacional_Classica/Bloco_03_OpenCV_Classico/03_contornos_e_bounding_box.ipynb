## 🎯 Objetivos deste notebook

Uma vez que a imagem é apenas Preto e Branco (Binarizada), o OpenCV pode encontrar todas as "ilhas brancas" e nos devolver as coordenadas `(x, y)` delas!

---

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((300, 300), dtype=np.uint8)
cv2.rectangle(img, (50, 50), (100, 100), 255, -1) # Quadrado pequeno
cv2.circle(img, (200, 200), 40, 255, -1)          # Circulo grande

# Achando contornos
contours, _ = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Criando imagem colorida para desenhar os quadrados vermelhos em cima
img_color = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

for c in contours:
    # Calculando a Bounding Box (A Caixa que envolve o objeto)
    x, y, w, h = cv2.boundingRect(c)
    cv2.rectangle(img_color, (x, y), (x+w, y+h), (255, 0, 0), 3) # Desenhando a caixa em VERMELHO
    
plt.imshow(img_color)
plt.title("Objetos rastreados com Bounding Boxes!")
plt.show()
```
