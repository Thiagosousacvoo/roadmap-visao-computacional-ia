## 🎯 Objetivos deste notebook

Aplicar aquelas matrizes de transformação da Fase 01 em imagens reais usando `warpAffine`.

---

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((200, 200), dtype=np.uint8)
cv2.rectangle(img, (50, 50), (150, 100), 255, -1)

# Matriz de Rotacao de 45 graus, ancorada no centro da imagem
centro = (100, 100)
matriz_rotacao = cv2.getRotationMatrix2D(centro, 45, escala=1.0)

rotacionada = cv2.warpAffine(img, matriz_rotacao, (200, 200))

plt.imshow(rotacionada, cmap="gray")
plt.title("Retângulo girado 45 Graus!")
plt.show()
```
