## 🎯 Objetivos deste notebook

Separar o objeto do fundo (background). 
Se um pixel for maior que X, vira 255 (Branco). Se for menor, vira 0 (Preto).

---

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Criando um Gradiente (Sombra indo do Preto ao Branco)
gradiente = np.linspace(0, 255, 300)
img = np.tile(gradiente, (300, 1)).astype(np.uint8)

# Binarizando com corte fixo em 127
_, binario = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.subplot(1, 2, 1)
plt.title("Imagem Gradiente Original")
plt.imshow(img, cmap="gray")

plt.subplot(1, 2, 2)
plt.title("Após cv2.threshold(127)")
plt.imshow(binario, cmap="gray")
plt.show()
```
