## 🎯 Objetivos deste notebook

No bloco anterior nós montamos o Sobel com Numpy. Agora usaremos a versão ultra rápida em C++ que vem no OpenCV, além do algoritmo Canny (O Rei das Bordas).

---

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Criando uma imagem de teste (Um circulo branco em fundo preto)
img = np.zeros((300, 300), dtype=np.uint8)
cv2.circle(img, (150, 150), 80, 255, -1) 

# Sobel
grad_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
grad_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

# Canny (Usa Sobel por baixo dos panos, mas é muito mais inteligente)
bordas_canny = cv2.Canny(img, 50, 150)

plt.subplot(1, 2, 1)
plt.title("Gradiente Y (Sobel)")
plt.imshow(grad_y, cmap="gray")

plt.subplot(1, 2, 2)
plt.title("Canny Inteligente")
plt.imshow(bordas_canny, cmap="gray")
plt.show()
```
