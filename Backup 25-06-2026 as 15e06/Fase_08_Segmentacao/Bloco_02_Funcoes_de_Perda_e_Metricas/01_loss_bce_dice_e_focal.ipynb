## 🎯 Objetivos deste notebook

Em segmentação médica, o câncer ocupa apenas 2% da chapa de raio-x. Se a sua IA chutar que "A foto inteira é saudável", a acurácia dela será de 98%.
Você precisa de Loss Functions que punam a IA baseada na Interseção da máscara, não nos pixels absolutos.

---

```python
import torch

def dice_loss(predicao, mascara_real, smooth=1.0):
    # 1. Aplica o Sigmoid para ter a probabilidade do pixel (0 a 1)
    pred = torch.sigmoid(predicao).view(-1)
    real = mascara_real.view(-1)
    
    # 2. Quanto eles se cruzam no mesmo exato pixel?
    intersecao = (pred * real).sum()
    
    # 3. Equação de Dice (F1-Score para áreas)
    dice = (2. * intersecao + smooth) / (pred.sum() + real.sum() + smooth)
    
    # Como Loss sempre tem que cair, retornamos 1 - Dice
    return 1 - dice

print("Dice Loss carregada. Se a IA errar o pixel minúsculo do tumor, a Loss vai explodir e ela vai ser punida!")
```
