## Dice vs IoU
Ambas avaliam a qualidade da máscara gerada comparada à máscara desenhada pelo humano.
- **Dice Score:** Penaliza um pouco menos os pequenos erros na borda.
- **IoU (Jaccard Index):** É mais rigoroso.

Na área Médica, o mundo acadêmico é apaixonado pelo Dice. Em self-driving cars, o IoU dita as regras.
