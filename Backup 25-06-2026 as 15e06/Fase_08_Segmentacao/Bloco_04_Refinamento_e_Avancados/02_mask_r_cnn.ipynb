## 🎯 Objetivos deste notebook

A U-Net falha se você precisa separar "Carro 1" de "Carro 2" que estão colados no trânsito. Para Segmentação de Instância, usamos a monstruosa **Mask R-CNN**.

Ela é literalmente uma rede de Object Detection (Bounding Boxes) que roda uma micro rede de Segmentação apenas DENTRO de cada caixinha encontrada.
