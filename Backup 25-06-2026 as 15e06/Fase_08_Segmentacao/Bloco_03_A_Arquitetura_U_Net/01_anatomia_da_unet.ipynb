## 🎯 Objetivos deste notebook

Por que a rede se chama U-Net? Porque ela esmaga a imagem descendo até um gargalo minúsculo, e depois expande ela de volta subindo para o tamanho original!

---

## O Problema do "Borrão"
Quando a rede chega no fundo do "U", ela sabe "O QUE" está na imagem, mas perdeu os detalhes finos das bordas porque a resolução ficou minúscula (ex: 8x8 pixels). Ao tentar expandir de volta para 256x256, a máscara fica borrada.

## A Genialidade: Skip Connections
Para não deixar borrar, a U-Net puxa um "fio" do lado esquerdo (Alta Resolução) e emenda com o lado direito (Expansão). Assim, o Decoder ganha dicas de onde estão as bordas exatas da imagem original!
