## 🎯 Objetivos deste notebook

Qual é a diferença entre pintar todas as pessoas de Vermelho, ou pintar o João de Vermelho e a Maria de Azul?

---

## 1️⃣ Semântica
Todos os pixels de uma classe recebem a mesma cor. Se tiverem 5 carros colados na rua, vira uma grande "massa" de pixels azuis. O modelo não sabe onde acaba um carro e começa o outro.

## 2️⃣ Instância
O modelo foca nos objetos independentes. Ele recorta o Carro 1 perfeitamente, e o Carro 2 perfeitamente. Porém ele ignora coisas "amorfas" como o céu, a estrada e a grama.

## 3️⃣ Panóptica
O Santo Graal. Ele faz a Semântica do Céu e da Estrada, e ao mesmo tempo identifica o Carro 1 diferente do Carro 2 no meio dessa estrada.
