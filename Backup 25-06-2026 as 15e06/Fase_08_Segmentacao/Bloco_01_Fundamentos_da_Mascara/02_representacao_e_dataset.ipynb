## 🎯 Objetivos deste notebook

Como o rótulo de segmentação é armazenado? Ele não é um `.txt`. Ele é literalmente outra IMAGEM (geralmente PNG), onde a cor do pixel é o número da classe.

---

```python
import cv2
import numpy as np

print("Imagine uma imagem 5x5. A máscara seria assim:")
mascara = np.zeros((5, 5), dtype=np.uint8)
mascara[2:4, 2:4] = 1 # A classe 1 (ex: Cachorro) ocupa esse quadrado
print(mascara)
```

> ⚠️ **A ARMADILHA DE OURO:**
> Se você for rotacionar a Imagem Original em 15 graus para treinar a rede (Data Augmentation), você **TEM QUE** rotacionar a Máscara PNG em exatos 15 graus também. Senão, o rótulo vai apontar para o lugar errado e destruir a sua Rede Neural!
