## 1️⃣ Como Autovetores comprimem dados?

1. Temos uma nuvem de pontos que parece um charuto inclinado.
2. Se tirarmos a **Matriz de Covariância** desses dados, ela mostra como X se correlaciona com Y.
3. Os **Autovetores** dessa Matriz de Covariância apontarão exatemente na direção onde os dados são mais "espalhados" (maior variância).
4. Se nós pegarmos apenas os Autovetores com os maiores Autovalores, teremos os "Eixos Principais". Podemos girar nosso gráfico para esses eixos e jogar os eixos fracos fora!
