## 🎯 Objetivos deste notebook

Um **Autovetor** (Eigenvector) de uma Matriz A é um vetor que, quando multiplicado por A, apenas sofre uma escala matemática, mas NÂO muda a sua direção da reta. O fator numérico que causou essa escala é o **Autovalor** (Eigenvalue).

---

## 1️⃣ Calculando no Numpy (`np.linalg.eig`)

```python
import numpy as np

# Matriz de cisalhamento (Empurra tudo em diagonal)
A = np.array([
    [2, 1],
    [0, 2]
])

# autovalores, autovetores
eigen_vals, eigen_vecs = np.linalg.eig(A)

print("Autovalores (Lambdas):", eigen_vals)
print("\nAutovetores (Cada COLUNA é um vetor!):\n", eigen_vecs)
```

## 2️⃣ Comprovando a Equação `A @ v = lambda * v`

```python
v1 = eigen_vecs[:, 0]  # Pegando a primeira coluna (Primeiro autovetor)
lambda_1 = eigen_vals[0] # Pegando o respectivo autovalor

print(f"Fazendo: Matriz @ Autovetor = {A @ v1}")
print(f"Fazendo: Lambda * Autovetor = {lambda_1 * v1}")
print("\nVeja que o resultado é absolutamente idêntico!")
```
