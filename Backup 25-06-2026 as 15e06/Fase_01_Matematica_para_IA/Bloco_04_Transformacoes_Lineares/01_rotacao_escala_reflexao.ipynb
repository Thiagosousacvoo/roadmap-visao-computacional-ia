## 1️⃣ Matriz de Rotação

A fórmula para girar um ponto `(x,y)` em um ângulo θ é aplicar a matriz:
```
[cos(θ)  -sin(θ)]
[sin(θ)   cos(θ)]
```

````python
import numpy as np
import matplotlib.pyplot as plt

# Nosso ponto inicial: x=1, y=0
ponto = np.array([1, 0])

# Criando Matriz de Rotação de 90 Graus (pi/2 radianos)
theta = np.pi / 2
matriz_rotacao = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

# Aplicando!
ponto_girado = matriz_rotacao @ ponto

# Note que o computador pode retornar algo como 6.12e-17 ao invés de exato zero. 
print(f"Original: {ponto}")
print(f"Rotacionado 90 graus: {np.round(ponto_girado, 2)}")
````
