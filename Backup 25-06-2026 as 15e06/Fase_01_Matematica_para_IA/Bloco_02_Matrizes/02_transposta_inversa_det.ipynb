## 1️⃣ Transposta

Gira a matriz sobre a sua diagonal. A primeira linha vira a primeira coluna. Essencial para forçar dimensões de matrizes a baterem no Produto Escalar.

```python
import numpy as np

A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Original Shape:", A.shape)
print(A)

print("\nTransposta Shape:", A.T.shape)
print(A.T)
```

## 2️⃣ Determinante

O determinante é um número escalar que nos diz o quanto essa matriz "escala" o espaço ao redor. 
Se o det = 0, significa que a matriz colapsa o espaço inteiro em uma linha ou ponto! (Perda irreversível de dados).

```python
M1 = np.array([
    [2, 0],
    [0, 2]
]) # Multiplica tudo por 2 (Dobra a área 2x2 = fator 4)
print(f"Det M1 = {np.linalg.det(M1)}")

M2 = np.array([
    [1, 2],
    [1, 2]
]) # Linhas lineares idênticas (ou dependentes). Colapsa tudo.
print(f"Det M2 = {np.linalg.det(M2)}")
```

## 3️⃣ A Matriz Inversa

Se uma matriz `A` aplica uma transformação em um vetor `x` (`A @ x = y`), a matriz Inversa `A_inv` consegue pegar `y` e transformar de volta em `x`! (`A_inv @ y = x`).

> ⚠️ **Fato vital:** Matrizes com Determinante = 0 **NÃO POSSUEM INVERSA**. Uma vez colapsada a dimensão, a informação não volta.

```python
A = np.array([
    [4, 7],
    [2, 6]
])

x = np.array([10, 20])
y_transformado = A @ x

print(f"Vetor Original   : {x}")
print(f"Após A@x         : {y_transformado}")

# Calculando a Inversa
A_inv = np.linalg.inv(A)

x_recuperado = A_inv @ y_transformado
print(f"Recuperado Invers: {np.round(x_recuperado).astype(int)}")
```
