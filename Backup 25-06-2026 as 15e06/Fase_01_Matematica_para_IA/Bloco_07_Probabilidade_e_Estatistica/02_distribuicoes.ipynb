## 🎯 Objetivos deste notebook

Toda inicialização de rede neural é feita tirando números aleatórios de uma "Distribuição Normal". Todo classificador converte saídas numa probabilidade "Softmax" ou "Bernoulli".

---

## 1️⃣ A Curva de Gauss (Distribuição Normal)

O famoso formato de "Sino". Encontrado na natureza em todo lugar: altura das pessoas, pontuações em provas, e erros de medição.

```python
import numpy as np
import matplotlib.pyplot as plt

# Tirando 10.000 amostras com Média=0, Desvio=1
amostras = np.random.normal(loc=0.0, scale=1.0, size=10000)

plt.hist(amostras, bins=50, density=True, color='skyblue', edgecolor='black')
plt.title("Distribuição Normal (Sino)")
plt.show()
```

## 2️⃣ Distribuição Uniforme

Todos os números dentro de um intervalo têm EXATAMENTE a mesma chance de sair. O `np.random.rand` usa isso!

```python
uniforme = np.random.uniform(low=-1.0, high=1.0, size=10000)
plt.hist(uniforme, bins=50, density=True, color='lightgreen', edgecolor='black')
plt.title("Distribuição Uniforme")
plt.show()
```
