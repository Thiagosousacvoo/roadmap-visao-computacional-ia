## 🎯 Objetivos deste notebook

A Entropia é a base da principal função de Loss de classificação do mundo: a **Cross-Entropy Loss**.

---

## 1️⃣ Entropia: O que é?

Entropia mede o nível de **Incerteza**.
- Uma moeda justa (50/50) tem Entropia MÁXIMA (Maior incerteza do universo, impossível prever).
- Uma moeda viciada de dois lados de "Cara" (100/0) tem Entropia ZERO (Nenhuma incerteza, eu já sei o que vai cair).

```python
import numpy as np

def entropia(probs):
    # Adicionamos um pequeno epsilon para evitar log(0) que quebra a matemática
    eps = 1e-9
    return -np.sum(probs * np.log2(probs + eps))

moeda_justa = np.array([0.5, 0.5])
moeda_viciada = np.array([0.99, 0.01])

print(f"Entropia Moeda Justa  : {entropia(moeda_justa):.3f} (Máxima)")
print(f"Entropia Moeda Viciada: {entropia(moeda_viciada):.3f} (Baixa)")
```
