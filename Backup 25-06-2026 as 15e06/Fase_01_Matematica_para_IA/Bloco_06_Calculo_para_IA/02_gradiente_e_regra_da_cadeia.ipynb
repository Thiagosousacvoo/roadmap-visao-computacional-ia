## 🎯 Objetivos deste notebook

Redes neurais não têm apenas 1 variável `x`. O GPT-4 tem trilhões de pesos. Precisamos de uma derivada para CADA UM deles.
- O **Gradiente** é apenas um vetor contendo todas as derivadas parciais.
- A **Regra da Cadeia** permite calcular o erro através de dezenas de camadas empilhadas.

---

## 1️⃣ O Vetor Gradiente

Se a nossa função de erro depende de 2 pesos `w1` e `w2`, o Gradiente será um vetor `[dw1, dw2]`.
Ele aponta para a direção onde o erro **CRESCE MAIS RÁPIDO**. Portanto, a IA dá um passo na direção OPOSTA do Gradiente (Descida do Gradiente).

## 2️⃣ A Regra da Cadeia (Backpropagation)

Em uma Rede Neural, temos `Entrada -> Camada 1 -> Camada 2 -> Erro`.
Isso é matematicamente `Erro( Camada2( Camada1( Entrada ) ) )`.

Como sabemos o quanto a Camada 1 afetou o Erro Final? Multiplicando as derivadas de trás pra frente!

`dErro/dCamada1 = dErro/dCamada2 * dCamada2/dCamada1`
