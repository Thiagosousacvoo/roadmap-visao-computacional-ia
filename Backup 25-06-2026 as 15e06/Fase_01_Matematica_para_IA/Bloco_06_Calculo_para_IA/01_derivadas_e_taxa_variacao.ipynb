## 1️⃣ O que é a Derivada?

A derivada de uma função `f(x)` nos diz: "Se eu aumentar `x` num pentelhésimo (quase nada), a função `f(x)` sobe ou desce? E com qual força?"
- Derivada positiva: A função está subindo.
- Derivada negativa: A função está caindo.
- Derivada zero: Chegamos no fundo de um vale ou no topo de uma montanha (O objetivo da IA é chegar no fundo do vale do Erro!).

```python
import numpy as np
import matplotlib.pyplot as plt

# Função de Custo Fictícia: f(x) = x^2
def f(x):
    return x**2

# Derivada analítica: f'(x) = 2x
def df(x):
    return 2*x

x = np.linspace(-5, 5, 100)
y = f(x)

plt.plot(x, y, label="Erro f(x) = x²")

# Analisando o ponto x = -3
ponto_x = -3
ponto_y = f(ponto_x)
inclinacao = df(ponto_x)

# Reta tangente y = mx + b
b = ponto_y - inclinacao * ponto_x
reta_y = inclinacao * x + b

plt.scatter([ponto_x], [ponto_y], color='red', zorder=5)
plt.plot(x, reta_y, color='red', linestyle='--', label=f"Derivada (inclinação) = {inclinacao}")
plt.ylim(-5, 30)
plt.legend()
plt.show()

print("A derivada é -6. Isso significa que para CADA passo positivo no X, a função cai 6 passos! Ou seja, devemos andar pra FRENTE para descer!")
```
