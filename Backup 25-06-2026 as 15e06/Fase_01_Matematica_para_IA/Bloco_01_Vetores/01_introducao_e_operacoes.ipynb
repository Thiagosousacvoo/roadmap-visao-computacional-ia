## 1️⃣ Representação de um Vetor

Em Python puro, um vetor é apenas uma lista de números. Em NumPy, é um array otimizado em C.

```python
import numpy as np
import matplotlib.pyplot as plt

# Criando vetores
v_python = [3, 4]
v_numpy = np.array([3, 4])

print("Lista normal:", v_python)
print("Array Numpy :", v_numpy)
print(f"Dimensão (Shape): {v_numpy.shape} -> (Tem 2 elementos, logo, vive em 2D)")
```

## 2️⃣ Operações Escalares

O que acontece se multiplicarmos um vetor por um número escalar? O vetor "estica", encolhe, ou inverte a direção.

```python
v = np.array([2, 1])

print("Original:    ", v)
print("Dobro (2*v): ", 2 * v)
print("Metade:      ", 0.5 * v)
print("Invertido:   ", -1 * v)

# Vamos visualizar!
plt.figure(figsize=(4,4))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='v original')
plt.quiver(0, 0, (2*v)[0], (2*v)[1], angles='xy', scale_units='xy', scale=1, color='r', alpha=0.5, label='2*v')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.legend()
plt.show()
```

## 3️⃣ Soma de Vetores

Somar dois vetores é matematicamente somar cada índice (x com x, y com y). Visualmente, é colocar a cauda de uma seta na ponta da outra.

```python
a = np.array([1, 3])
b = np.array([2, -1])

resultado = a + b
print(f"A: {a} | B: {b} | A+B: {resultado}")

# ✏️ EXERCÍCIO: Calcule o vetor que liga o ponto A(10, 5) ao ponto B(15, 2).
# Dica: O vetor deslocamento é (B - A).

# Seu código aqui
# vetor_deslocamento = ...
```
