## 🎯 Objetivos deste notebook

Como saber se a frase "Cachorro late" é parecida com "Cão ladra"? Usamos IA para transformá-las em vetores, e depois usamos o **Cosseno** para ver o ângulo entre esses vetores!

---

## 1️⃣ O Problema da Distância Euclidiana

Se usarmos a distância euclidiana para medir a diferença entre vetores, um vetor `[1, 1]` será considerado "muito distante" do vetor `[100, 100]`. Mas perceba: eles têm exatamente a mesma direção (a mesma intenção de sentido). 
O ângulo entre eles não considera a magnitude (comprimento).

## 2️⃣ A Fórmula Mágica

$$ \text{cosine\_similarity}(A, B) = \frac{A \cdot B}{||A|| \times ||B||} $$

Basicamente: O dot product, dividido pelo produto de suas magnitudes. O resultado sempre estará entre `-1` e `1`.

```python
import numpy as np

def similaridade_cosseno(a, b):
    dot_prod = np.dot(a, b)
    norma_a = np.linalg.norm(a)
    norma_b = np.linalg.norm(b)
    return dot_prod / (norma_a * norma_b)

# Testando com o exemplo acima
v1 = np.array([1, 1])
v2 = np.array([100, 100])

# Distância enorme, mas o ângulo...
print(f"Similaridade de Cosseno: {similaridade_cosseno(v1, v2)}") # Resultado: 1.0 (Idênticos!)
```

## 3️⃣ Aplicação em Sistemas de Recomendação

Imagine que as notas dadas por 3 usuários para os filmes `[Ação, Comédia, Terror]` sejam representadas por vetores.

```python
user_A = np.array([5, 0, 1])  # Gosta muito de ação
user_B = np.array([4, 1, 0])  # Gosta de ação também
user_C = np.array([0, 5, 4])  # Odeia ação, prefere comédia

print(f"Sim(A, B): {similaridade_cosseno(user_A, user_B):.3f} -> Perfis Parecidos!")
print(f"Sim(A, C): {similaridade_cosseno(user_A, user_C):.3f} -> Perfis Opostos!")
```
