## 1️⃣ Configurando o Logging

Ao usar logging, informamos a gravidade da mensagem. Assim, no servidor, podemos filtrar para ver apenas os "ERROS".

```python
import logging

# Configuração básica para escrever tanto no console quanto num arquivo
logging.basicConfig(
    level=logging.INFO, # Mostra apenas de INFO para cima (ignora DEBUG)
    format="%(asctime)s [%(levelname)s] %(message)s",
    handlers=[
        logging.StreamHandler(), # Joga na tela
        # logging.FileHandler("app.log") # Descomente para salvar num arquivo de verdade!
    ],
    force=True  # Força sobrescrever configs anteriores do jupyter
)

# Usando os níveis
logging.debug("Detalhe inútil (não vai aparecer porque configuramos para INFO)")
logging.info("O modelo concluiu a época 1/10.")
logging.warning("O uso de CPU está passando de 90%!")
logging.error("Falha ao salvar pesos na nuvem. HD Cheio.")
logging.critical("O SISTEMA VAI CRASHAR!")
```
