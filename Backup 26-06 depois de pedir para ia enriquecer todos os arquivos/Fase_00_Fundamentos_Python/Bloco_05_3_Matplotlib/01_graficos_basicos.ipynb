## 🎯 Objetivos deste notebook

Uma imagem vale mais que mil números. O Matplotlib (`plt`) é a engine de plotagem gráfica clássica do Python.
- Plot de Linha (Tendências temporais).
- Plot de Dispersão / Scatter (Correlações).
- Gráfico de Barras.

---

## 1️⃣ O Gráfico de Linhas Clássico

```python
import matplotlib.pyplot as plt
import numpy as np

# Épocas de treinamento vs Acurácia
epocas = np.arange(1, 11)
acc_treino = [0.4, 0.55, 0.70, 0.82, 0.88, 0.91, 0.94, 0.95, 0.96, 0.97]
acc_validacao = [0.38, 0.50, 0.65, 0.75, 0.80, 0.82, 0.85, 0.86, 0.86, 0.85]  # Overfitting no final!

plt.figure(figsize=(8, 4)) # Largura, Altura em polegadas

# Plotando as linhas com label para a legenda
plt.plot(epocas, acc_treino, marker='o', linestyle='-', color='blue', label='Treino')
plt.plot(epocas, acc_validacao, marker='s', linestyle='--', color='red', label='Validação')

plt.title("Acurácia do Modelo durante Treinamento")
plt.xlabel("Épocas")
plt.ylabel("Acurácia")
plt.legend()
plt.grid(True)

plt.show()
```

## 2️⃣ Gráficos de Dispersão (Scatter Plots)

Essenciais na Análise Exploratória para ver se dois eixos possuem correlação linear (ex: Tamanho da Casa vs Preço).

```python
tamanho_m2 = np.random.randint(40, 200, 50)
preco_milhares = tamanho_m2 * 3.5 + np.random.normal(0, 50, 50) # Relação linear + Ruído

plt.figure(figsize=(6, 4))
plt.scatter(tamanho_m2, preco_milhares, alpha=0.7, color="purple")
plt.title("Tamanho vs Preço do Imóvel")
plt.xlabel("Tamanho (m²)")
plt.ylabel("Preço (R$ 1000s)")
plt.show()
```
