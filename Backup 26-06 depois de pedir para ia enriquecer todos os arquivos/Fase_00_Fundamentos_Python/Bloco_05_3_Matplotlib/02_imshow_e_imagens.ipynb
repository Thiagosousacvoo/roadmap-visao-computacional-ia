## 🎯 Objetivos deste notebook

Para quem estuda **Visão Computacional**, o Matplotlib é a tela do nosso microscópio. O comando `imshow` pega um Tensor/Array e o colore na tela.

```python
import matplotlib.pyplot as plt
import numpy as np

# Vamos criar uma imagem totalmente preta (zeros) de 10x10 pixels em escala de cinza
img_pb = np.zeros((10, 10))

# Vamos desenhar uma "cruz" branca nela (Pixel branco = 255)
img_pb[:, 4:6] = 255  # Colunas 4 e 5 inteiras brancas
img_pb[4:6, :] = 255  # Linhas 4 e 5 inteiras brancas

plt.figure(figsize=(4, 4))
# cmap="gray" é ESSENCIAL para arrays 2D, senão ele tenta pintar com o mapa de calor bizarro padrão (viridis)
plt.imshow(img_pb, cmap="gray")
plt.title("A Cruz de Zeros e 255s")
plt.axis("off")  # Desliga as réguas de gráfico laterais
plt.show()
```

## 2️⃣ Imagens Coloridas (RGB)

Imagens RGB precisam de 3 canais no Numpy (shape H, W, 3) com dados entre 0 e 1 (floats) ou 0 e 255 (inteiros uint8).

```python
img_cor = np.zeros((10, 10, 3), dtype=np.uint8)

# Pintando o topo de Vermelho (R=255, G=0, B=0)
img_cor[0:3, :, 0] = 255

# Pintando o meio de Verde (R=0, G=255, B=0)
img_cor[3:7, :, 1] = 255

# Pintando o fundo de Azul (R=0, G=0, B=255)
img_cor[7:, :, 2] = 255

plt.imshow(img_cor)
plt.title("Bandeira RGB Numérica")
plt.show()
```
