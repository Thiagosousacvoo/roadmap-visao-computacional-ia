## 🎯 Objetivos deste notebook

O ditado em IA é: *Garbage In, Garbage Out*. Dados reais sempre vêm sujos, com células vazias e duplicadas.
- Lidar com dados nulos (NaN).
- Usar `dropna()` e `fillna()`.
- Remover duplicações.

---

## 1️⃣ Detectando Dados Nulos (NaN - Not a Number)

O Pandas usa o NaN do NumPy para representar a ausência de dados.

```python
import pandas as pd
import numpy as np

dados_sujos = {
    "Nome": ["Ana", "Bruno", "Caio", "Diana", "Ana", "Caio"],
    "Idade": [25, np.nan, 19, 31, 25, 19],
    "Salario": [4000, 5000, np.nan, 3000, 4000, np.nan]
}
df = pd.DataFrame(dados_sujos)
display(df)

# Contar quantos vazios temos por coluna
print("\nTotal de Nulos por Coluna:\n", df.isnull().sum())
```

## 2️⃣ Lidando com Nulos: Dropar ou Preencher?

Dependendo da modelagem de dados, você descarta a linha com defeito ou preenche com uma média/mediana (Imputação).

```python
print("Opção 1: Jogar fora a linha se faltar QUALQUER coisa (dropna)")
display(df.dropna())

print("\nOpção 2: Imputar o valor médio (fillna)")
media_salario = df["Salario"].mean()
print(f"(A média de salário dos não-nulos é {media_salario})")

df_limpo = df.copy()
df_limpo["Salario"] = df_limpo["Salario"].fillna(media_salario)
df_limpo["Idade"] = df_limpo["Idade"].fillna(0) # Pessoas sem idade vão ficar com 0 (talvez não seja uma boa ideia dependendo do modelo)

display(df_limpo)
```

## 3️⃣ Removendo Linhas Duplicadas

Nossa tabela tem as linhas "Ana" e "Caio" repetidas. Isso cria viés de treino!

```python
print("As linhas duplicadas são:")
display(df_limpo[df_limpo.duplicated()])

df_final = df_limpo.drop_duplicates()
print("\nBase purificada:")
display(df_final)
```
