## 1️⃣ Agregação Global

Se não passarmos o parâmetro `axis`, as funções do Numpy achatam (flatten) a matriz inteira e retornam um único valor global.

```python
import numpy as np

dados = np.random.randn(100, 100) * 15 + 50 # 10.000 amostras (média 50, desvio 15)

print(f"Mínimo: {dados.min():.2f}")
print(f"Máximo: {dados.max():.2f}")
print(f"Média:  {dados.mean():.2f}")
print(f"Desvio: {dados.std():.2f}")
print(f"Índice do Máximo Absoluto (argmax): {dados.argmax()}")
```

## 2️⃣ Entendendo os Eixos (`axis`)

Se tivermos uma matriz tabular (linhas e colunas), podemos querer a "Soma de cada Coluna" ou a "Soma de cada Linha". O `axis` resolve isso.

> 💡 **Regra de Ouro:** O `axis` informa qual dimensão será COLAPSADA (esmagada).
> - `axis=0`: Esmaga as linhas (anda de cima para baixo). O resultado será calculado por Coluna.
> - `axis=1`: Esmaga as colunas (anda da esquerda para direita). O resultado será por Linha.

```python
tabela = np.array([
    [10, 20, 30],
    [ 1,  2,  3]
])
print("Tabela:\n", tabela)

# Esmagando as linhas (Soma vertical -> resulta numa soma por coluna)
soma_colunas = tabela.sum(axis=0)
print("\nSoma com axis=0 (Colunas):", soma_colunas)

# Esmagando as colunas (Soma horizontal -> resulta numa soma por linha)
soma_linhas = tabela.sum(axis=1)
print("Soma com axis=1 (Linhas) :", soma_linhas)
```
