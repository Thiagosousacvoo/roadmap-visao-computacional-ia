## 1️⃣ O Problema da Memória

Imagine que você tem 1 milhão de imagens para treinar um modelo de Visão Computacional. Você tenta carregar tudo em uma lista Python... e seu computador trava. A memória RAM lotou.

Como resolver? Processando "uma de cada vez" (ou em pequenos lotes/batches).

```python
import sys

# Compreensão de Lista: cria TUDO na memória de uma vez.
lista_gigante = [x ** 2 for x in range(1_000_000)]
print(f"Tamanho da lista na memória: {sys.getsizeof(lista_gigante) / 1024 / 1024:.2f} MB")

# Generator Expression: cria as regras, mas não os valores. Gera sob demanda!
gerador_gigante = (x ** 2 for x in range(1_000_000))
print(f"Tamanho do gerador na memória: {sys.getsizeof(gerador_gigante)} BYTES!")
```

## 2️⃣ Como funciona um Iterador?

Usamos a função embutida `next()` para pedir o próximo valor a um gerador. Quando acabar, ele lança uma exceção `StopIteration`.

```python
meu_gerador = (x * 2 for x in [10, 20, 30])

print(next(meu_gerador))
print(next(meu_gerador))
print(next(meu_gerador))

# Se rodarmos next(meu_gerador) novamente, causaria StopIteration!
# O `for` loop do Python basicamente roda `next()` escondido e trata esse erro sozinho.
```

## 3️⃣ Criando suas próprias funções geradoras (`yield`)

Em vez de `return`, usamos `yield`. O `yield` "pausa" a função, entrega o valor, e quando a função for chamada de novo, ela continua exatamente da linha onde parou!

```python
def contagem_regressiva(n):
    print("Iniciando a contagem...")
    while n > 0:
        yield n
        print(f"Retomando do {n}...")
        n -= 1
    print("Fogo! 🚀")

contador = contagem_regressiva(3)
print("Primeiro next:", next(contador))
print("Segundo next:", next(contador))
print("Terceiro next:", next(contador))
```

## 4️⃣ O Padrão Ouro em IA: Geradores em Batches

Para treinar redes neurais de forma eficiente, processamos imagens em pequenos grupos (ex: 32 por vez).

```python
def gerar_batches(dataset, batch_size):
    """Função clássica de processamento em blocos."""
    for i in range(0, len(dataset), batch_size):
        yield dataset[i : i + batch_size]

dataset_simulado = list(range(1, 101))  # 100 amostras (IDs de 1 a 100)

# Gerar blocos de 32 itens
meu_data_loader = gerar_batches(dataset_simulado, batch_size=32)

for num_batch, batch in enumerate(meu_data_loader, start=1):
    print(f"Treinando Batch {num_batch} -> Tamanho: {len(batch)} | Início: {batch[0]} Fim: {batch[-1]}")
```
