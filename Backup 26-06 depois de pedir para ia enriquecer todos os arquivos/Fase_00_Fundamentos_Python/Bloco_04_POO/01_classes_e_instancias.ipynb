## 1️⃣ O que é uma Classe?

Em Ciência de Dados, muitas vezes queremos agrupar dados (variáveis) e comportamentos (funções) que pertencem à mesma "entidade". Uma classe atua como um molde para criar essas entidades de forma padronizada.

```python
class ModeloIA:
    """Molde genérico para qualquer modelo de IA no nosso sistema."""
    pass

# Criando objetos (instâncias) a partir do molde
modelo_visao = ModeloIA()
modelo_npl = ModeloIA()

print(type(modelo_visao))
print("Eles são o mesmo objeto na memória?", modelo_visao is modelo_npl)
```

## 2️⃣ O Construtor `__init__` e Atributos

O `__init__` é chamado automaticamente assim que a classe é instanciada. Usamos ele para inicializar os atributos daquele objeto. O argumento `self` sempre aponta para a "instância atual".

```python
class Dataset:
    def __init__(self, nome: str, total_amostras: int):
        # "self.nome" é o atributo que pertence ao objeto.
        # "nome" é a variável local passada por parâmetro.
        self.nome = nome
        self.total_amostras = total_amostras
        self.carregado = False
        
# Ao criar, o Python passa o 'self' automaticamente e nós passamos o resto.
mnist = Dataset(nome="MNIST", total_amostras=70000)
imagenet = Dataset("ImageNet", 1200000)

print(f"Dataset 1: {mnist.nome} | Amostras: {mnist.total_amostras}")
print(f"Dataset 2: {imagenet.nome} | Amostras: {imagenet.total_amostras}")
```
