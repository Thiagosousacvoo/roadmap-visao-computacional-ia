## 1️⃣ A Falha da Detecção em Vídeo

Se você passa 30 frames num YOLO, ele vai achar uma pessoa 30 vezes. Mas ele não sabe que é a **mesma** pessoa. 
Para contar carros numa rodovia, ou monitorar clientes numa loja, precisamos de um ID único para cada pessoa (Tracking).

## 2️⃣ Como funciona o SORT e ByteTrack

1. Detectamos os bboxes no Frame T.
2. Um 'Filtro de Kalman' tenta adivinhar, matematicamente, onde essas caixas estarão no Frame T+1 baseado na velocidade.
3. No Frame T+1, calculamos o IoU entre a adivinhação e a detecção real do YOLO. Se baterem, é o mesmo objeto (mantém o ID).
