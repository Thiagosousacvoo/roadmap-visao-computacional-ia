## 1️⃣ O que muda da Classificação?

Até agora, a rede dizia 'Isto é um Gato'. Em Detecção, ela diz: 'Há um Gato aqui: [X, Y, Largura, Altura] com 95% de certeza'.
A saída não é mais um vetor de `[10]` probabilidades, mas uma matriz `[N_objetos, 4_coords + 1_conf + Classes]`.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 2️⃣ Formato Pascal VOC (x_min, y_min, x_max, y_max)
# Usado em Faster R-CNN, SSD. Coordenadas absolutas em Pixels.
bbox_voc = (50, 60, 200, 250)  # Canto sup-esq e inf-dir

# Imagem preta falsa de 300x300
img = np.zeros((300, 300, 3), dtype=np.uint8)
cv2.rectangle(img, (bbox_voc[0], bbox_voc[1]), (bbox_voc[2], bbox_voc[3]), (0,255,0), 2)

plt.imshow(img)
plt.title('Pascal VOC Bounding Box')
plt.show()
```

## 3️⃣ Formato YOLO (Centro_X, Centro_Y, Largura, Altura) Normalizado

O YOLO padroniza tudo entre 0 e 1 dividindo pela largura/altura da imagem. Isso ajuda a rede a ser invariante à escala.

```python
def convert_voc_to_yolo(voc_box, img_w, img_h):
    x_min, y_min, x_max, y_max = voc_box
    
    x_center = ((x_min + x_max) / 2) / img_w
    y_center = ((y_min + y_max) / 2) / img_h
    width = (x_max - x_min) / img_w
    height = (y_max - y_min) / img_h
    
    return (x_center, y_center, width, height)

bbox_yolo = convert_voc_to_yolo(bbox_voc, 300, 300)
print(f'Bbox YOLO: {bbox_yolo}')
print('Notou que todos os valores estão entre 0 e 1?')
```
