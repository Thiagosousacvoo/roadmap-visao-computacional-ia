## 🎯 Objetivos

- Entender por que treinar do zero é quase sempre má ideia na indústria
- Baixar um modelo pré-treinado no ImageNet usando Torchvision

---

## 1️⃣ Não reinvente a roda

O ImageNet tem 1.2 milhão de imagens e levou semanas em data centers para treinar. As primeiras camadas dessa rede já sabem detectar bordas perfeitamente. Por que fazer sua rede aprender 'bordas' do zero para um projeto de gatos?

```python
import torch
import torchvision.models as models

# Carregando o modelo com os pesos já treinados
resnet = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)

print('O Classificador final padrão da ResNet (fc):')
print(resnet.fc)
```

## 2️⃣ O Problema: 1000 Classes

O `fc` (Fully Connected) final da ResNet-18 tem 1000 saídas, porque o ImageNet tem 1000 categorias (cães, gatos, carros, pianos). 
Mas nosso problema (ex: Dog vs Cat) tem só 2 classes!

```python
import torch.nn as nn

# Vamos sobreescrever a última camada!
num_classes_novas = 2
num_features_entrada = resnet.fc.in_features

resnet.fc = nn.Linear(num_features_entrada, num_classes_novas)

print('Novo Classificador final:')
print(resnet.fc)
```
