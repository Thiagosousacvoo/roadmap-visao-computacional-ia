## 1️⃣ CIFAR-10 vs MNIST

O MNIST é preto e branco, objetos no centro, muito fácil.
O CIFAR-10 tem **3 canais** (RGB), **32x32** pixels, objetos complexos (caminhões, pássaros, cavalos). A CNN precisará de mais parâmetros!

```python
import torch
import torch.nn as nn

class CNN_CIFAR(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            # Entrada: (3, 32, 32) - NOTA: 3 canais de entrada!
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),  # Cai pra 16x16
            
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2)   # Cai pra 8x8
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 8 * 8, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, 10)
        )
        
    def forward(self, x):
        return self.classifier(self.features(x))

modelo = CNN_CIFAR()
x = torch.randn(4, 3, 32, 32)
print('Shape final:', modelo(x).shape)
```

## 2️⃣ O Segredo: Blocos VGG-Style

Note que fizemos `Conv -> Conv -> Pool`. Essa é uma técnica da arquitetura VGG. Duas convs 3x3 seguidas têm o mesmo campo de visão de uma 5x5, mas usam menos parâmetros e introduzem mais não-linearidades!
