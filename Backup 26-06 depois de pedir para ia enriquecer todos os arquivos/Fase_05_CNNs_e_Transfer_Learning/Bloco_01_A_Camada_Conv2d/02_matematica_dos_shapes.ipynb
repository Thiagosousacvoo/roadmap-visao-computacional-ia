## 1️⃣ A Fórmula do Shape (Decore!)

`Saída = floor( (Entrada + 2*Padding - Kernel) / Stride ) + 1`

```python
import torch
import torch.nn as nn

def calcular_shape_saida(tamanho_in, padding, kernel, stride):
    import math
    return math.floor((tamanho_in + 2*padding - kernel) / stride) + 1

# Exemplo 1: Mantendo o tamanho (SAME)
# In=224, Pad=1, Kernel=3, Stride=1
print(f'Exemplo 1 (Padding=1, Stride=1): {calcular_shape_saida(224, 1, 3, 1)}')

# Exemplo 2: Reduzindo pela metade
# In=224, Pad=1, Kernel=3, Stride=2
print(f'Exemplo 2 (Padding=1, Stride=2): {calcular_shape_saida(224, 1, 3, 2)}')
```

## 2️⃣ Na Prática com PyTorch

Vamos confirmar se a matemática bate com a realidade.

```python
entrada = torch.randn(1, 3, 224, 224)

conv_reduz = nn.Conv2d(3, 16, kernel_size=3, stride=2, padding=1)
saida = conv_reduz(entrada)

print(f'Shape real do PyTorch: {saida.shape}')
print('Confirmado! A imagem encolheu para 112x112.')
```

## 3️⃣ Sem Padding (VALID)

Se `padding=0`, a imagem sempre encolhe um pouco, pois o kernel não consegue processar as bordas extremas.

```python
conv_sem_pad = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=0)
print(f'Shape sem padding: {conv_sem_pad(entrada).shape}')
print(f'Formula (224 + 0 - 3)/1 + 1 = 222')
```
