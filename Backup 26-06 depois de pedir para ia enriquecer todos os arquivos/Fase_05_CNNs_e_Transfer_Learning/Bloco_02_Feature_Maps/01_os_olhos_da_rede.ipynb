## 1️⃣ Cada Filtro Procura Uma Coisa Diferente

Se a camada `Conv2d` tem 32 filtros (`out_channels=32`), a saída terá 32 'Feature Maps'. 
Isso significa que a rede gerou 32 interpretações da mesma imagem (uma realçando bordas verticais, outra horizontais, etc).

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import torchvision.models as models

# Vamos pegar um modelo pré-treinado real (ResNet18) para ver features reais!
modelo = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
primeira_camada = modelo.conv1

print(primeira_camada)
print('A ResNet18 usa um kernel gigante de 7x7 no começo e gera 64 mapas!')
```

## 2️⃣ Passando uma Imagem e Vendo a Saída

Vamos criar uma imagem de teste sintética com bolinhas e listras.

```python
# Imagem sintética RGB 1x3x224x224
img_tensor = torch.zeros(1, 3, 224, 224)
img_tensor[:, :, 50:150, 50:150] = 1.0  # Um quadrado branco
img_tensor[:, :, 100:200, 100:200] = 0.5 # Outro quadrado cinza

# Passar pela primeira camada
with torch.no_grad():
    feature_maps = primeira_camada(img_tensor)

print(f'Shape da saída: {feature_maps.shape}')
```

## 3️⃣ Plotando os Mapas de Características (Feature Maps)

```python
# Plotar os primeiros 16 (dos 64)
fig, axes = plt.subplots(4, 4, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    # Pegar o i-ésimo mapa: feature_maps[0, i, :, :]
    mapa = feature_maps[0, i].numpy()
    ax.imshow(mapa, cmap='viridis')
    ax.set_title(f'Filtro {i}')
    ax.axis('off')
plt.suptitle('Os "Olhos" da ResNet (O que a camada Conv enxerga)', fontsize=14)
plt.tight_layout()
plt.show()
```
