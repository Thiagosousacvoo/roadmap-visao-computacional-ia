## 1️⃣ Torchvision Transforms

É o módulo padrão do PyTorch para modificar imagens (PIL ou Tensores) on-the-fly (durante o carregamento do dataset).

```python
import torch
import torchvision.transforms as T
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Vamos criar uma imagem PIL falsa para testar
array_img = np.zeros((100, 100, 3), dtype=np.uint8)
array_img[20:80, 20:80] = [255, 0, 0]  # Quadrado vermelho
img_pil = Image.fromarray(array_img)

# Transformação de Treinamento
transform_treino = T.Compose([
    T.RandomHorizontalFlip(p=0.5),
    T.RandomRotation(degrees=15),
    T.ColorJitter(brightness=0.2, contrast=0.2),
    # T.ToTensor(), # Comentado só para a gente ver o resultado da imagem PIL
])

fig, axes = plt.subplots(1, 4, figsize=(15, 4))
axes[0].imshow(img_pil); axes[0].set_title('Original')
for i in range(1, 4):
    img_aug = transform_treino(img_pil)
    axes[i].imshow(img_aug); axes[i].set_title(f'Augmentada {i}')
plt.show()
```
