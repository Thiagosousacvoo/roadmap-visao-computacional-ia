## 1️⃣ Transformers e Patches

CNNs usam janelas deslizantes (Convolução). Transformers usam "Atenção", feita originalmente para textos. 
Para aplicar a imagens, quebramos a imagem num quebra-cabeça (Patches), onde cada pedaço vira como se fosse uma 'palavra' de uma frase!

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from PIL import Image
import urllib.request
import numpy as np

# Simular imagem colorida fake
img = np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8)
patch_size = 16

fig, ax = plt.subplots(figsize=(6, 6))
ax.imshow(img)
# Desenhar o grid dos patches
for i in range(0, 224, patch_size):
    ax.axhline(i, color='white', linewidth=0.5)
    ax.axvline(i, color='white', linewidth=0.5)
ax.set_title(f'Imagem dividida em {int(224/patch_size)}x{int(224/patch_size)} = {int(224/patch_size)**2} Patches')
plt.axis('off')
plt.show()
```
