## 1️⃣ Gerando Ruído

Para testarmos os filtros, precisamos de uma imagem com ruído. O ruído normal (Gaussiano) simula sensores de câmera ruins em pouca luz.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Imagem com formas básicas
img = np.zeros((200, 300), dtype=np.uint8)
cv2.rectangle(img, (50, 50), (250, 150), 200, -1)

# Adicionar ruído
ruido = np.random.normal(0, 30, img.shape).astype(np.int16)
img_ruidosa = np.clip(img.astype(np.int16) + ruido, 0, 255).astype(np.uint8)
```

## 2️⃣ Blur de Média vs Gaussiano

O blur de média dá o mesmo peso a todos os vizinhos. O Gaussiano dá mais peso ao pixel central, preservando melhor os detalhes.

```python
# Filtros básicos
media = cv2.blur(img_ruidosa, (5, 5))
gaussiano = cv2.GaussianBlur(img_ruidosa, (5, 5), 0)
mediana = cv2.medianBlur(img_ruidosa, 5)
bilateral = cv2.bilateralFilter(img_ruidosa, 9, 75, 75)

fig, axes = plt.subplots(2, 3, figsize=(15, 9))
imgs = [img, img_ruidosa, media, gaussiano, mediana, bilateral]
nomes = ['Original', 'Com Ruído', 'Blur Média', 'Gaussiano', 'Mediana', 'Bilateral']
for ax, im, nome in zip(axes.flat, imgs, nomes):
    ax.imshow(im, cmap='gray')
    ax.set_title(nome, fontsize=11, fontweight='bold')
    ax.axis('off')
plt.suptitle('Comparação de Filtros de Suavização', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 3️⃣ Quando usar qual?

- **Média**: Rápido, mas borra muito as bordas.
- **Gaussiano**: O padrão ouro. Suaviza bem preservando mais formato.
- **Mediana**: Excelente para ruído "Salt and Pepper" (pontos brancos/pretos extremos).
- **Bilateral**: Preserva as bordas perfeitamente, mas é mais pesado para processar.
