## 1️⃣ A Operação de Convolução — Passo a Passo

Para cada posição (i,j) da imagem:
1. Extrair janela (vizinhança) do tamanho do kernel
2. Multiplicar elemento a elemento: janela × kernel
3. Somar tudo → resultado[i,j]

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def convolucao_2d(imagem, kernel):
    """Convolução 2D manual — SEM cv2.filter2D."""
    h, w = imagem.shape
    kh, kw = kernel.shape
    pad_h, pad_w = kh // 2, kw // 2
    
    # Padding (zeros ao redor)
    img_padded = np.pad(imagem, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant')
    resultado = np.zeros_like(imagem, dtype=np.float64)
    
    for i in range(h):
        for j in range(w):
            vizinhanca = img_padded[i:i+kh, j:j+kw]
            resultado[i, j] = np.sum(vizinhanca * kernel)
    
    return resultado

print('Função convolucao_2d criada com sucesso!')
```

## 2️⃣ Testar com Imagem Real

```python
# Imagem de teste
img = np.zeros((200, 200), dtype=np.uint8)
cv2.rectangle(img, (40, 40), (160, 160), 255, -1)
cv2.circle(img, (100, 100), 30, 128, -1)

# Kernels
kernel_blur = np.ones((5, 5)) / 25
kernel_borda = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])

# Aplicar manualmente
manual_blur = convolucao_2d(img.astype(np.float64), kernel_blur)
manual_borda = convolucao_2d(img.astype(np.float64), kernel_borda)

# Aplicar com OpenCV para comparar
cv2_blur = cv2.filter2D(img, cv2.CV_64F, kernel_blur)
cv2_borda = cv2.filter2D(img, cv2.CV_64F, kernel_borda)

# Verificar se são iguais
diff_blur = np.abs(manual_blur - cv2_blur).max()
diff_borda = np.abs(manual_borda - cv2_borda).max()
print(f'Diferença máxima blur: {diff_blur:.10f}')
print(f'Diferença máxima borda: {diff_borda:.10f}')
print('✅ Valores devem ser ~0 (idênticos!)' if diff_blur < 0.01 else '❌ Algo está errado')

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes[0,0].imshow(img, cmap='gray'); axes[0,0].set_title('Original')
axes[0,1].imshow(manual_blur, cmap='gray'); axes[0,1].set_title('Blur (manual)')
axes[1,0].imshow(np.clip(manual_borda, 0, 255), cmap='gray'); axes[1,0].set_title('Borda (manual)')
axes[1,1].imshow(np.clip(cv2_borda, 0, 255), cmap='gray'); axes[1,1].set_title('Borda (OpenCV)')
for ax in axes.flat: ax.axis('off')
plt.suptitle('Manual vs OpenCV — Devem ser Idênticos', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
