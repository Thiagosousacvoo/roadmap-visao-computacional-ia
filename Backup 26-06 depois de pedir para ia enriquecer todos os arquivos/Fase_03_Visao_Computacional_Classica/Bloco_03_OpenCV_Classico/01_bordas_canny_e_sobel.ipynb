## 🎯 Objetivos

- Usar Canny e Sobel para detectar bordas
- Entender o efeito dos thresholds do Canny

---

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Imagem com formas
img = np.zeros((300, 300), dtype=np.uint8)
cv2.rectangle(img, (30, 30), (120, 120), 200, -1)
cv2.circle(img, (220, 100), 50, 180, -1)
cv2.ellipse(img, (150, 230), (80, 40), 0, 0, 360, 220, -1)

# Blur primeiro (reduz ruído)
blurred = cv2.GaussianBlur(img, (5, 5), 0)

# Canny com thresholds diferentes
fig, axes = plt.subplots(1, 4, figsize=(20, 4))
axes[0].imshow(img, cmap='gray'); axes[0].set_title('Original')
for ax, (t1, t2) in zip(axes[1:], [(10, 50), (50, 150), (100, 200)]):
    edges = cv2.Canny(blurred, t1, t2)
    ax.imshow(edges, cmap='gray')
    ax.set_title(f'Canny ({t1}, {t2})', fontsize=11, fontweight='bold')
for ax in axes: ax.axis('off')
plt.suptitle('Efeito dos Thresholds no Canny', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Sobel: Bordas Direcionais

```python
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
direcao = np.arctan2(sobel_y, sobel_x)

fig, axes = plt.subplots(1, 4, figsize=(20, 4))
for ax, im, t in zip(axes, [img, np.abs(sobel_x), np.abs(sobel_y), magnitude],
                      ['Original', 'Sobel X (vert.)', 'Sobel Y (horiz.)', 'Magnitude']):
    ax.imshow(im, cmap='gray')
    ax.set_title(t, fontsize=11, fontweight='bold')
    ax.axis('off')
plt.tight_layout()
plt.show()
```
