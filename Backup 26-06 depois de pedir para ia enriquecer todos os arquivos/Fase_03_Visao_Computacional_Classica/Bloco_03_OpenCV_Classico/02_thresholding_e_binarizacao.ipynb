## 1️⃣ O Problema da Iluminação Desigual

O maior inimigo da visão computacional tradicional é a luz. Uma sombra pode arruinar um threshold fixo.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Simular imagem com gradiente de luz
img = np.zeros((200, 400), dtype=np.uint8)
for i in range(400):
    img[:, i] = max(0, min(255, 50 + i // 3))
cv2.rectangle(img, (30, 50), (90, 150), 20, -1)   # Quadrado escuro na área escura
cv2.rectangle(img, (310, 50), (370, 150), 180, -1) # Quadrado claro na área clara
```

## 2️⃣ Fixo vs Otsu vs Adaptativo

- **Fixo**: Você define um valor absoluto (ex: 127).
- **Otsu**: Acha o melhor valor separando o histograma em dois picos.
- **Adaptativo**: Calcula o threshold para cada pedaço da imagem independentemente!

```python
# Threshold fixo
_, bin_fixo = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Otsu (calcula o valor ideal global)
valor_otsu, bin_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print(f'Valor ideal calculado pelo Otsu: {valor_otsu}')

# Adaptativo
bin_adapt = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                   cv2.THRESH_BINARY, 51, 5)

fig, axes = plt.subplots(2, 2, figsize=(14, 8))
for ax, im, t in zip(axes.flat, [img, bin_fixo, bin_otsu, bin_adapt],
                      ['Original (ilum. desigual)', 'Fixo (127)', 'Otsu (auto)', 'Adaptativo (local)']):
    ax.imshow(im, cmap='gray')
    ax.set_title(t, fontsize=12, fontweight='bold')
    ax.axis('off')
plt.suptitle('Fixo e Otsu falham com gradientes. Adaptativo vence!', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
