## 🎯 Objetivos

- Calcular e plotar histogramas de intensidade
- Equalizar histograma para melhorar contraste
- Usar HSV para detecção por cor

---

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Criar imagem de baixo contraste
img = np.random.randint(80, 180, (200, 300), dtype=np.uint8)  # Faixa estreita

# Equalização de histograma
equalizada = cv2.equalizeHist(img)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes[0,0].imshow(img, cmap='gray'); axes[0,0].set_title('Original (baixo contraste)')
axes[0,1].imshow(equalizada, cmap='gray'); axes[0,1].set_title('Equalizada (alto contraste)')
axes[1,0].hist(img.ravel(), 256, [0, 256], color='steelblue'); axes[1,0].set_title('Histograma Original')
axes[1,1].hist(equalizada.ravel(), 256, [0, 256], color='tomato'); axes[1,1].set_title('Histograma Equalizado')
plt.suptitle('Equalização de Histograma — Expande o Contraste', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Detecção por Cor (HSV)

```python
# Criar imagem colorida com regiões de cor específica
img_color = np.zeros((200, 300, 3), dtype=np.uint8)
cv2.rectangle(img_color, (20, 20), (100, 180), (0, 0, 255), -1)   # Vermelho BGR
cv2.rectangle(img_color, (120, 20), (200, 180), (0, 255, 0), -1)  # Verde BGR
cv2.rectangle(img_color, (220, 20), (280, 180), (255, 0, 0), -1)  # Azul BGR

hsv = cv2.cvtColor(img_color, cv2.COLOR_BGR2HSV)

# Detectar vermelho (H ~0 ou ~170-180 no OpenCV)
mask_red = cv2.inRange(hsv, np.array([0, 100, 100]), np.array([10, 255, 255]))
resultado = cv2.bitwise_and(img_color, img_color, mask=mask_red)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)); axes[0].set_title('Original')
axes[1].imshow(mask_red, cmap='gray'); axes[1].set_title('Máscara Vermelha')
axes[2].imshow(cv2.cvtColor(resultado, cv2.COLOR_BGR2RGB)); axes[2].set_title('Resultado')
for ax in axes: ax.axis('off')
plt.tight_layout()
plt.show()
```
