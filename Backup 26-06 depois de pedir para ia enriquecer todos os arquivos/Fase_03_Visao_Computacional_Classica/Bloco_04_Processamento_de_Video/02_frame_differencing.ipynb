## 🎯 Objetivos

- Detectar movimento usando diferença entre frames
- Aplicar threshold na diferença para binarizar
- Encontrar contornos do movimento

---

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Simular 3 frames com objeto se movendo
h, w = 200, 300

frame1 = np.zeros((h, w), dtype=np.uint8)
cv2.circle(frame1, (80, 100), 30, 200, -1)

frame2 = np.zeros((h, w), dtype=np.uint8)
cv2.circle(frame2, (150, 100), 30, 200, -1)

frame3 = np.zeros((h, w), dtype=np.uint8)
cv2.circle(frame3, (220, 100), 30, 200, -1)

# Diferença absoluta entre frames consecutivos
diff_12 = cv2.absdiff(frame1, frame2)
diff_23 = cv2.absdiff(frame2, frame3)

# Binarizar
_, thresh_12 = cv2.threshold(diff_12, 25, 255, cv2.THRESH_BINARY)
_, thresh_23 = cv2.threshold(diff_23, 25, 255, cv2.THRESH_BINARY)

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes[0,0].imshow(frame1, cmap='gray'); axes[0,0].set_title('Frame 1')
axes[0,1].imshow(frame2, cmap='gray'); axes[0,1].set_title('Frame 2')
axes[0,2].imshow(frame3, cmap='gray'); axes[0,2].set_title('Frame 3')
axes[1,0].imshow(diff_12, cmap='gray'); axes[1,0].set_title('|Frame2 - Frame1|')
axes[1,1].imshow(thresh_12, cmap='gray'); axes[1,1].set_title('Threshold → Movimento!')
axes[1,2].imshow(thresh_23, cmap='gray'); axes[1,2].set_title('Threshold 2-3')
for ax in axes.flat: ax.axis('off')
plt.suptitle('Frame Differencing — Detecção de Movimento', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Encontrar Contornos do Movimento

```python
# Dilatar para conectar regiões
dilated = cv2.dilate(thresh_12, None, iterations=2)

contours, _ = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Desenhar bounding boxes no frame 2
frame2_color = cv2.cvtColor(frame2, cv2.COLOR_GRAY2BGR)
for c in contours:
    if cv2.contourArea(c) < 100:
        continue
    x, y, bw, bh = cv2.boundingRect(c)
    cv2.rectangle(frame2_color, (x, y), (x+bw, y+bh), (0, 255, 0), 2)

plt.figure(figsize=(8, 5))
plt.imshow(cv2.cvtColor(frame2_color, cv2.COLOR_BGR2RGB))
plt.title('Movimento Detectado!', fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
```
