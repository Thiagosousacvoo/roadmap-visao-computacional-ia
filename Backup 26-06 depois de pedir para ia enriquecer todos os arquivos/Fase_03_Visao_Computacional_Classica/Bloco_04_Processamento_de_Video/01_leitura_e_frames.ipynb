## 1️⃣ Conceito: Vídeo = Sequência de Imagens

Um vídeo de 30fps com 10 segundos = **300 imagens** (frames).
Cada frame é um array NumPy normal que pode ser processado.

````python
import numpy as np
import cv2

# Simulação: criar 'frames' de vídeo sintético
# (Sem precisar de webcam ou arquivo de vídeo)
n_frames = 60
h, w = 240, 320
fps = 30

print(f'Vídeo sintético: {n_frames} frames, {w}x{h}, {fps} FPS')
print(f'Duração: {n_frames/fps:.1f} segundos')

# Gerar frames com um círculo que se move
frames = []
for i in range(n_frames):
    frame = np.zeros((h, w, 3), dtype=np.uint8)
    x = int(30 + (w - 60) * i / n_frames)
    y = int(h // 2 + 50 * np.sin(2 * np.pi * i / n_frames))
    cv2.circle(frame, (x, y), 20, (0, 255, 0), -1)
    cv2.putText(frame, f'Frame {i}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1)
    frames.append(frame)

print(f'Shape de cada frame: {frames[0].shape}')
print(f'dtype: {frames[0].dtype}')
````

## 2️⃣ Processar Frames

````python
import matplotlib.pyplot as plt

# Mostrar 6 frames como grid
indices = [0, 10, 20, 30, 40, 50]
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
for ax, idx in zip(axes.flat, indices):
    # Processar: converter para grayscale + Canny
    gray = cv2.cvtColor(frames[idx], cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 150)
    
    # Combinar original + bordas
    combined = frames[idx].copy()
    combined[edges > 0] = [0, 0, 255]  # Bordas em vermelho
    
    ax.imshow(cv2.cvtColor(combined, cv2.COLOR_BGR2RGB))
    ax.set_title(f'Frame {idx}', fontsize=11, fontweight='bold')
    ax.axis('off')
plt.suptitle('Processamento Frame a Frame', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````
