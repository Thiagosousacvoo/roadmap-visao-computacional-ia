## O Problema

Na computação gráfica, translações, escalas e cisalhamentos são feitos puramente multiplicando matrizes.
Vamos desenhar um "X" na tela usando vetores e você deve aplicar matrizes para alterá-lo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Pontos originais do desenho (Uma matriz de vetores 2D)
# Cada coluna é um ponto (x, y)
pontos = np.array([
    [0, 1, 2, 0, 1, 2],
    [0, 1, 2, 2, 1, 0]
]) # Shape (2, 6)

def plotar_desenho(p, titulo):
    plt.scatter(p[0, :], p[1, :], color="red")
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.grid()
    plt.title(titulo)
    plt.axhline(0, color="black",linewidth=1)
    plt.axvline(0, color="black",linewidth=1)
    plt.show()

plotar_desenho(pontos, "Original")
```
