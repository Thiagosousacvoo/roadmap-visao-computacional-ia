## 🎯 Objetivos deste notebook

Entender matrizes como coleções de vetores. O shape da matriz dita quem pode se multiplicar com quem.
- Representação matricial.
- O perigo do Broadcasting oculto.
- Multiplicação de matrizes `(m x n) @ (n x p)`.

---

## 1️⃣ Shape: As dimensões mandam no Universo

Uma imagem colorida de 1920x1080 é apenas uma matriz 3D de shape `(1080, 1920, 3)`.

```python
import numpy as np

matriz = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(f"A matriz tem shape: {matriz.shape}")
print(f"Isto significa {matriz.shape[0]} Linhas e {matriz.shape[1]} Colunas.")
```

## 2️⃣ Matriz × Vetor

Quando multiplicamos uma matriz por um vetor, estamos essencialmente passando o vetor por uma "Transformação Linear".
O vetor entra de um tamanho e sai de outro.

```python
W = np.array([
    [1, 0, -1],  # Linha 1
    [2, 2,  0]   # Linha 2
]) # Shape (2, 3)

x = np.array([1, 2, 3]) # Vetor input, Shape (3,)

# Multiplicação via Produto Ponto Matricial
resultado = W @ x

print("Matriz (2,3) multiplicada por Vetor (3,) resulta num Vetor:", resultado.shape)
print("Transformou 3 dimensões em 2 dimensões! A rede neural diminuiu a resolução dos dados!")
```

## 3️⃣ Matriz × Matriz

Para multiplicar duas matrizes, o número de colunas da esquerda DEVE bater com o número de linhas da direita.

```python
A = np.ones((4, 2))
B = np.zeros((2, 5))

# (4 x 2) @ (2 x 5) = (4 x 5)
C = A @ B
print("A @ B shape final:", C.shape)

# O que aconteceria se você tentasse B @ A?
# B tem colunas=5, A tem linhas=4.
# ValueError: matmul: Input operand 1 has a mismatch in its core dimension 0...
```
