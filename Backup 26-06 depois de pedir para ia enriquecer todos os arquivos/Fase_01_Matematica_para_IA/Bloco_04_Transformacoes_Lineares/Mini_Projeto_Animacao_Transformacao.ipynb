## 🎯 Objetivo deste Mini-Projeto

Criar uma **animação** que mostra um conjunto de pontos sendo transformado por matrizes de rotação aplicadas frame a frame. Isso solidifica a intuição de que matrizes são **movimentos no espaço**.

---

## 1️⃣ Preparação: Criar a Forma Base

Vamos criar um triângulo e uma seta para ver claramente a orientação.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

# Triângulo + seta de orientação
triangulo = np.array([
    [0, 0], [1, 0], [0.5, 0.87], [0, 0]  # triângulo equilátero
]).T  # shape (2, 4)

# Verificar shape
print(f'Shape do triângulo: {triangulo.shape}')
print(f'Pontos:\n{triangulo}')
```

## 2️⃣ Função Genérica de Transformação + Plot

Vamos criar uma função reutilizável que aplica qualquer matriz e plota o resultado:

```python
def aplicar_e_plotar(pontos, matrizes, titulos, figsize=(16, 5)):
    """Aplica uma lista de matrizes ao mesmo conjunto de pontos e plota lado a lado."""
    n = len(matrizes)
    fig, axes = plt.subplots(1, n, figsize=figsize)
    if n == 1:
        axes = [axes]
    
    for ax, M, titulo in zip(axes, matrizes, titulos):
        resultado = M @ pontos
        ax.fill(pontos[0], pontos[1], alpha=0.3, color='blue', label='Original')
        ax.plot(pontos[0], pontos[1], 'b-o', linewidth=2)
        ax.fill(resultado[0], resultado[1], alpha=0.3, color='red', label='Transformado')
        ax.plot(resultado[0], resultado[1], 'r-s', linewidth=2)
        ax.set_xlim(-2, 2)
        ax.set_ylim(-1.5, 2)
        ax.set_aspect('equal')
        ax.grid(True, alpha=0.3)
        ax.legend(fontsize=9)
        ax.set_title(titulo, fontsize=12, fontweight='bold')
        det = np.linalg.det(M)
        ax.text(0.02, 0.02, f'det = {det:.2f}', transform=ax.transAxes,
                fontsize=10, color='darkred', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
```

## 3️⃣ Demonstração: 3 Transformações Lado a Lado

```python
def rot(graus):
    t = np.radians(graus)
    return np.array([[np.cos(t), -np.sin(t)], [np.sin(t), np.cos(t)]])

matrizes = [
    rot(45),
    np.array([[1.5, 0], [0, 0.5]]),
    np.array([[1, 0.8], [0, 1]])  # Cisalhamento
]
titulos = ['Rotação 45°', 'Escala (1.5x, 0.5y)', 'Cisalhamento (k=0.8)']

aplicar_e_plotar(triangulo, matrizes, titulos)
```

## 4️⃣ Rotação Progressiva (Simulação de Animação)

Vamos simular frames de uma animação: o triângulo girando de 0° a 360° em passos de 30°.

```python
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
axes = axes.flatten()

for i, angulo in enumerate(range(0, 360, 30)):
    M = rot(angulo)
    resultado = M @ triangulo
    ax = axes[i]
    ax.fill(resultado[0], resultado[1], alpha=0.5, color=plt.cm.hsv(angulo/360))
    ax.plot(resultado[0], resultado[1], 'k-', linewidth=1.5)
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.2)
    ax.set_title(f'θ = {angulo}°', fontsize=11, fontweight='bold')

plt.suptitle('Rotação Progressiva — Frames de Animação', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 5️⃣ Composição Acumulada

Agora algo mais sofisticado: a cada frame, aplicamos rotação + escala crescente. O resultado é uma espiral.

```python
fig, ax = plt.subplots(figsize=(10, 10))
cores = plt.cm.cool(np.linspace(0, 1, 24))

for i, angulo in enumerate(range(0, 720, 30)):
    fator_escala = 1 + i * 0.1  # Cresce a cada frame
    M = rot(angulo) @ np.array([[fator_escala, 0], [0, fator_escala]])
    resultado = M @ triangulo
    ax.fill(resultado[0], resultado[1], alpha=0.3, color=cores[i % len(cores)])
    ax.plot(resultado[0], resultado[1], '-', color=cores[i % len(cores)], linewidth=1)

ax.set_aspect('equal')
ax.grid(True, alpha=0.2)
ax.set_title('Espiral: Rotação + Escala Crescente', fontsize=14, fontweight='bold')
plt.show()
```
