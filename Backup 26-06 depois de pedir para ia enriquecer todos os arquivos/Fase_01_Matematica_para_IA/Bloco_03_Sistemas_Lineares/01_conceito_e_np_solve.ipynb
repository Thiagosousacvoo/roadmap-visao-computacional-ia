## 🎯 Objetivos deste notebook

Sistemas Lineares buscam encontrar as variáveis desconhecidas (`X`) sabendo os coeficientes (`A`) e o resultado final (`Y`). Basicamente, todo treinamento de Machine Learning tenta resolver um sistema linear gigante (só que com erro).

---

## 1️⃣ A Estrutura Matricial `Ax = b`

Imagine o sistema:
- `2x + y = 5`
- `x - y = 1`

Isso pode ser escrito como Matrizes:
```
[2  1]   [x]   [5]
[1 -1] * [y] = [1]
   A      X     B
```

````python
import numpy as np

# Montando a matriz A (Coeficientes)
A = np.array([
    [2,  1],
    [1, -1]
])

# Montando o vetor B (Resultados)
b = np.array([5, 1])

# Resolvendo para X!
x = np.linalg.solve(A, b)
print(f"A solução do sistema é: x = {x[0]:.1f}, y = {x[1]:.1f}")
````

## 2️⃣ Por que np.linalg.solve() e não A_inversa * b?

Matematicamente, `x = A^(-1) @ b` é correto. 
No entanto, calcular matrizes inversas computacionalmente é lento e altamente instável para grandes volumes de dados. A função `solve` usa decomposições otimizadas no nível de processador.

````python
A_inv = np.linalg.inv(A)

# O resultado matemático é o mesmo:
x_alternativo = A_inv @ b
print("Solução via Inversa:", x_alternativo)
print("Use np.linalg.solve sempre que puder em vez de calcular a Inversa pura!")
````
