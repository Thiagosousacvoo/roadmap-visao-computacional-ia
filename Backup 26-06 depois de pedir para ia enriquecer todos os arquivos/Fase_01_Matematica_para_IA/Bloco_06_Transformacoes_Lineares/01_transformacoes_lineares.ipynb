## 1️⃣ Matriz = Transformação

Uma matriz não é só uma tabela de números. Ela é um 'Motor' que pega um vetor e o transforma (estica, gira ou reflete).

```
import numpy as np

vetor = np.array([1, 0])
# Matriz de Rotação 90 graus
matriz_rotacao = np.array([[0, -1], [1, 0]])

vetor_girado = matriz_rotacao @ vetor
print(f'Vetor após rotação: {vetor_girado}')
```
