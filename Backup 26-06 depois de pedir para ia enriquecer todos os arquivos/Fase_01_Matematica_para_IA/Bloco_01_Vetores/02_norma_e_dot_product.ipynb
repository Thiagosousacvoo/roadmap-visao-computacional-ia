## 🎯 Objetivos deste notebook

Como medimos o "tamanho" de um vetor? Como multiplicamos dois vetores juntos? O Produto Escalar é literalmente a operação mais executada por uma Rede Neural a cada milissegundo.

---

## 1️⃣ Norma (Magnitude)

A norma `L2` de um vetor é o seu comprimento. Calculado usando o Teorema de Pitágoras em N dimensões: `√(x² + y² + z² ...)`

```python
import numpy as np

vetor = np.array([3, 4])

# Método Raiz de Pitágoras manual
norma_manual = np.sqrt((3**2) + (4**2))

# O jeito NumPy de fazer
norma_numpy = np.linalg.norm(vetor)

print(f"O comprimento do vetor {vetor} é {norma_numpy}.")
```

## 2️⃣ Produto Escalar (Dot Product)

Multiplica os elementos correspondentes de dois vetores e soma TUDO num único número (escalar).

$$ a \cdot b = a_1b_1 + a_2b_2 + a_3b_3 $$

```python
pesos = np.array([0.2, 0.8, -0.5])
inputs = np.array([1.0, 2.0, -1.0])

# Modo Element-wise e Sum (didático)
dot_manual = np.sum(pesos * inputs)

# Modo Otimizado C (NumPy)
dot_numpy = np.dot(pesos, inputs)

# Modo Pythônico 3.5+ (Recomendado)
dot_operador = pesos @ inputs

print("Dot product calculado:", dot_operador)
```

## 3️⃣ O Significado do Produto Escalar

O Dot Product mede "o quanto dois vetores apontam para a mesma direção".
- Positivo Alto: Mesma direção.
- Zero: Ortogonais (90 graus, independentes).
- Negativo Alto: Direções opostas.

```python
v_dir = np.array([1, 0])
v_cima = np.array([0, 1])
v_esq = np.array([-1, 0])

print("v_dir @ v_cima (Ortogonais):", v_dir @ v_cima)  # Deve dar 0
print("v_dir @ v_esq  (Opostos)   :", v_dir @ v_esq)   # Deve dar -1
print("v_dir @ v_dir  (Idênticos) :", v_dir @ v_dir)   # Positivo máximo (1)
```
