## 1️⃣ A Visão Geométrica

O sistema `2x + y = 5` e `x - y = 1` representa duas retas no plano. A solução do sistema é exatamente o ponto onde essas retas se cruzam.

```
import numpy as np
import matplotlib.pyplot as plt

# A = [[2, 1], [1, -1]], b = [5, 1]
A = np.array([[2, 1], [1, -1]])
b = np.array([5, 1])

solucao = np.linalg.solve(A, b)
print(f'Ponto de Interseção (x, y): {solucao}')
```
