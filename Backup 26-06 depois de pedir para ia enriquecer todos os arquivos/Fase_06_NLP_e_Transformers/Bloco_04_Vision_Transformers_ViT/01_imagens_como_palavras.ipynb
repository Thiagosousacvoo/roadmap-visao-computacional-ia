## 1️⃣ 'A Image is Worth 16x16 Words'

O paper que revolucionou a Visão Computacional propôs: e se quebrássemos uma imagem em quadrados de 16x16, e tratássemos cada quadrado como uma 'palavra' de um texto?

```python
import torch
import torch.nn as nn

class PatchEmbedding(nn.Module):
    def __init__(self, img_size=224, patch_size=16, in_channels=3, d_model=768):
        super().__init__()
        # A mágica: Uma Conv2d cujo kernel E stride são iguais ao tamanho do patch!
        # Isso gera blocos não-sobrepostos.
        self.proj = nn.Conv2d(in_channels, d_model, kernel_size=patch_size, stride=patch_size)
        
    def forward(self, x):
        x = self.proj(x)           # (Batch, 768, 14, 14)
        x = x.flatten(2)           # (Batch, 768, 196)
        x = x.transpose(1, 2)      # (Batch, 196, 768) -> Agora virou formato de Sequência!
        return x

patcher = PatchEmbedding()
print('Extrator de Patches criado.')
```

## 2️⃣ Transformando Imagem em Sequência

Vamos passar uma imagem real de 224x224 RGB e ver ela virar uma 'frase' de 196 palavras!

```python
img_rgb = torch.randn(1, 3, 224, 224)
sequencia_patches = patcher(img_rgb)

print(f'Shape Original:  {img_rgb.shape}')
print(f'Shape Patches:   {sequencia_patches.shape}')
print('\nLê-se: 1 batch, 196 posições (patches), onde cada patch é um vetor de 768 números.')
print('Agora é só jogar isso num bloco Encoder do Transformer (mais Positional Encoding) e classificar!')
```
