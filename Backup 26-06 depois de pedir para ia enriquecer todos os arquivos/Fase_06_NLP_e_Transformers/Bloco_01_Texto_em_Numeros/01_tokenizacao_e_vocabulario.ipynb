## 1️⃣ O Problema do Texto

Redes Neurais só multiplicam matrizes. Elas não entendem 'Gato' ou 'Cachorro'. 
O primeiro passo é mapear palavras para números inteiros (IDs do vocabulário).

```python
# Exemplo manual ultra-simples (Word-level)
frase = 'o rato roeu a roupa'
vocab = {'o': 1, 'rato': 2, 'roeu': 3, 'a': 4, 'roupa': 5, '<UNK>': 0}

tokens_manuais = [vocab.get(palavra, vocab['<UNK>']) for palavra in frase.split()]
print(f'Frase: {frase}')
print(f'Tokens Manuais: {tokens_manuais}')
```

## 2️⃣ Tokenização de Subpalavras (O Padrão da Indústria)

Se o vocabulário tem 30 mil palavras, o que acontece com 'otorrinolaringologista'? Ela vira `<UNK>` (desconhecido). 
A solução (usada por GPT, BERT) é quebrar palavras raras em pedacinhos (Subwords).

```python
# Você precisará rodar: pip install transformers
try:
    from transformers import AutoTokenizer
    # Vamos baixar o tokenizador do BERT
    tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
    
    frase = 'ChatGPT is revolutionizing artificial intelligence!'
    tokens = tokenizer.tokenize(frase)
    ids = tokenizer.encode(frase, add_special_tokens=False)
    
    print(f'Tokens: {tokens}')
    print(f'IDs:    {ids}')
    print('\nNotou como ChatGPT foi quebrado em pedaços (ex: chat, ##g, ##pt)?')
except ImportError:
    print('Por favor, instale a HuggingFace: pip install transformers')
```
