## 1️⃣ O Problema da Ordem

'O cão mordeu o homem' tem os mesmos tokens que 'O homem mordeu o cão'.
RNNs e LSTMs processam palavras da esquerda pra direita, logo, sabem a ordem. 
Transformers processam TODAS as palavras de uma vez (em paralelo). Como eles sabem quem vem primeiro?

## 2️⃣ A Injeção de Posição

Nós somamos um vetor de posição ao vetor da palavra (Embedding). 
`Vetor Final = Word_Embedding + Positional_Embedding`

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

def get_positional_encoding(seq_len, d_model):
    # Implementação oficial do Paper 'Attention Is All You Need'
    PE = np.zeros((seq_len, d_model))
    for pos in range(seq_len):
        for i in range(0, d_model, 2):
            # Para índices pares usamos Seno
            PE[pos, i] = np.sin(pos / (10000 ** (i / d_model)))
            # Para ímpares usamos Cosseno
            if i + 1 < d_model:
                PE[pos, i + 1] = np.cos(pos / (10000 ** (i / d_model)))
    return PE

pe = get_positional_encoding(seq_len=50, d_model=128)

plt.figure(figsize=(10, 5))
plt.imshow(pe, cmap='viridis', aspect='auto')
plt.title('Positional Encoding (Senos e Cossenos entrelaçados)')
plt.xlabel('Dimensões (d_model)')
plt.ylabel('Posição da Palavra na Frase')
plt.colorbar()
plt.show()
```

## 3️⃣ Por que Senos e Cossenos?

Ao usar ondas com frequências diferentes, a rede aprende a deduzir a distância relativa entre as palavras, não importando o tamanho da frase!
