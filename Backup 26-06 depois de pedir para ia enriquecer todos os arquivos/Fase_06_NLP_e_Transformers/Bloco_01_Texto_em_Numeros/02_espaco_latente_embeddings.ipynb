## 1️⃣ A Camada nn.Embedding

A camada `Embedding` é apenas uma tabela de consulta gigante (Lookup Table). 
O Token 42 vai lá na tabela e busca um vetor de 512 números que o representa.

```python
import torch
import torch.nn as nn

vocab_size = 10000  # Nosso dicionário tem 10 mil palavras
dimensao = 4        # Em redes reais (GPT, ViT) isso é 512, 768 ou 4096

embedding_layer = nn.Embedding(num_embeddings=vocab_size, embedding_dim=dimensao)

token_id = torch.tensor([42])
vetor_palavra = embedding_layer(token_id)

print(f'ID 42 virou: {vetor_palavra.detach().numpy()}')
print('Durante o treino, a rede ajusta esses números para agrupar palavras similares!')
```

## 2️⃣ O Espaço Latente (A Magia Semântica)

Quando a rede está bem treinada, a matemática semântica funciona: 
`Vetor(Rei) - Vetor(Homem) + Vetor(Mulher) ≈ Vetor(Rainha)`

```python
import numpy as np

# Valores ilustrativos simplificados (Realeza, Feminilidade)
rei = np.array([0.9, 0.1])
homem = np.array([0.1, 0.1])
mulher = np.array([0.1, 0.9])
rainha_esperada = np.array([0.9, 0.9])

conta = rei - homem + mulher
print(f'Rei - Homem + Mulher = {conta}')
print(f'Rainha original      = {rainha_esperada}')
print('A matemática semântica bateu quase perfeito!')
```
