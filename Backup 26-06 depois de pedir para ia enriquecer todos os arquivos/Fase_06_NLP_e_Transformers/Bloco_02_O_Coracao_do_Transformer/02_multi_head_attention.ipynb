## 1️⃣ A Limitação de Uma Cabeça Só

Se tivermos apenas uma operação de atenção, a palavra 'banco' só pode olhar para UM aspecto do contexto (ex: a relação gramatical de substantivo). 
Com **Multi-Head Attention**, dividimos o vetor da palavra em pedaços. 
- A *Head 1* procura o sujeito do verbo.
- A *Head 2* resolve pronomes (ex: 'ela' refere-se a 'Maria').
- A *Head 3* resolve localização.

```python
import torch
import torch.nn as nn

class MultiHeadAttentionUltraSimples(nn.Module):
    def __init__(self, d_model=512, num_heads=8):
        super().__init__()
        # Cria as matrizes lineares que vão gerar o Q, K, V para TODAS as cabeças juntas
        self.W_q = nn.Linear(d_model, d_model)
        self.W_k = nn.Linear(d_model, d_model)
        self.W_v = nn.Linear(d_model, d_model)
        self.W_o = nn.Linear(d_model, d_model) # Projeção final de saída
        self.num_heads = num_heads
        
    def forward(self, x):
        # No self-attention, Q=K=V=x da camada anterior!
        Q = self.W_q(x)
        K = self.W_k(x)
        V = self.W_v(x)
        
        # Em um código real, aqui dividimos (reshape) o d_model pelo num_heads, 
        # calculamos as 8 atenções em paralelo e concatenamos no final.
        # Por fim, passa no W_o.
        return self.W_o(V) # Pseudo-retorno para o exemplo

print('Multi-Head Attention mapeada!')
```

## 2️⃣ Na Prática, o PyTorch já faz isso

A classe `nn.MultiheadAttention` faz todo esse particionamento em C/CUDA de forma absurdamente rápida.

```python
mha = nn.MultiheadAttention(embed_dim=512, num_heads=8, batch_first=True)

# Batch=2, Seq=10 palavras, Dim=512
dummy_input = torch.randn(2, 10, 512)

# Chamada: Q, K, V (iguais no self-attention)
saida_mha, pesos_mha = mha(dummy_input, dummy_input, dummy_input)

print(f'Shape da Entrada: {dummy_input.shape}')
print(f'Shape da Saída:   {saida_mha.shape}')
```
