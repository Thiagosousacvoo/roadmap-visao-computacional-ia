## 1️⃣ Como mostrar isso pro cliente?

Ninguém entende uma matriz numpy cheia de 0s e 1s. Precisamos pintar os 1s de vermelho semi-transparente sobre a foto RGB.

```
import numpy as np
import matplotlib.pyplot as plt

# Simulando Alpha Blending
def overlay_mask(img, mask, color=[255, 0, 0], alpha=0.5):
    overlay = img.copy()
    overlay[mask == 1] = color
    return (img * (1 - alpha) + overlay * alpha).astype(np.uint8)

print('A função usa uma média ponderada entre o pixel da foto original e a cor da máscara.')
```
