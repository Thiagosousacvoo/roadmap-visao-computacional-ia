## 1️⃣ Semântica vs Instância

**Segmentação Semântica:** A rede pinta TODOS os gatos de azul e TODOS os cachorros de vermelho. Se dois gatos estiverem abraçados, viram uma única mancha azul.
**Segmentação de Instância:** A rede pinta o Gato 1 de azul e o Gato 2 de verde. Ela sabe que são dois objetos diferentes (Geralmente usando Bounding Boxes primeiro, como na Mask R-CNN).

Nesta fase, focaremos muito na **Semântica** (U-Net).

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Simulando uma máscara semântica (0=Fundo, 1=Pessoa, 2=Carro)
mascara = np.zeros((100, 100), dtype=np.uint8)
mascara[20:80, 20:40] = 1 # Pessoa 1
mascara[30:90, 35:50] = 1 # Pessoa 2 (sobreposta à Pessoa 1)
mascara[60:90, 70:95] = 2 # Carro

plt.imshow(mascara, cmap='tab10')
plt.title('Segmentação Semântica (Pessoas se fundem)')
plt.colorbar(ticks=[0, 1, 2])
plt.show()
```

## 2️⃣ Segmentação Panóptica

É o Santo Graal atual. Ela une as duas: O Fundo (Rua, Céu, Grama) não tem instâncias, é apenas semântico ('Stuff'). Os objetos ('Things') são segmentados por instâncias (Carro 1, Carro 2).
