## 1️⃣ U-Net vs Foundation Models

Na U-Net, você treina 1 modelo exclusivo para segmentar 'Rachaduras em Asfalto'.

Você passa a imagem e um **Prompt** (Um ponto, uma caixa ou um texto). Ele corta exatamente aquele objeto, **sem nunca ter sido treinado para a sua classe**.
