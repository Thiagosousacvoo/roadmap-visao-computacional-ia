## 1️⃣ Escolhendo a Infraestrutura

- **SageMaker:** Caro, mas gerenciado e perfeito para treinamento distribuído.
- **EC2 com T4 GPU:** Flexível, mas requer setup manual de drivers.
- **AWS Lambda:** Serverless barato, mas INÚTIL para modelos pesados devido ao limite de MB e timeout de RAM/Processador.
