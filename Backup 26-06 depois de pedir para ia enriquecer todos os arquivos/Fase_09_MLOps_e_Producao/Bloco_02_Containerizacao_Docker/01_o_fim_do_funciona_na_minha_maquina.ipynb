## 1️⃣ O Inferno das Dependências de ML

Seu código roda no Windows com PyTorch 2.0 e CUDA 11.8. O servidor de produção é um Ubuntu com drivers genéricos. Se você tentar rodar o código lá, vai quebrar.
**Docker** resolve isso: Ele empacota seu código, o Python, o PyTorch e as bibliotecas C++ do sistema operacional dentro de uma 'caixa' (Container) isolada.

## 2️⃣ Imagem vs Container

- **Imagem:** É o arquivo ZIP congelado. É a 'receita de bolo' (Dockerfile).
- **Container:** É a Imagem rodando viva na memória. Você pode rodar 10 containers iguais da mesma imagem.

```python
# Comandos vitais que todo engenheiro de ML deve saber de cor:
"""
docker build -t minha_api_visao .    # Constrói a imagem lendo o Dockerfile
docker run -p 8000:8000 minha_api_visao  # Roda a imagem expondo a porta
docker ps                            # Lista containers vivos
docker stop <id>                     # Mata o container
"""
print('Docker é obrigatório para deploy em AWS, GCP ou Azure.')
```
