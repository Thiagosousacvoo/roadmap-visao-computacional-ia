## 1️⃣ Por que sair do PyTorch (.pt)?

O PyTorch é um framework de **Pesquisa e Treinamento**. Ele carrega grafos dinâmicos pesados. 
Para produção, queremos algo enxuto. Compilamos a rede para **ONNX** (Open Neural Network Exchange). O modelo vira um binário agnóstico, rodável até em Javascript ou C#.

## 2️⃣ TensorRT (A Bomba Atômica da NVIDIA)
