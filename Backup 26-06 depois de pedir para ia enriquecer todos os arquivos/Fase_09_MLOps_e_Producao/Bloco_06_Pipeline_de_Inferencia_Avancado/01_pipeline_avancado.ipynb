## 1️⃣ Separando API de GPU

A API que recebe a requisição web NÃO deve morar na mesma máquina da Placa de Vídeo se você quiser escalar. A API (barata em CPU) enfileira imagens num broker. O Worker (Máquina com GPU cara) consome a fila e devolve a resposta.
