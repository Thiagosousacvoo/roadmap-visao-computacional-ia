## 1️⃣ O Problema do Flask em ML

No Flask clássico, quando uma requisição chega, ele trava a thread para processar. Se a inferência do seu modelo demora 1 segundo, seu servidor só atende 1 pessoa por segundo. A 2ª pessoa fica esperando na fila.
O **FastAPI** suporta `async/await`. Enquanto o modelo roda na GPU, ele pode receber milhares de outras requisições e enfileirá-las sem travar.

```python
print('Além da velocidade, o FastAPI traz:')
print('1. Validação de dados automática com Pydantic (Adeus ifs validando JSON)')
print('2. Documentação automática interativa (Swagger UI em /docs)')
print('3. Tipagem estática (Mais seguro contra bugs)')
```
