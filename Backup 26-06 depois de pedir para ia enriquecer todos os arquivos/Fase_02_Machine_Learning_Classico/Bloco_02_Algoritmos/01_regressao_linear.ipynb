## 1️⃣ O Modelo Mais Simples: y = wx + b

Regressão linear encontra a reta que minimiza a soma dos erros quadráticos (MSE).
- **w** (peso/coeficiente): inclinação da reta
- **b** (bias/intercepto): onde a reta cruza o eixo y

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados: preço = 2000*m² + 50000 + ruído
np.random.seed(42)
m2 = np.random.uniform(30, 200, 80)
preco = 2000 * m2 + 50000 + np.random.normal(0, 20000, 80)

plt.scatter(m2, preco, alpha=0.5, s=30)
plt.xlabel('Área (m²)')
plt.ylabel('Preço (R$)')
plt.title('Dados: Preço vs Área', fontsize=13, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.show()
```

## 2️⃣ Implementação Manual (Gradient Descent)

```python
# Normalizar (ESSENCIAL para gradient descent)
X = (m2 - m2.mean()) / m2.std()
y = (preco - preco.mean()) / preco.std()

# Gradient descent
w, b = 0.0, 0.0
lr = 0.1
n = len(X)
losses = []

for epoch in range(200):
    y_pred = w * X + b
    loss = np.mean((y_pred - y) ** 2)
    losses.append(loss)
    
    dw = (2/n) * np.sum((y_pred - y) * X)
    db = (2/n) * np.sum(y_pred - y)
    w -= lr * dw
    b -= lr * db

print(f'w = {w:.4f}, b = {b:.4f}')
print(f'Loss final: {losses[-1]:.6f}')
```

## 3️⃣ Comparar com sklearn

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X.reshape(-1,1), y)
print(f'Manual:  w={w:.4f}, b={b:.4f}')
print(f'Sklearn: w={model.coef_[0]:.4f}, b={model.intercept_:.4f}')
print(f'\n✅ Valores devem ser praticamente iguais!')

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(losses[:50], 'b-', linewidth=2)
axes[0].set_title('Convergência da Loss', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE')
axes[0].grid(True, alpha=0.3)

axes[1].scatter(X, y, alpha=0.5, s=30)
axes[1].plot(np.sort(X), w*np.sort(X)+b, 'r-', linewidth=2, label=f'y = {w:.2f}x + {b:.2f}')
axes[1].set_title('Regressão Linear', fontsize=13, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
