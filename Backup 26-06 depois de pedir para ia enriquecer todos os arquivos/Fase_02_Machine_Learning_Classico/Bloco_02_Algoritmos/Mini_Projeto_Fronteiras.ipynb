## 🎯 Mini-Projeto: Comparar Fronteiras de Decisão de 5 Modelos

O melhor jeito de entender cada algoritmo é **ver** como ele divide o espaço.

---

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# 3 datasets com complexidade crescente
datasets = [
    ('Linear', make_classification(n_samples=300, n_features=2, n_redundant=0,
                                   n_clusters_per_class=1, random_state=42)),
    ('Moons', make_moons(n_samples=300, noise=0.25, random_state=42)),
    ('Circles', make_circles(n_samples=300, noise=0.15, factor=0.5, random_state=42))
]

modelos = [
    ('Logistic', LogisticRegression()),
    ('Tree (d=5)', DecisionTreeClassifier(max_depth=5)),
    ('RF (100)', RandomForestClassifier(n_estimators=100)),
    ('KNN (5)', KNeighborsClassifier(n_neighbors=5)),
    ('SVM-RBF', SVC(kernel='rbf'))
]

fig, axes = plt.subplots(3, 5, figsize=(25, 15))

for i, (ds_name, (X, y)) in enumerate(datasets):
    X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3, random_state=42)
    for j, (m_name, model) in enumerate(modelos):
        model.fit(X_tr, y_tr)
        acc = model.score(X_te, y_te)
        ax = axes[i, j]
        # Boundary
        h = 0.05
        xx, yy = np.meshgrid(
            np.arange(X[:,0].min()-1, X[:,0].max()+1, h),
            np.arange(X[:,1].min()-1, X[:,1].max()+1, h))
        Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
        ax.contourf(xx, yy, Z, alpha=0.3, cmap='RdBu')
        ax.scatter(X_tr[y_tr==0,0], X_tr[y_tr==0,1], c='blue', s=15, alpha=0.5)
        ax.scatter(X_tr[y_tr==1,0], X_tr[y_tr==1,1], c='red', s=15, alpha=0.5)
        ax.set_title(f'{m_name}\n{acc:.0%}', fontsize=10)
        if j == 0: ax.set_ylabel(ds_name, fontsize=12, fontweight='bold')

plt.suptitle('Fronteiras de Decisão: 5 Modelos × 3 Datasets', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
