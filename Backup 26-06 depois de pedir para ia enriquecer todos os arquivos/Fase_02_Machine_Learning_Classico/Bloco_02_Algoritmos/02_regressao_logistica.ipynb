## 1️⃣ A Curva Sigmoid

A sigmoid transforma qualquer número real em uma probabilidade entre 0 e 1:
```
σ(z) = 1 / (1 + e^(-z))
```

Propriedades:
- `σ(0) = 0.5` → ponto de decisão
- `z → +∞ → σ(z) → 1`
- `z → -∞ → σ(z) → 0`

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = np.linspace(-10, 10, 200)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(z, sigmoid(z), 'b-', linewidth=2.5)
ax.axhline(0.5, color='r', linestyle='--', alpha=0.5, label='Threshold 0.5')
ax.axvline(0, color='gray', linestyle=':', alpha=0.5)
ax.fill_between(z, sigmoid(z), 0.5, where=sigmoid(z)>0.5, alpha=0.1, color='green', label='Classe 1')
ax.fill_between(z, sigmoid(z), 0.5, where=sigmoid(z)<0.5, alpha=0.1, color='red', label='Classe 0')
ax.set_xlabel('z = w·x + b', fontsize=12)
ax.set_ylabel('σ(z) = Probabilidade', fontsize=12)
ax.set_title('Função Sigmoid: Transforma Score em Probabilidade', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.show()
````

## 2️⃣ Treinando na Prática

````python
# Dataset binário
X, y = make_classification(n_samples=300, n_features=2, n_redundant=0,
                           n_clusters_per_class=1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

# Previsões
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)  # Probabilidades!

print(f'Accuracy: {accuracy_score(y_test, y_pred):.2%}')
print(f'\nPrimeiras 5 previsões:')
print(f'  Classe: {y_pred[:5]}')
print(f'  Prob 0: {y_proba[:5, 0].round(3)}')
print(f'  Prob 1: {y_proba[:5, 1].round(3)}')
````

## 3️⃣ Decision Boundary

````python
fig, ax = plt.subplots(figsize=(10, 8))

# Plotar boundary
h = 0.05
x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1].reshape(xx.shape)

contour = ax.contourf(xx, yy, Z, levels=20, cmap='RdBu_r', alpha=0.6)
plt.colorbar(contour, label='P(classe 1)')
ax.scatter(X_test[y_test==0, 0], X_test[y_test==0, 1], c='blue', s=40, label='Classe 0', edgecolors='white')
ax.scatter(X_test[y_test==1, 0], X_test[y_test==1, 1], c='red', s=40, label='Classe 1', edgecolors='white')
ax.set_title('Decision Boundary — Regressão Logística', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.2)
plt.show()
````

## 4️⃣ Mudando o Threshold

O default é 0.5, mas em problemas médicos ou de fraude, você pode querer outro valor:

````python
from sklearn.metrics import precision_score, recall_score

thresholds = [0.3, 0.5, 0.7]
print(f'{"Threshold":>10} | {"Precision":>10} | {"Recall":>10} | {"Accuracy":>10}')
print('-' * 50)
for t in thresholds:
    y_custom = (y_proba[:, 1] >= t).astype(int)
    p = precision_score(y_test, y_custom)
    r = recall_score(y_test, y_custom)
    a = accuracy_score(y_test, y_custom)
    print(f'{t:>10.1f} | {p:>10.4f} | {r:>10.4f} | {a:>10.4f}')

print('\n→ Threshold baixo = mais recall (menos falsos negativos)')
print('→ Threshold alto = mais precision (menos falsos positivos)')
````
