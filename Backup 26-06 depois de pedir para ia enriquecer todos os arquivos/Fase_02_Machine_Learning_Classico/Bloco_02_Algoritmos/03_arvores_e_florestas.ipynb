## 1️⃣ Decision Tree: O Algoritmo que Faz Perguntas

Uma árvore de decisão aprende regras como:
```
Se petal_length <= 2.45:
  → Setosa (100% certo)
Senão:
  Se petal_width <= 1.75:
    → Versicolor (90%)
  Senão:
    → Virginica (95%)
```

O critério para escolher cada split usa **Gini** ou **Entropia** (da Fase 01!).

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

tree = DecisionTreeClassifier(max_depth=3, random_state=42)
tree.fit(X_train, y_train)

# Visualizar a árvore
fig, ax = plt.subplots(figsize=(20, 10))
plot_tree(tree, feature_names=load_iris().feature_names, 
          class_names=load_iris().target_names,
          filled=True, rounded=True, ax=ax, fontsize=10)
ax.set_title('Decision Tree — Iris', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print(f'Accuracy: {tree.score(X_test, y_test):.2%}')
````

## 2️⃣ O Perigo: Overfitting

Sem limitar a profundidade, a árvore cria uma regra para CADA amostra:

````python
# Comparar profundidades
depths = [1, 3, 5, 10, None]
print(f'{"max_depth":>10} | {"Acc Treino":>10} | {"Acc Teste":>10} | {"Gap":>8}')
print('-' * 45)
for d in depths:
    t = DecisionTreeClassifier(max_depth=d, random_state=42).fit(X_train, y_train)
    acc_tr = t.score(X_train, y_train)
    acc_te = t.score(X_test, y_test)
    print(f'{str(d):>10} | {acc_tr:>10.2%} | {acc_te:>10.2%} | {acc_tr-acc_te:>8.2%}')

print('\n→ Sem limite (None): 100% treino mas gap alto = OVERFITTING')
````

## 3️⃣ Random Forest: A Sabedoria da Multidão

Em vez de 1 árvore que overfita, treinar **100 árvores** em subconjuntos aleatórios e votar:

````python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
rf.fit(X_train, y_train)

print(f'Decision Tree: {tree.score(X_test, y_test):.2%}')
print(f'Random Forest: {rf.score(X_test, y_test):.2%}')

# Feature importance
importancias = rf.feature_importances_
features = load_iris().feature_names

fig, ax = plt.subplots(figsize=(10, 5))
ax.barh(features, importancias, color='steelblue')
ax.set_xlabel('Importância')
ax.set_title('Feature Importance — Random Forest', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````
