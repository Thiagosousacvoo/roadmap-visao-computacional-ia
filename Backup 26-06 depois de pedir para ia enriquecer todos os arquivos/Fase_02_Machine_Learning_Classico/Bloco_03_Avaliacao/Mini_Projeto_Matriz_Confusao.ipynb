## 🎯 Mini-Projeto: Confusion Matrix Completa e Interpretação

---

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (confusion_matrix, classification_report,
                             precision_score, recall_score, f1_score)

# Dataset: câncer de mama (binário)
data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.3, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Maligno', 'Benigno'],
            yticklabels=['Maligno', 'Benigno'], ax=ax,
            annot_kws={'size': 18})
ax.set_ylabel('Real', fontsize=13)
ax.set_xlabel('Predito', fontsize=13)
ax.set_title('Confusion Matrix — Câncer de Mama', fontsize=14, fontweight='bold')
plt.show()

print(classification_report(y_test, y_pred, target_names=['Maligno', 'Benigno']))
```

## Interpretação

| Célula | Significado | Consequência |
|---|---|---|
| **TP** (canto inf-dir) | Benigno correto | ✅ |
| **TN** (canto sup-esq) | Maligno correto | ✅ |
| **FP** (sup-dir) | Maligno classificado como Benigno | ⚠️ Paciente vai embora sem tratamento! |
| **FN** (inf-esq) | Benigno classificado como Maligno | 😰 Cirurgia desnecessária |

**Neste problema médico, o que importa mais?** → **Recall de Maligno** (não deixar câncer passar).
