## 🎯 Objetivos

- Implementar MSE, RMSE, MAE e R² do zero
- Saber interpretar R² (e quando ele mente)

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados simulados
np.random.seed(42)
y_true = np.array([3, -0.5, 2, 7, 4.2, 5.5, 1.2, 8, 6.3, 3.8])
y_pred = np.array([2.5, 0.0, 2.1, 7.8, 4.0, 5.2, 1.5, 7.5, 6.0, 4.0])

# Métricas manuais
MSE = np.mean((y_true - y_pred)**2)
RMSE = np.sqrt(MSE)
MAE = np.mean(np.abs(y_true - y_pred))
SS_res = np.sum((y_true - y_pred)**2)
SS_tot = np.sum((y_true - y_true.mean())**2)
R2 = 1 - SS_res / SS_tot

print(f'MSE:  {MSE:.4f}  ← erro quadrático médio')
print(f'RMSE: {RMSE:.4f}  ← mesma unidade que y')
print(f'MAE:  {MAE:.4f}  ← erro absoluto médio')
print(f'R²:   {R2:.4f}  ← 1.0 = perfeito, 0 = modelo prediz a média')
```

## Visualização: Resíduos

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Real vs Predito
axes[0].scatter(y_true, y_pred, s=60, alpha=0.7)
lim = [min(y_true.min(), y_pred.min())-1, max(y_true.max(), y_pred.max())+1]
axes[0].plot(lim, lim, 'r--', label='Perfeito')
axes[0].set_xlabel('Real')
axes[0].set_ylabel('Predito')
axes[0].set_title('Real vs Predito', fontsize=13, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Resíduos
residuos = y_true - y_pred
axes[1].bar(range(len(residuos)), residuos, color=['green' if r>0 else 'red' for r in residuos])
axes[1].axhline(0, color='k', linewidth=0.5)
axes[1].set_xlabel('Amostra')
axes[1].set_ylabel('Resíduo (real - pred)')
axes[1].set_title('Análise de Resíduos', fontsize=13, fontweight='bold')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
