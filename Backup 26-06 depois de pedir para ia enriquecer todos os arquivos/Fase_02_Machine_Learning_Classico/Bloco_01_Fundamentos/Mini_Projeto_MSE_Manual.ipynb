## 🎯 Mini-Projeto: Calcular MSE e Treinar Regressão Linear com NumPy Puro

Sem sklearn — apenas NumPy e Matplotlib. Isso é o que acontece DENTRO do `.fit()`.

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset: preço de casas simplificado
np.random.seed(42)
areas = np.random.uniform(30, 200, 100)  # m²
precos = 1500 * areas + 50000 + np.random.normal(0, 15000, 100)  # R$

# Normalizar (CRÍTICO para gradient descent funcionar)
X = (areas - areas.mean()) / areas.std()
y = (precos - precos.mean()) / precos.std()

# Split manual
X_train, X_test = X[:80], X[80:]
y_train, y_test = y[:80], y[80:]

print(f'Treino: {X_train.shape[0]} | Teste: {X_test.shape[0]}')
```

## Gradient Descent Manual

```python
# Inicializar
w = 0.0
b = 0.0
lr = 0.1
epochs = 100
n = len(X_train)
historico_loss = []

for epoch in range(epochs):
    # Forward
    y_pred = w * X_train + b
    loss = np.mean((y_pred - y_train) ** 2)
    historico_loss.append(loss)
    
    # Gradientes
    dw = (2/n) * np.sum((y_pred - y_train) * X_train)
    db = (2/n) * np.sum(y_pred - y_train)
    
    # Atualizar
    w -= lr * dw
    b -= lr * db
    
    if epoch % 20 == 0:
        print(f'Epoch {epoch:3d} | Loss: {loss:.6f} | w: {w:.4f} | b: {b:.4f}')

# MSE final no teste
y_test_pred = w * X_test + b
mse_test = np.mean((y_test_pred - y_test) ** 2)
print(f'\nMSE no TESTE: {mse_test:.6f}')
```

## Visualização

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(historico_loss, 'b-', linewidth=2)
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE')
axes[0].set_title('Convergência da Loss', fontsize=13, fontweight='bold')
axes[0].grid(True, alpha=0.3)

axes[1].scatter(X_test, y_test, alpha=0.6, s=50, label='Dados teste')
axes[1].plot(np.sort(X_test), w*np.sort(X_test)+b, 'r-', linewidth=2, label=f'y={w:.2f}x+{b:.2f}')
axes[1].set_title('Regressão Linear Manual', fontsize=13, fontweight='bold')
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
