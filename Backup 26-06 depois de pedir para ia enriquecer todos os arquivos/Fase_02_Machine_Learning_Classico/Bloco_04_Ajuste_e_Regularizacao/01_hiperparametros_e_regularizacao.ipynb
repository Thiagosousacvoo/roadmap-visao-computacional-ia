## 1️⃣ Parâmetros vs Hiperparâmetros

| | Parâmetros | Hiperparâmetros |
|---|---|---|
| Definidos por | O modelo aprende | Você define antes |
| Exemplos | Pesos (w), bias (b) | learning_rate, max_depth, C |
| Ajuste | Gradient descent | Grid Search, experiência |

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Dataset com overfitting potencial
np.random.seed(42)
X = np.sort(np.random.uniform(0, 4, 20)).reshape(-1, 1)
y = np.sin(X).ravel() + np.random.normal(0, 0.2, 20)

# Polinômio grau alto (vai overfitar sem regularização)
poly = PolynomialFeatures(degree=12)
X_poly = poly.fit_transform(X)
X_plot = np.linspace(0, 4, 200).reshape(-1, 1)
X_plot_poly = poly.transform(X_plot)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
modelos = [
    ('Sem Regularização', LinearRegression()),
    ('Ridge (L2, α=1)', Ridge(alpha=1.0)),
    ('Lasso (L1, α=0.01)', Lasso(alpha=0.01))
]

for ax, (nome, model) in zip(axes, modelos):
    model.fit(X_poly, y)
    y_plot = model.predict(X_plot_poly)
    ax.scatter(X, y, s=50, alpha=0.7)
    ax.plot(X_plot, y_plot, 'r-', linewidth=2)
    ax.plot(X_plot, np.sin(X_plot), 'g--', alpha=0.5, label='sin(x) real')
    ax.set_ylim(-2, 2)
    ax.set_title(f'{nome}\nMSE treino: {mean_squared_error(y, model.predict(X_poly)):.4f}', fontsize=11, fontweight='bold')
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)

plt.suptitle('Regularização Controla Overfitting', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 2️⃣ Efeito nos Pesos

L1 (Lasso) **zera** pesos irrelevantes → seleção de features.
L2 (Ridge) **reduz** todos os pesos → mais estável.

```python
fig, ax = plt.subplots(figsize=(12, 5))
lr = LinearRegression().fit(X_poly, y)
ridge = Ridge(alpha=1).fit(X_poly, y)
lasso = Lasso(alpha=0.01).fit(X_poly, y)

x_pos = np.arange(len(lr.coef_))
w = 0.25
ax.bar(x_pos - w, np.abs(lr.coef_), w, label='Sem reg.', alpha=0.7)
ax.bar(x_pos, np.abs(ridge.coef_), w, label='Ridge', alpha=0.7)
ax.bar(x_pos + w, np.abs(lasso.coef_), w, label='Lasso', alpha=0.7)
ax.set_xlabel('Coeficiente #')
ax.set_ylabel('|Peso|')
ax.set_title('Magnitude dos Pesos', fontsize=13, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_yscale('log')
plt.show()

print(f'Pesos zerados pelo Lasso: {np.sum(np.abs(lasso.coef_) < 1e-6)} de {len(lasso.coef_)}')
```
