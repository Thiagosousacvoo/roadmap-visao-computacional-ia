## 1️⃣ O Take-Home Project

A empresa te dá um CSV e uma pasta de imagens, e diz 'Crie um classificador em 3 dias'.
Eles NÃO estão interessados em ver 99% de mAP. Eles estão avaliando o seu **Processo**.

**Como gabaritar:**
1. Crie um script `train.py` limpo e modular (sem usar Jupyter Notebook para o código final).
2. Crie um `requirements.txt`.
3. Escreva um README impecável de como rodar.
4. Adicione um arquivo `ANALISE_ERROS.md` mostrando imagens que a sua rede errou e a sua teoria do porquê. (Isso é atitude de Tech Lead).

## 2️⃣ Negociação e Contrato

- Saiba o seu valor: Profissionais de Visão Computacional são muito mais raros que Cientistas de Dados tradicionais. Você mexe com tensores multidimensionais, CUDA e matrizes de câmera.
- Nunca aceite a primeira oferta cegamente. Mostre seu portfólio como alavanca: 'Pela senioridade e problemas que já resolvi no projeto X, acredito que a faixa Y seja mais condizente'.
