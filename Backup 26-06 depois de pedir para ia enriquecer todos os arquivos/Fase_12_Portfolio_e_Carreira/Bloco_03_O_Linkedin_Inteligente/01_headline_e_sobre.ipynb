## 🎯 Objetivos

- SEO no LinkedIn
- Headline que gera convites de entrevista

---

## 1️⃣ A Headline (O Título)

**Ruim:** 'Estudante de IA' ou 'Buscando recolocação'
**Bom:** 'Computer Vision Engineer | PyTorch & MLOps | OpenCV'
O algoritmo do LinkedIn usa o Título como palavra-chave principal. Use os termos técnicos exatos da vaga que você quer.

## 2️⃣ A Seção 'Sobre'

Conte uma história técnica.
- 'Comecei programando em Python, migrei para IA e hoje sou fascinado por ensinar máquinas a enxergar.'
- Liste explicitamente: 'Stack: PyTorch, YOLO, OpenCV, Docker, FastAPI, AWS'.
