## 🎯 Objetivos

- Acabar com a ilusão do 'Portfólio de Cursos'
- Entender o que os recrutadores realmente procuram

---

## 1️⃣ O Que NÃO É Portfólio

Clone do Netflix, classificador de Gato/Cachorro (Kaggle), tutorial do YOLO copiado do YouTube. 
Isso mostra que você sabe apertar Shift+Enter, não que sabe resolver problemas.

**Regra de Ouro:** Se o dataset veio limpinho de um `import torchvision.datasets` ou Kaggle, o projeto vale 10x menos. O mundo real é sujo.

## 2️⃣ O Que É um Portfólio Sênior

- **Problema Real:** 'Onde eu resolvi uma dor?' (Ex: Contar quantas pessoas entram num supermercado usando uma webcam genérica).
- **Coleta de Dados:** 'Fui na rua, tirei 500 fotos de placas de carro do Brasil e anotei no Roboflow'.
- **Métricas Claras:** 'Alcancei mAP de 0.85 em CPU com latência de 30ms'.
- **Deploy:** O modelo não morre no `.ipynb`. Ele tem uma API FastAPI com Docker.
