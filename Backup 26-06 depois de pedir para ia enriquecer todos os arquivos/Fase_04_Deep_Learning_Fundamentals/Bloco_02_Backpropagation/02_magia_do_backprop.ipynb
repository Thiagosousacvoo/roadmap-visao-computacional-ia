## 1️⃣ A Intuição do Gradiente

Se o erro (Loss) foi alto, quem foi o culpado? 
O Backprop distribui a culpa do erro final para trás, de neurônio em neurônio. 
Usamos a derivada parcial `dLoss / dPeso` para saber se aumentar o peso vai piorar ou melhorar o erro.

```python
import numpy as np

# Valores de exemplo
w = 2.0
x = 3.0
y_real = 10.0
lr = 0.01

# Forward
y_pred = w * x
loss = (y_pred - y_real) ** 2  # MSE simplificado

print(f'Loss inicial: {loss}')
```

## 2️⃣ Derivada à Mão (Regra da Cadeia)

Loss = (wx - y_real)². 
Derivada da Loss em relação a `w`:
`dLoss/dw = 2 * (wx - y_real) * x`

```python
# Gradiente
dLoss_dw = 2 * (y_pred - y_real) * x
print(f'Gradiente (dLoss_dw): {dLoss_dw}')

# Atualização (Descida do Gradiente)
w = w - lr * dLoss_dw

# Novo Forward
y_pred_novo = w * x
loss_nova = (y_pred_novo - y_real) ** 2

print(f'Novo w: {w:.4f}')
print(f'Nova Loss: {loss_nova:.4f} (Diminuiu!)')
```

## 3️⃣ O Fluxo numa Rede Real

Na rede real, o PyTorch faz isso automaticamente para matrizes gigantescas usando `loss.backward()`. O processo é: `Erros → dZ2 → dW2 → dA1 → dZ1 → dW1`.
