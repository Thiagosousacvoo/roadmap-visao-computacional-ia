## 🎯 Mini-Projeto: MLP Completa em NumPy

Aqui você vai juntar Forward Pass e Backprop numa rede real de 2 camadas, SEM usar PyTorch. Isso te tornará imune ao erro de não entender os tensores!

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset XOR (não linear)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

# Funções
def sigmoid(z): return 1 / (1 + np.exp(-z))
def sigmoid_deriv(z): return sigmoid(z) * (1 - sigmoid(z))

# Inicialização (2 entradas, 4 ocultos, 1 saída)
np.random.seed(42)
W1 = np.random.randn(2, 4)
b1 = np.zeros((1, 4))
W2 = np.random.randn(4, 1)
b2 = np.zeros((1, 1))

lr = 0.5
losses = []

# Loop de Treino
for epoch in range(10000):
    # FORWARD
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)
    
    # CUSTO (MSE)
    erro = a2 - y
    loss = np.mean(erro ** 2)
    losses.append(loss)
    
    # BACKPROPAGATION (A Matemática Pura!)
    dz2 = erro * sigmoid_deriv(z2)
    dW2 = np.dot(a1.T, dz2)
    db2 = np.sum(dz2, axis=0, keepdims=True)
    
    dz1 = np.dot(dz2, W2.T) * sigmoid_deriv(z1)
    dW1 = np.dot(X.T, dz1)
    db1 = np.sum(dz1, axis=0, keepdims=True)
    
    # UPDATE
    W1 -= lr * dW1
    b1 -= lr * db1
    W2 -= lr * dW2
    b2 -= lr * db2

print('Treino concluído!')
print('Previsões:\n', np.round(a2, 3))

plt.plot(losses)
plt.title('Curva de Loss (Rede aprendeu o XOR!)')
plt.show()
```
