## 1️⃣ O Modelo Simplificado

Um neurônio recebe entradas (X), multiplica por pesos (W), soma um bias (b) e passa por uma função de ativação.

`z = x1*w1 + x2*w2 + ... + xn*wn + b`
Em matrizes: `z = W · X + b`

```python
import numpy as np

x = np.array([1.5, 2.0, -1.0])  # 3 features (ex: peso, altura, idade)
w = np.array([0.5, -0.2, 0.8])  # 3 pesos (importância de cada feature)
b = 0.1                         # bias (tendência do neurônio)

# Cálculo manual
z_manual = (x[0]*w[0]) + (x[1]*w[1]) + (x[2]*w[2]) + b

# Cálculo vetorizado (Produto Escalar)
z_vetor = np.dot(x, w) + b

print(f'Z manual: {z_manual}')
print(f'Z vetorizado: {z_vetor}')
print('→ Eles devem ser exatamente iguais!')
```

## 2️⃣ A Ativação (Step)

No perceptron original, a ativação é um "degrau" (Step Function).
Se `z >= 0` → 1. Se `z < 0` → 0.

```python
def step_function(z):
    return 1 if z >= 0 else 0

print(f'Saída do neurônio: {step_function(z_vetor)}')
```

## 3️⃣ Processando Vários Exemplos de uma Vez (Batch)

Em Deep Learning, nunca passamos uma imagem de cada vez. Passamos várias (Batch).

```python
X_batch = np.array([
    [1.5, 2.0, -1.0],
    [0.5, 1.0, 1.0],
    [-2.0, 0.5, 0.0]
])

# W e b continuam os mesmos
Z_batch = np.dot(X_batch, w) + b
print(f'Z do batch: {Z_batch}')

# Ativação vetorizada
saidas = (Z_batch >= 0).astype(int)
print(f'Saídas do batch: {saidas}')
```
