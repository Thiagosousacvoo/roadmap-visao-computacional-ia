## 1️⃣ A Regra de Aprendizado

O perceptron aprende ajustando os pesos quando erra:
`w_novo = w_antigo + (lr * erro * X)`
Onde `erro = Real - Previsto`.

```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    def __init__(self, input_size, lr=0.1):
        self.W = np.zeros(input_size)
        self.b = 0.0
        self.lr = lr
        self.erros = []

    def predict(self, x):
        z = np.dot(x, self.W) + self.b
        return 1 if z >= 0 else 0

    def fit(self, X, y, epochs=10):
        for _ in range(epochs):
            erro_epoch = 0
            for xi, yi in zip(X, y):
                pred = self.predict(xi)
                erro = yi - pred
                
                # Atualização
                self.W += self.lr * erro * xi
                self.b += self.lr * erro
                
                erro_epoch += int(erro != 0)
            self.erros.append(erro_epoch)
            if erro_epoch == 0:
                break  # Aprendeu 100%
print('Classe criada!')
```

## 2️⃣ Porta Lógica OR

Vamos ensinar o perceptron a resolver o problema lógico OU (OR).

```python
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y_or = np.array([0, 1, 1, 1])

model_or = Perceptron(input_size=2, lr=0.1)
model_or.fit(X, y_or, epochs=20)

plt.plot(model_or.erros, marker='o')
plt.title('Erros por Época (OR)')
plt.xlabel('Época')
plt.ylabel('Total de Erros')
plt.show()

print(f'Pesos finais: W={model_or.W}, b={model_or.b}')
```

## 3️⃣ O Limite (XOR)

O perceptron clássico desenha apenas uma RETA. Ele não consegue resolver o XOR (Ou Exclusivo).

```python
y_xor = np.array([0, 1, 1, 0])

model_xor = Perceptron(input_size=2, lr=0.1)
model_xor.fit(X, y_xor, epochs=20)

plt.plot(model_xor.erros, marker='x', color='red')
plt.title('Erros por Época (XOR) - Nunca converge!')
plt.show()
```
