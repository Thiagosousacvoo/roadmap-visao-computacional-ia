## 1️⃣ Learning Rate (A Taxa de Aprendizado)

É o tamanho do passo que damos montanha abaixo no Gradient Descent. 
- **Alto demais:** A rede pula o fundo e explode (Diverge).
- **Baixo demais:** A rede demora anos para aprender, ou fica presa em buracos rasos (Mínimo local).

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulação de curvas de Loss de acordo com LR
epocas = np.arange(1, 50)
loss_lr_bom = 1 / epocas
loss_lr_baixo = 1 - 0.01 * epocas
loss_lr_alto = np.exp(epocas * 0.1) # Explode!

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].plot(loss_lr_baixo, 'b'); axes[0].set_title('LR muito Baixo (1e-6)\nDemora a descer')
axes[1].plot(loss_lr_bom, 'g'); axes[1].set_title('LR Bom (1e-3)\nDesce rápido e estabiliza')
axes[2].plot(loss_lr_alto, 'r'); axes[2].set_title('LR muito Alto (10.0)\nExplode para infinito!')
for ax in axes: ax.set_xlabel('Epochs'); ax.set_ylabel('Loss')
plt.tight_layout()
plt.show()
```

## 2️⃣ Batch Size (Tamanho do Lote)

Quantas imagens a rede vê antes de atualizar os pesos.
- **Batch = 1 (Stochastic):** Muito rápido, mas caótico. Erros disparam.
- **Batch = 32 ou 64 (Mini-batch):** O padrão. Bom equilíbrio entre memória e estabilidade matemática.
- **Batch = Todo o Dataset (Full-batch):** Perfeito matematicamente, mas não cabe na RAM da GPU.

## 3️⃣ Épocas (Epochs)

Uma ÉPOCA ocorre quando a rede viu TODO o dataset de treino 1 vez. 
Se o dataset tem 1000 imagens e o Batch Size é 100, precisamos de 10 passos (iterações) para completar 1 Época.
