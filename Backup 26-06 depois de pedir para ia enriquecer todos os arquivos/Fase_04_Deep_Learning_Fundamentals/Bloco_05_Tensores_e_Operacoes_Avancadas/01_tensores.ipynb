## 1️⃣ Arrays vs Tensores

Um array do NumPy vive na Memória RAM. Um Tensor do PyTorch pode viver na Memória de Vídeo (VRAM) da Placa de Vídeo. Eles são a espinha dorsal de qualquer rede neural.

```
import torch

tensor_cpu = torch.randn(3, 3)
if torch.cuda.is_available():
    tensor_gpu = tensor_cpu.to('cuda')
    print('Tensor movido para a GPU!')
else:
    print('CUDA não disponível, operando em CPU.')
```
