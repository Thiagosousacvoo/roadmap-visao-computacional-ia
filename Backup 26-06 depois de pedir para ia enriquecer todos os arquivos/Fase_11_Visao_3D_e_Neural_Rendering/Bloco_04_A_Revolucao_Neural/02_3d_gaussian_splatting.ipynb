## 1️⃣ Como resolver a lentidão do NeRF?

Jogar uma rede neural MLP milhares de vezes *por pixel* para gerar um vídeo em tempo real é inviável.
Pesquisadores da Inria apresentaram o **3D Gaussian Splatting**.
Eles substituíram a Rede Neural por **Milhões de Manchas em formato de Ovo (Gaussianas 3D)**.

## 2️⃣ Como funciona o Splatting?

1. Pegamos a Nuvem de Pontos Esparsa do COLMAP.
2. Transformamos cada ponto num 'Ovo' matemático (com Posição, Rotação, Escala, Cor e Opacidade).
3. Projetamos ('Splat') esses Ovos 3D na tela 2D usando placas de vídeo comuns (Rasterização hiper-rápida, como nos jogos de videogame).
4. Comparamos a imagem gerada com a Real, e usamos Backpropagation para atualizar a Escala, Cor e Rotação dos Ovos!
