## 1️⃣ Matriz Intrínseca (O DNA da Lente)

Descreve o hardware da câmera. A Distância Focal (Focal Length) e o Centro Ótico. Essa matriz NUNCA muda se você mover a câmera de lugar, só muda se você der Zoom ou trocar a Lente.

```python
import numpy as np

fx, fy = 800, 800 # Distância Focal (em pixels)
cx, cy = 640, 360 # Centro Ótico (Geralmente metade da resolução 1280x720)

K = np.array([
    [fx, 0, cx],
    [0, fy, cy],
    [0,  0,  1]
])
print('Matriz Intrínseca K:\n', K)
```

## 2️⃣ Matriz Extrínseca (Onde estou no Mundo?)

Descreve a Posição (Translação 3D) e a Rotação (Roll, Pitch, Yaw) da câmera em relação ao cenário. Essa matriz muda toda vez que você dá um passo pra frente ou vira o rosto.

```python
R = np.eye(3) # Rotação neutra (Câmera olhando reto)
T = np.array([[0], [2], [-5]]) # Câmera a 2m de altura, 5m para trás

Extrinseca = np.hstack((R, T))
print('\nMatriz Extrínseca [R|T] (3x4):\n', Extrinseca)
```
