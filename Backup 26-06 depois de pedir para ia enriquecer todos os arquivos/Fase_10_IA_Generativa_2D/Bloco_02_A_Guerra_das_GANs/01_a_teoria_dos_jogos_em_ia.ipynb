## 1️⃣ A Metáfora do Falsificador e do Policial

- **Gerador (O Falsificador):** Nunca vê as imagens reais (Dataset). Ele cria imagens falsas a partir de ruído aleatório e tenta enganar o Policial.
- **Discriminador (O Policial):** Recebe imagens reais do Dataset e imagens Falsas do Gerador. O objetivo dele é classificar corretamente (1 = Real, 0 = Falso).

Eles treinam simultaneamente. Se o Policial fica muito bom, o Falsificador é obrigado a criar obras-primas para enganá-lo.

```python
# Loss do Discriminador (Quer acertar Real=1, Falso=0)
"""
loss_D_real = BCE(Discriminador(img_real), 1)
loss_D_fake = BCE(Discriminador(img_fake), 0)
loss_D_total = loss_D_real + loss_D_fake
loss_D_total.backward() # Treina o Policial
"""

# Loss do Gerador (Quer que o Policial diga Real=1 para suas imagens falsas)
"""
loss_G = BCE(Discriminador(img_fake), 1)
loss_G.backward() # Treina o Falsificador
"""
```

## 2️⃣ Por que as GANs são tão Difíceis?

Se o Policial (Discriminador) ficar perfeito muito rápido (Acurácia = 100%), a Loss do Gerador zera. Ele não aprende mais nada, pois qualquer coisa que ele faz é reprovada sumariamente. Treinar GAN é andar na corda bamba.
