## 1️⃣ A DCGAN

As primeiras GANs usavam camadas Lineares (Dense). A DCGAN revolucionou ao usar CNNs sem MaxPooling (usando Strided Convolutions para descer) e Transposed Convolutions (para subir).

```python
import torch.nn as nn

class DCGAN_Generator(nn.Module):
    def __init__(self, z_dim=100):
        super().__init__()
        # Transforma ruído de 100 numeros em uma Imagem 64x64
        self.net = nn.Sequential(
            # Input (Batch, 100, 1, 1)
            nn.ConvTranspose2d(z_dim, 512, 4, 1, 0, bias=False),
            nn.BatchNorm2d(512),
            nn.ReLU(True),
            # ... (mais camadas) ...
            nn.ConvTranspose2d(64, 3, 4, 2, 1, bias=False),
            nn.Tanh() # Imagem colorida normalizada entre -1 e 1
        )
    def forward(self, z): return self.net(z)

print('Gerador DCGAN. Repare no Tanh final (muito melhor que Sigmoid para imagens RGB em GANs).')
```

## 2️⃣ Mode Collapse (O Medo do Falsificador)

Imagine que o Falsificador descobre que toda vez que ele desenha um Círculo Preto, o Policial erra e acha que é real. 
O Falsificador é preguiçoso: ele vai parar de desenhar qualquer outra coisa. **Todas** as imagens geradas pela rede, não importa o ruído de entrada, viram o maldito Círculo Preto. 
Isso é o Colapso de Modo.
