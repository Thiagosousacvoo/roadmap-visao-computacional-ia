## 1️⃣ Caminhando no Espaço Latente (Morphing)

Como o espaço gerador (VAE ou GAN) é contínuo, se você gerar uma imagem de um Homem (Z1) e uma imagem de uma Mulher (Z2), e caminhar devagar em linha reta entre os pontos `Z1` e `Z2` e passar os valores do caminho no modelo, você verá o Homem se transformando na Mulher suavemente. A imagem não 'esmaece' uma sobre a outra, ela **muta** a estrutura 3D e texturas gradativamente.

```python
import numpy as np

z_homem = np.array([-1.0, 0.5, 2.0])
z_mulher = np.array([2.0, -0.5, 1.0])

# Interpolação Linear Simples (Morphing de 5 quadros)
for alpha in [0.0, 0.25, 0.50, 0.75, 1.0]:
    z_atual = (1 - alpha) * z_homem + alpha * z_mulher
    print(f'Frame com alpha {alpha}: vetor {z_atual}')
    # img = generator(z_atual)
    # save(img)
```
