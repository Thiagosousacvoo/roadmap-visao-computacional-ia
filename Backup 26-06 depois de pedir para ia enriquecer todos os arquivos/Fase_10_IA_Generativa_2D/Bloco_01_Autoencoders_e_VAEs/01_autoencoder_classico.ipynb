## 1️⃣ O que é um Autoencoder?

É uma rede neural treinada para recriar a própria entrada. 
`Imagem -> ENCODER -> Vetor(Gargalo) -> DECODER -> Imagem Recriada`

O truque está no Gargalo (Espaço Latente). Se a imagem tem 784 pixels (28x28) e o Gargalo tem apenas 2 números, a rede é **obrigada** a compactar a essência visual da imagem nesses 2 números.

```python
import torch
import torch.nn as nn

class Autoencoder(nn.Module):
    def __init__(self):
        super().__init__()
        # Encoder amassa a imagem
        self.encoder = nn.Sequential(
            nn.Linear(784, 128),
            nn.ReLU(),
            nn.Linear(128, 2)  # O Gargalo Mágico (Latent Space Z)
        )
        # Decoder tenta reconstruir
        self.decoder = nn.Sequential(
            nn.Linear(2, 128),
            nn.ReLU(),
            nn.Linear(128, 784),
            nn.Sigmoid() # Saída entre 0 e 1 (pixels)
        )
    def forward(self, x):
        z = self.encoder(x)
        return self.decoder(z)

print('Autoencoder Linear Básico! A Loss usada é o MSE (Erro Quadrático Médio) entre a imagem original e a reconstruída.')
```

## 2️⃣ Por que Autoencoders Falham como Geradores?

Se você pegar um Autoencoder treinado, jogar fora o Encoder, e passar um número aleatório (ex: `z = [0.99, -3.14]`) no Decoder... geralmente sai um borrado inútil. 
Isso ocorre porque o espaço latente de um Autoencoder clássico é cheio de 'buracos vazios' onde a rede nunca foi treinada.
