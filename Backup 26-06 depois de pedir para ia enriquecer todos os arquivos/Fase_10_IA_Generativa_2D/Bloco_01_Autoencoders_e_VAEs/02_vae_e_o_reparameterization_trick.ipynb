## 1️⃣ Variational Autoencoder (VAE)

No VAE, o Encoder não devolve um número exato para o Espaço Latente. Ele devolve uma **Distribuição de Probabilidade** (Uma Média `μ` e um Desvio Padrão `σ`).
Assim, a imagem de um 'Rosto Sorrindo' não vira um ponto fixo, vira uma 'nuvem' de pontos. Isso força o espaço latente a ser contínuo e preenchido.

```python
class VAE_Encoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 128)
        self.fc_mu = nn.Linear(128, 2)    # Média
        self.fc_logvar = nn.Linear(128, 2) # Log da Variância
        
    def forward(self, x):
        h = torch.relu(self.fc1(x))
        mu = self.fc_mu(h)
        logvar = self.fc_logvar(h)
        return mu, logvar

print('O VAE prevê parâmetros estatísticos em vez de valores absolutos.')
```

## 2️⃣ O Reparameterization Trick

As redes neurais treinam usando Backpropagation (Cálculo de Derivadas). Você não pode calcular a derivada de um processo aleatório (Como um `random.normal()`).
O truque: `z = μ + σ * epsilon` (Onde epsilon é um ruído fixo externo). A rede aprende a derivar o `μ` e o `σ`, e ignora o `epsilon`.

```python
def reparameterize(mu, logvar):
    std = torch.exp(0.5 * logvar)
    eps = torch.randn_like(std) # O ruído puro (Epsilon)
    return mu + eps * std       # O Z amostrado de forma diferenciável!

print('Magia Matemática!')
```
