## 1️⃣ A Nova Era (Diffusion Models)

As GANs eram instáveis. Os pesquisadores criaram uma nova abordagem: Pegar imagens reais, adicionar Ruído Gaussiano lentamente até virar TV sem sinal, e **treinar uma U-Net para fazer o caminho inverso** (Remover o ruído).

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Forward Process Simulado (Destruição em passos T)
def destroy_image(img, step, max_steps=1000):
    # Quanto maior o step, mais ruído
    noise = torch.randn_like(img)
    alpha = 1 - (step / max_steps) # simplificação brutal da matemática real
    return alpha * img + (1 - alpha) * noise

print('O modelo nunca prevê a imagem inteira de uma vez. Ele prevê qual ruído deve ser SUBTRAÍDO da imagem atual para ir do passo T para o T-1.')
```

## 2️⃣ Por que Difusão venceu as GANs?

Estabilidade. Não há dois modelos brigando. Há apenas uma U-Net fazendo o trabalho de Denoising repetidas vezes, e a função de Perda (Loss) é o MSE puro entre o Ruído Inserido e o Ruído Previsto pela rede.
