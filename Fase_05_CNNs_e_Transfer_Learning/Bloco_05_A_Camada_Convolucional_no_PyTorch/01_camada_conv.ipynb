## 1️⃣ Anatomia da Camada

Se a imagem tem 224x224, e aplicarmos convoluções sem Padding, ela vai encolher a cada camada. O Padding adiciona bordas de zeros para manter o tamanho.

```
import torch
import torch.nn as nn

# Imagem Fake RGB (Batch=1, Channels=3, H=224, W=224)
img = torch.randn(1, 3, 224, 224)

# Camada: Entra 3 canais, sai 16 feature maps. Filtro 3x3.
conv = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, padding=1, stride=1)

saida = conv(img)
print('Shape de saída:', saida.shape)
```
