## 1️⃣ O Problema das Redes Muito Profundas

Antes de 2015, se você colocasse 50 camadas empilhadas, a precisão da rede CAÍA em vez de melhorar. Por que? Porque o gradiente se 'perde' após multiplicar por pesos menores que 1 por 50 vezes.

## 2️⃣ A Mágica de He (Skip Connections)

Se o gradiente não consegue passar pelas convoluções, vamos criar uma 'estrada auxiliar' (atalho). 
A saída passa a ser: `F(x) + x`.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ResidualBlock(nn.Module):
    def __init__(self, canais):
        super().__init__()
        self.conv1 = nn.Conv2d(canais, canais, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(canais)
        
        self.conv2 = nn.Conv2d(canais, canais, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(canais)
        
    def forward(self, x):
        residual = x                      # Salva a entrada original! (O Atalho)
        
        out = self.conv1(x)
        out = self.bn1(out)
        out = F.relu(out)
        
        out = self.conv2(out)
        out = self.bn2(out)
        
        out += residual                   # SOMA COM O ATALHO ANTES DO ÚLTIMO RELU
        out = F.relu(out)
        return out

bloco = ResidualBlock(64)
x = torch.randn(4, 64, 32, 32)
print('Saída do Bloco Residual:', bloco(x).shape)
```

## 3️⃣ O Impacto

Isso permitiu treinar redes com mais de 100 camadas (ResNet-152) e revolucionou a IA. Hoje, quase todo modelo (inclusive Transformers) usa algum tipo de skip connection.
