## 1️⃣ Anatomia da nn.Conv2d

Na Fase 03, você aplicou convolução com um filtro fixo (blur, sobel). Na CNN, a rede **aprende** esses filtros.

```python
import torch
import torch.nn as nn

# nn.Conv2d(in_channels, out_channels, kernel_size)
conv = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, stride=1, padding=1)

print('Camada Convolucional criada!')
print(conv)
```

## 2️⃣ Inspecionando os Pesos

Um kernel 3x3 para uma imagem RGB (3 canais) tem, na verdade, `3x3x3 = 27` pesos. 
Se temos 32 filtros, teremos `32 x 27 = 864` pesos nessa camada (além dos 32 bias).

```python
print(f'Shape dos pesos (W): {conv.weight.shape}')
print('Lê-se: (Filtros, Canais_Entrada, Altura_Kernel, Largura_Kernel)')

print(f'Shape do bias (b): {conv.bias.shape}')
print('Lê-se: (1 bias para cada filtro de saída)')
```

## 3️⃣ Aplicando a Convolução

Sempre lembre do formato esperado pelo PyTorch: `(Batch, Canais, Altura, Largura)`.

```python
# 1 imagem RGB de 224x224 (batch=1)
img_falsa = torch.randn(1, 3, 224, 224)

saida = conv(img_falsa)

print(f'Entrada shape: {img_falsa.shape}')
print(f'Saída shape:   {saida.shape}')
print('\nNotou que os canais mudaram de 3 para 32? A altura/largura continuou 224 porque usamos padding=1.')
```
