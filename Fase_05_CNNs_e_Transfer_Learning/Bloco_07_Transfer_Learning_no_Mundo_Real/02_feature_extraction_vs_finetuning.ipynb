## 1️⃣ Feature Extraction (Congelando tudo)

Se você tem pouquíssimos dados (ex: 200 imagens), não mude a ResNet! Ela vai desaprender (Catastrophic Forgetting). 
Trave a rede inteira e treine APENAS a última camada (o classificador).

```python
import torch
import torch.nn as nn
import torchvision.models as models

modelo = models.resnet18(weights='DEFAULT')

# 1. Congelar TODOS os parâmetros
for param in modelo.parameters():
    param.requires_grad = False

# 2. Criar a nova camada final (ela vem com requires_grad=True por padrão)
modelo.fc = nn.Linear(modelo.fc.in_features, 3) # 3 Classes

# 3. Passar para o otimizador APENAS a camada final
import torch.optim as optim
otimizador = optim.Adam(modelo.fc.parameters(), lr=0.001)

print('Otimizador configurado para atualizar SÓ o modelo.fc!')
```

## 2️⃣ Fine-Tuning (Descongelamento Seletivo)

E se temos mais dados? Podemos treinar as últimas convoluções junto com o classificador!

```python
modelo2 = models.resnet18(weights='DEFAULT')

# Vamos descongelar a 'layer4' (último bloco da ResNet) e a 'fc'
for name, param in modelo2.named_parameters():
    if 'layer4' in name or 'fc' in name:
        param.requires_grad = True
    else:
        param.requires_grad = False

print('Camadas que vão atualizar:')
for name, param in modelo2.named_parameters():
    if param.requires_grad:
        print('-', name)
```
