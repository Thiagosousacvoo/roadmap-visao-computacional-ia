## 🎯 Objetivos

- Juntar Conv2d + ReLU + MaxPool
- Construir a clássica LeNet do Zero

## 1️⃣ O Pipeline End-to-End

Extratores de features (Conv) seguidos de classificadores finais (Linear).

```
import torch.nn as nn

class CNN_Simples(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 16, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(16 * 16 * 16, 10)
        )
    def forward(self, x):
        return self.classifier(self.features(x))

print('Sua primeira arquitetura profissional!')
```
