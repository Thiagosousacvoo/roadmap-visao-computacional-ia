## 1️⃣ A Arquitetura da CNN

Vamos projetar uma rede para o MNIST (imagens 1x28x28 grayscale).

```python
import torch
import torch.nn as nn

class CNN_MNIST(nn.Module):
    def __init__(self):
        super().__init__()
        
        # Bloco de Extração de Features
        self.features = nn.Sequential(
            # Entrada: (1, 28, 28)
            nn.Conv2d(1, 16, kernel_size=3, padding=1),  # -> (16, 28, 28)
            nn.ReLU(),
            nn.MaxPool2d(2, 2),                          # -> (16, 14, 14)
            
            nn.Conv2d(16, 32, kernel_size=3, padding=1), # -> (32, 14, 14)
            nn.ReLU(),
            nn.MaxPool2d(2, 2)                           # -> (32, 7, 7)
        )
        
        # Bloco de Classificação
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(32 * 7 * 7, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 10)  # 10 dígitos (0 a 9)
        )
        
    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

modelo = CNN_MNIST()
print('Modelo instanciado!')
```

## 2️⃣ Teste de Sanidade (Forward Pass em Dummy Data)

Sempre antes de treinar, passe uma imagem falsa para ver se não há erros de dimensões (Size Mismatch).

```python
batch_falso = torch.randn(8, 1, 28, 28)  # Batch de 8 imagens

try:
    saida = modelo(batch_falso)
    print(f'✅ SUCESSO! Shape de saída: {saida.shape} (Esperado: 8, 10)')
except Exception as e:
    print(f'❌ ERRO: {e}')
```
