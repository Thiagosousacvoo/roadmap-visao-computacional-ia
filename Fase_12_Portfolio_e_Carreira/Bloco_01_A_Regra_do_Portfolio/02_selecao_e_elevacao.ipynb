- A Regra do 'Menos é Mais'

---

## 1️⃣ Seleção Brutal

Se você tem 15 repositórios no GitHub com scripts soltos, apague ou deixe privados. 
Mantenha APENAS 3 projetos excepcionais no seu portfólio. 
Os recrutadores olham seu GitHub por no máximo 45 segundos. Eles não vão garimpar o seu ouro.

## 2️⃣ Os 3 Pilares do Portfólio

1. **O Projeto de Base (CNN/Classification):** Um dataset montado por você (ex: Sucata vs Peça Boa). Mostra que você domina PyTorch e Data Augmentation.
2. **O Projeto de Engajamento (YOLO/Tracking):** Algo visual. Contador de carros, detector de EPI, análise de movimento em esportes. Mostra que você entende de vídeo e tempo real.
3. **O Projeto Moderno (Segmentação ou Generativa):** Remoção de fundo de fotos de e-commerce com U-Net, ou um Fine-tuning de Stable Diffusion com LoRA. Mostra que você está na vanguarda.
