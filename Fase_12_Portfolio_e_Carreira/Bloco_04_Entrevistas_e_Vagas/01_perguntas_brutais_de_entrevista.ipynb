## 1️⃣ O Banco de Perguntas (Estude isso!)

**Q1.** 'Seu modelo de classificação de falhas industriais tem 99% de acurácia, mas o cliente reclamou que ele não serve. Qual é o problema?'
> *Resposta Esperada:* A classe de falha provavelmente é rara (desbalanceamento). 99% de acurácia é alcançada chutando que TODAS as peças são boas. Precisamos olhar o Recall (Falsos Negativos) da classe de falha.

**Q2.** 'Qual a diferença prática entre Softmax e Sigmoid na última camada?'
> *Resposta Esperada:* Softmax é para multi-classe excludente (Gato OU Cachorro, a soma dá 1). Sigmoid é para multi-label (Foto de Praia E Foto de Dia, as saídas são independentes).

**Q3.** 'A loss do seu modelo de detecção de repente convergiu para um número estranho (NaN). O que você investiga primeiro?'
> *Resposta Esperada:* Learning Rate explodiu (Diminuir LR), divisões por zero (adicionar epsilon), ou sujeira no Ground Truth (Bboxes com tamanho negativo ou vazios).

**Q4.** 'Por que você escolheu YOLO e não Mask R-CNN no seu projeto?'
> *Resposta Esperada:* Trade-off de latência vs necessidade da tarefa. Eu precisava apenas contar os objetos em tempo real numa Jetson Nano (30 FPS), a bounding box era suficiente e o YOLO entregava isso. A Mask R-CNN me daria os contornos exatos mas rodaria a 2 FPS, inviabilizando o produto.
