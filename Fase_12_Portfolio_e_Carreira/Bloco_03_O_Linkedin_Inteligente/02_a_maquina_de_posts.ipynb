## 1️⃣ O Post 'Build in Public'

Não poste apenas: 'Estou feliz com meu certificado'. Poste a jornada.

**Fórmula do Post de IA:**
1. **Gancho:** 'É mais difícil do que parece detectar se alguém está usando capacete ou não. Aqui está o porquê.'
2. **O Desafio Técnico:** 'O YOLOv8 estava confundindo bonés com capacetes amarelos.'
3. **A Solução:** 'Resolvi aplicando color jitter e coletando mais 200 imagens focadas em luz baixa.'
4. **O Visual:** Um VÍDEO (não imagem estática) do modelo rodando liso.
5. **Call to Action:** 'Código completo no meu GitHub nos comentários.'

Isso prova para a rede inteira que você sabe resolver problemas e atrairá Headhunters ativamente.
