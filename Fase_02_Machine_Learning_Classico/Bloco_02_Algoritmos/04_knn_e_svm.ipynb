## 1️⃣ KNN — K-Nearest Neighbors

O KNN **não treina** — ele memoriza o dataset inteiro. Para classificar um ponto novo, encontra os K vizinhos mais próximos e vota pela maioria.

- `K pequeno` (1-3) → fronteiras complexas → overfitting
- `K grande` (50+) → fronteiras suaves → underfitting

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# KNN do zero
def knn_predict(X_train, y_train, X_new, k=3):
    predictions = []
    for x in X_new:
        # Calcular distância euclidiana para todos os pontos de treino
        dists = np.sqrt(np.sum((X_train - x) ** 2, axis=1))
        # Pegar os K mais próximos
        k_idx = np.argsort(dists)[:k]
        k_labels = y_train[k_idx]
        # Votação por maioria
        pred = np.bincount(k_labels).argmax()
        predictions.append(pred)
    return np.array(predictions)

# Dataset
X, y = make_classification(n_samples=200, n_features=2, n_redundant=0,
                           n_clusters_per_class=1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Testar
y_pred = knn_predict(X_train, y_train, X_test, k=5)
acc = np.mean(y_pred == y_test)
print(f'KNN (k=5) do zero — Accuracy: {acc:.2%}')
```

## 2️⃣ Visualizar Decision Boundary

```python
def plot_boundary(X, y, predict_fn, title, ax):
    h = 0.05
    x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
    y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = predict_fn(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3, cmap='RdBu')
    ax.scatter(X[y==0, 0], X[y==0, 1], c='blue', s=30, alpha=0.6)
    ax.scatter(X[y==1, 0], X[y==1, 1], c='red', s=30, alpha=0.6)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.2)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, k in zip(axes, [1, 5, 30]):
    fn = lambda Xn, k=k: knn_predict(X_train, y_train, Xn, k=k)
    plot_boundary(X_train, y_train, fn, f'KNN (k={k})', ax)

plt.suptitle('Efeito do K na Decision Boundary', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 3️⃣ SVM — Support Vector Machine

O SVM encontra o **hiperplano** que maximiza a **margem** entre as classes.

- **Kernel linear**: fronteira é uma reta/plano
- **Kernel RBF**: projeta dados em dimensão superior → fronteira curva

```python
from sklearn.svm import SVC

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, kernel in zip(axes, ['linear', 'rbf']):
    svm = SVC(kernel=kernel, C=1.0)
    svm.fit(X_train, y_train)
    acc = svm.score(X_test, y_test)
    plot_boundary(X_train, y_train, svm.predict, f'SVM ({kernel}) — Acc: {acc:.2%}', ax)

plt.tight_layout()
plt.show()
```
