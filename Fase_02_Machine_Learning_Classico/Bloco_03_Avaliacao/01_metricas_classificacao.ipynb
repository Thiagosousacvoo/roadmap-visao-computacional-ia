## 1️⃣ A Armadilha da Accuracy

Dataset com 99% classe A e 1% classe B:
- Modelo que **sempre chuta A** → 99% accuracy! 🎉
- Mas **nunca detecta B** → inútil para fraude, câncer, etc.

```python
import numpy as np
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, classification_report)

# Simular: detector de fraude
# y=1 é fraude (rara), y=0 é normal
np.random.seed(42)
y_true = np.concatenate([np.zeros(950), np.ones(50)])  # 5% fraude

# Modelo burro: sempre diz "normal"
y_pred_burro = np.zeros(1000)

# Modelo decente: acerta 80% das fraudes
y_pred_decente = np.zeros(1000)
y_pred_decente[950:990] = 1  # Acerta 40 fraudes, erra 10
y_pred_decente[900:920] = 1  # 20 falsos positivos

print('=== Modelo Burro (sempre diz Normal) ===')
print(f'  Accuracy: {accuracy_score(y_true, y_pred_burro):.2%} ← Parece bom!')
print(f'  Recall:   {recall_score(y_true, y_pred_burro):.2%} ← Nunca detecta fraude!')

print('\n=== Modelo Decente ===')
print(f'  Accuracy:  {accuracy_score(y_true, y_pred_decente):.2%}')
print(f'  Precision: {precision_score(y_true, y_pred_decente):.2%}')
print(f'  Recall:    {recall_score(y_true, y_pred_decente):.2%}')
print(f'  F1:        {f1_score(y_true, y_pred_decente):.2%}')
```

## 2️⃣ Implementação Manual das Métricas

```python
# Confusion matrix: [[TN, FP], [FN, TP]]
cm = confusion_matrix(y_true, y_pred_decente)
TN, FP, FN, TP = cm.ravel()

precision_manual = TP / (TP + FP)
recall_manual = TP / (TP + FN)
f1_manual = 2 * precision_manual * recall_manual / (precision_manual + recall_manual)
accuracy_manual = (TP + TN) / (TP + TN + FP + FN)

print(f'TP={TP}  FP={FP}  FN={FN}  TN={TN}')
print(f'\nPrecision: {precision_manual:.4f}  ← Dos que disse fraude, quantos eram?')
print(f'Recall:    {recall_manual:.4f}  ← Das fraudes reais, quantas encontrou?')
print(f'F1:        {f1_manual:.4f}  ← Equilíbrio entre Precision e Recall')
print(f'Accuracy:  {accuracy_manual:.4f}')
```

## 3️⃣ Quando Usar Cada Métrica

| Contexto | Métrica principal | Por quê |
|---|---|---|
| **Câncer** | Recall | FN = paciente doente sem diagnóstico |
| **Spam** | Precision | FP = email legítimo na lixeira |
| **Geral (balanceado)** | F1 ou Accuracy | Equilíbrio |
| **Competição Kaggle** | Depende da avaliação! | Leia as regras |

```python
# Classification report completo
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

data = load_breast_cancer()
X_tr, X_te, y_tr, y_te = train_test_split(data.data, data.target, test_size=0.3, random_state=42)
model = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_tr, y_tr)
y_pred = model.predict(X_te)

print(classification_report(y_te, y_pred, target_names=['Maligno', 'Benigno']))
```
