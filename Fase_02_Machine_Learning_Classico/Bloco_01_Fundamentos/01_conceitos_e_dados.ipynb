## 1️⃣ O que é Machine Learning?

ML é ensinar o computador a aprender padrões **a partir de dados**, em vez de programar regras explícitas.

| Tipo | Tem resposta (y)? | Exemplo |
|---|---|---|
| **Supervisionado** | ✅ Sim | Prever preço de casa, classificar email spam |
| **Não-Supervisionado** | ❌ Não | Agrupar clientes similares (clustering) |
| **Reforço** | 🎮 Recompensa | Treinar agente para jogar Atari |

## 2️⃣ Anatomia de um Dataset

```
X (features/inputs)     y (labels/target)
┌──────┬──────┬────┐   ┌────────┐
│ m²   │ quartos │ ...│   │ preço  │
│ 80   │ 2    │    │   │ 250k   │
│ 120  │ 3    │    │   │ 380k   │
│ 45   │ 1    │    │   │ 150k   │
└──────┴──────┴────┘   └────────┘
```

- **Features (X):** Informações de entrada — shape `(n_amostras, n_features)`
- **Labels (y):** O que queremos prever — shape `(n_amostras,)`

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Carregar dataset real
iris = load_iris()
X = iris.data
y = iris.target

print(f'X shape: {X.shape} → {X.shape[0]} amostras, {X.shape[1]} features')
print(f'y shape: {y.shape} → {len(np.unique(y))} classes')
print(f'\nFeatures: {iris.feature_names}')
print(f'Classes:  {iris.target_names}')

# Converter para DataFrame para exploração
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y
print(f'\nPrimeiras linhas:')
df.head()
````

## 3️⃣ Exploração Inicial (EDA Rápido)

````python
# Estatísticas descritivas
print(df.describe().round(2))

# Distribuição das classes
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Contagem de classes
classes, counts = np.unique(y, return_counts=True)
axes[0].bar(iris.target_names, counts, color=['#e74c3c', '#2ecc71', '#3498db'])
axes[0].set_title('Distribuição das Classes', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Contagem')

# Scatter 2 features
for i, (name, cor) in enumerate(zip(iris.target_names, ['#e74c3c','#2ecc71','#3498db'])):
    mask = y == i
    axes[1].scatter(X[mask, 0], X[mask, 1], c=cor, label=name, s=40, alpha=0.7)
axes[1].set_xlabel(iris.feature_names[0])
axes[1].set_ylabel(iris.feature_names[1])
axes[1].set_title('Scatter: 2 Features', fontsize=13, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 4️⃣ Tipos de Problemas

| Problema | Tipo de y | Exemplo | Métrica |
|---|---|---|---|
| **Regressão** | Contínuo (float) | Preço de casa | MSE, R² |
| **Classificação binária** | 0 ou 1 | Spam ou não-spam | Precision, Recall |
