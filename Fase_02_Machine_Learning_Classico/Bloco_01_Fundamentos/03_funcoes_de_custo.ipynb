## 1️⃣ MSE — Mean Squared Error (Regressão)

```
MSE = (1/n) Σ (ŷᵢ - yᵢ)²
```

Penaliza erros grandes **quadraticamente** — um erro de 10 pesa 100x mais que um erro de 1.

````python
import numpy as np
import matplotlib.pyplot as plt

# Implementação manual
def mse(y_true, y_pred):
    return np.mean((y_pred - y_true) ** 2)

# Exemplo
y_real = np.array([3.0, 5.0, 2.5, 7.0])
y_pred_bom = np.array([2.8, 5.1, 2.6, 6.8])  # Previsão boa
y_pred_ruim = np.array([5.0, 2.0, 6.0, 3.0])  # Previsão ruim

print(f'MSE (bom):  {mse(y_real, y_pred_bom):.4f}')
print(f'MSE (ruim): {mse(y_real, y_pred_ruim):.4f}')
````

## 2️⃣ Visualizando MSE como Superfície

Para `y = wx + b` com um peso `w`, a MSE forma uma **parábola** — e queremos encontrar o fundo.

````python
# Dados simples: y = 2x + 1 (com ruído)
np.random.seed(42)
X = np.linspace(0, 5, 50)
y_true = 2 * X + 1 + np.random.normal(0, 0.5, 50)

# MSE para diferentes valores de w (com b=1 fixo)
w_range = np.linspace(-1, 5, 200)
losses = [mse(y_true, w * X + 1) for w in w_range]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Curva de Loss
axes[0].plot(w_range, losses, 'b-', linewidth=2)
axes[0].axvline(2.0, color='r', linestyle='--', label='w ótimo = 2')
axes[0].set_xlabel('Peso w', fontsize=12)
axes[0].set_ylabel('MSE', fontsize=12)
axes[0].set_title('MSE vs Peso — O Fundo é o Ótimo', fontsize=13, fontweight='bold')
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

# Dados + regressão
axes[1].scatter(X, y_true, alpha=0.5, s=30)
axes[1].plot(X, 2*X+1, 'r-', linewidth=2, label='y = 2x + 1')
axes[1].set_title('Dados + Reta Ótima', fontsize=13, fontweight='bold')
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 3️⃣ Cross-Entropy Loss (Classificação)

```
BCE = -(1/n) Σ [yᵢ·log(ŷᵢ) + (1-yᵢ)·log(1-ŷᵢ)]
```

Penaliza **confiança errada**: se o modelo diz 0.99 e a resposta é 0, a loss explode.

````python
def binary_cross_entropy(y_true, y_pred):
    eps = 1e-15
    y_pred = np.clip(y_pred, eps, 1 - eps)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# Comparar: modelo confiante certo vs confiante errado
y = np.array([1, 1, 0, 0])
pred_bom = np.array([0.95, 0.88, 0.1, 0.05])
pred_ruim = np.array([0.2, 0.3, 0.8, 0.9])

print(f'BCE (bom):  {binary_cross_entropy(y, pred_bom):.4f}')
print(f'BCE (ruim): {binary_cross_entropy(y, pred_ruim):.4f}')
print(f'\n→ BCE pune severamente confiança na direção errada!')
````

## 4️⃣ MAE — Mean Absolute Error

```
MAE = (1/n) Σ |ŷᵢ - yᵢ|
```
Mais **robusta a outliers** que MSE (não eleva ao quadrado).

````python
def mae(y_true, y_pred):
    return np.mean(np.abs(y_pred - y_true))

# Comparar MSE vs MAE com outlier
y_normal = np.array([3.0, 5.0, 2.5, 7.0])
y_com_outlier = np.array([3.0, 5.0, 2.5, 50.0])  # Outlier!
y_pred = np.array([3.1, 4.9, 2.6, 7.1])

print('Sem outlier:')
print(f'  MSE = {mse(y_normal, y_pred):.4f}  |  MAE = {mae(y_normal, y_pred):.4f}')
print('Com outlier no y_true:')
print(f'  MSE = {mse(y_com_outlier, y_pred):.4f}  |  MAE = {mae(y_com_outlier, y_pred):.4f}')
print('\n→ MSE explode com outlier, MAE cresce linearmente!')
````
