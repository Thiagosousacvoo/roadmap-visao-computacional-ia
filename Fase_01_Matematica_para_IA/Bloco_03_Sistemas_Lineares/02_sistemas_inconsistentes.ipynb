## 🎯 Objetivos deste notebook

Na vida real, a quantidade de equações (dados de clientes) é milhões de vezes maior que o número de variáveis (idade, salário). Sistemas assim não possuem solução exata.
- Introdução à Pseudo-inversa (Least Squares).

---

## 1️⃣ O Problema do Mundo Real

Temos 3 equações para apenas 2 variáveis:
- `x + y = 2`
- `x + y = 3` (Contradição!)
- `2x + y = 5`

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([
    [1, 1],
    [1, 1],
    [2, 1]
])
b = np.array([2, 3, 5])

# Se rodarmos np.linalg.solve(A, b), vai dar erro (LinAlgError) porque A não é quadrada.
# Usamos np.linalg.lstsq (Least Squares / Mínimos Quadrados) para achar a "melhor aproximação possível".

x_aprox, residuals, rank, s = np.linalg.lstsq(A, b, rcond=None)
print(f"Solução mais próxima encontrada: {x_aprox}")
print(f"Erro residual (O quanto erramos no total): {residuals}")
```
