## 🎯 Objetivos deste notebook

Antes de treinar uma IA, precisamos entender como os dados estão espalhados.
- Média vs Mediana.
- Variância e Desvio Padrão.
- Detecção de Outliers com Z-Score.

---

## 1️⃣ Média vs Mediana

A média é incrivelmente sensível a valores extremos (Outliers). A mediana não liga para eles.

```python
import numpy as np

salarios = np.array([2000, 2500, 2200, 2800, 3000, 50000]) # O último cara ganha 50k!

print(f"Média  : R$ {np.mean(salarios):.2f} (Distorcido pelo ricasso)")
print(f"Mediana: R$ {np.median(salarios):.2f} (Mais realista para o grupo)")
```

## 2️⃣ Variância e Desvio Padrão (Standard Deviation - σ)

Variância mede o quão espalhado estão os dados em relação à média. O Desvio padrão é a raiz da variância, trazendo a medida de volta para a unidade original (ex: Reais em vez de Reais Quadrados).

```python
dados_consistentes = np.array([50, 51, 49, 50, 50])
dados_caoticos = np.array([10, 90, 50, 0, 100])

print(f"Desvio Padrão (Consistente): {np.std(dados_consistentes):.2f}")
print(f"Desvio Padrão (Caótico)    : {np.std(dados_caoticos):.2f}")
print("\nRedes Neurais odeiam dados caóticos. Sempre que possível, nós padronizamos os dados para ter média 0 e desvio 1!")
```
