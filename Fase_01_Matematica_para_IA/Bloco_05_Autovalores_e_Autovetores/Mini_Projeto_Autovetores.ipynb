## 🎯 Objetivo deste Mini-Projeto

Calcular autovalores e autovetores de matrizes 2×2 e **visualizá-los** sobrepostos à transformação que a matriz causa. Isso vai cimentar de vez a intuição: autovetores = direções que a matriz não muda.

---

```python
import numpy as np
import matplotlib.pyplot as plt

def visualizar_autovetores(A, titulo=''):
    """Visualiza a transformação de um círculo unitário e os autovetores da matriz A."""
    # Gerar pontos em um círculo unitário
    theta = np.linspace(0, 2*np.pi, 100)
    circulo = np.array([np.cos(theta), np.sin(theta)])  # shape (2, 100)
    
    # Aplicar transformação
    transformado = A @ circulo
    
    # Calcular autovalores e autovetores
    vals, vecs = np.linalg.eig(A)
    
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.plot(circulo[0], circulo[1], 'b-', alpha=0.4, label='Círculo original')
    ax.plot(transformado[0], transformado[1], 'r-', linewidth=2, label='Transformado')
    
    # Plotar autovetores (escalados pelo autovalor)
    cores_seta = ['#00cc00', '#ff8800']
    for i in range(len(vals)):
        if np.isreal(vals[i]):  # Só plotar se for real
            v = np.real(vecs[:, i]) * np.real(vals[i])
            ax.annotate('', xy=v, xytext=(0, 0),
                        arrowprops=dict(arrowstyle='->', color=cores_seta[i], lw=3))
            ax.text(v[0]*1.15, v[1]*1.15, f'λ={np.real(vals[i]):.2f}', 
                    fontsize=12, color=cores_seta[i], fontweight='bold')
    
    lim = max(3, np.abs(transformado).max() * 1.3)
    ax.set_xlim(-lim, lim)
    ax.set_ylim(-lim, lim)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=11)
    ax.set_title(titulo or f'A = {A.tolist()}', fontsize=13, fontweight='bold')
    plt.show()
    
    print(f'Autovalores: {np.round(vals, 4)}')
    print(f'Autovetores (colunas):\n{np.round(vecs, 4)}')
```

## 1️⃣ Matriz de Escala — Autovetores são os Eixos

```python
A1 = np.array([[3, 0],
               [0, 1]])
visualizar_autovetores(A1, 'Escala: estica 3x no eixo X')
```

## 2️⃣ Matriz Simétrica — Autovetores Ortogonais

```python
A2 = np.array([[2, 1],
               [1, 2]])
visualizar_autovetores(A2, 'Matriz simétrica → autovetores ortogonais')
```

## 3️⃣ Rotação Pura — Sem Autovetores Reais!

```python
theta = np.radians(45)
A3 = np.array([[np.cos(theta), -np.sin(theta)],
               [np.sin(theta),  np.cos(theta)]])

vals, vecs = np.linalg.eig(A3)
print('Autovalores de rotação 45°:', vals)
print('São COMPLEXOS! Rotação não tem direção que não muda.')
print(f'|λ| = {abs(vals[0]):.4f} → preserva comprimento (como esperado)')
```
