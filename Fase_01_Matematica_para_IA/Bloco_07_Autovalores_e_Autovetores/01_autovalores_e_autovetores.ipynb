## 🎯 Objetivos

- O que são 'Direções Especiais' (Autovetores)
- A equação `Av = λv`

## 1️⃣ A Intuição

Quando aplicamos uma Matriz de Transformação em um espaço 2D, a maioria dos vetores muda de direção. Porém, alguns vetores muito especiais continuam na mesma linha reta original, eles apenas são 'esticados' ou 'encolhidos'. Estes são os **Autovetores**.

```
import numpy as np

A = np.array([[3, 1], [0, 2]])
autovalores, autovetores = np.linalg.eig(A)
print(f'Autovalores (Fator de Escala): {autovalores}')
print(f'Autovetores (Direções Inalteradas):\n{autovetores}')
```
