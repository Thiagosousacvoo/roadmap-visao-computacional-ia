## 2️⃣ O Dataset PyTorch (Cuidado Extremo com Augmentation!)

Se você girar a foto original em 90 graus (Data Augmentation), **DEVE** girar a máscara nos mesmos exatos 90 graus. Senão a rede vai aprender tudo errado.

```python
import torch
from torch.utils.data import Dataset
from torchvision import transforms
import torchvision.transforms.functional as TF
import random

class SegDataset(Dataset):
    def __init__(self, img_paths, mask_paths):
        self.imgs = img_paths
        self.masks = mask_paths
        
    def __len__(self): return len(self.imgs)
    
    def __getitem__(self, idx):
        # Leria com PIL ou OpenCV aqui
        img = torch.randn(3, 256, 256) # Fake image
        mask = torch.zeros(1, 256, 256) # Fake mask
        
        # Augmentation Conjunto!
        if random.random() > 0.5:
            img = TF.hflip(img)
            mask = TF.hflip(mask)
            
        return img, mask

print('Dataset criado. Augmentation geométrico sempre aplicado a ambos!')
```
