## 1️⃣ A Saída Real

A rede neural não é perfeita. Uma artéria pode ser segmentada com falhas no meio. Como consertar sem treinar de novo?

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Criar uma máscara 'suja'
mascara = np.zeros((100, 100), dtype=np.uint8)
mascara[30:70, 30:70] = 1
mascara[45:55, 45:55] = 0 # Buraco no meio
mascara[10:15, 80:85] = 1 # Ruído longe

kernel = np.ones((5,5), np.uint8)

# 1. Abertura (Opening) tira ruídos externos
abertura = cv2.morphologyEx(mascara, cv2.MORPH_OPEN, kernel)

# 2. Fechamento (Closing) tampa buracos internos
fechamento = cv2.morphologyEx(abertura, cv2.MORPH_CLOSE, kernel)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(mascara); axes[0].set_title('Mascara Suja')
axes[1].imshow(abertura); axes[1].set_title('Após Opening (Ruído sumiu)')
axes[2].imshow(fechamento); axes[2].set_title('Após Closing (Buraco sumiu)')
plt.show()
```
