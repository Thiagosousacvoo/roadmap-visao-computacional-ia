## 1️⃣ O Encoder (Descida)

Pense no Encoder como uma CNN clássica (ResNet). Ela pega a imagem 256x256 e comprime até um gargalo 16x16, extraindo 'O QUÊ' a imagem contém (O contexto semântico), mas perdendo o 'ONDE' exato.

## 2️⃣ O Decoder (Subida)

O Decoder precisa pegar a imagem 16x16 e aumentar (Upsampling / ConvTranspose) de volta para 256x256. 
Problema: Como ele vai reconstruir bordas finas perfeitas se a imagem original foi amassada?

## 3️⃣ Skip Connections (A Ponte)

A U-Net pega os mapas de alta resolução salvos lá no Encoder (antes deles serem amassados) e **CONCATENA** diretamente no Decoder. 
O Decoder passa a ter o 'Contexto' + 'Detalhes das Bordas'.

```python
# Representação Mental do Shape:
"""
Encoder 1: 64 canais, 256x256
   |
   ↓
  (Pooling -> Bottleneck 1024 canais, 16x16)
   |
   ↓
Decoder 1: Upsample -> 512 canais, 256x256
   + 
Skip Connection (Concatena Encoder 1) -> 512 + 64 canais, 256x256
"""
print('As Skip Connections da U-Net usam CONCATENAÇÃO, diferente da ResNet que usa SOMA.')
```
