## 1️⃣ Forward Pass em 2 Camadas

Em vez de apenas entradas → saída. Vamos fazer: `Entrada → Oculta → Saída`.

```python
import numpy as np

def relu(z):
    return np.maximum(0, z)

X = np.array([[0.5, -1.0]])  # 1 exemplo, 2 features

# Camada Oculta (2 entradas -> 3 neurônios)
W1 = np.random.randn(2, 3)
b1 = np.zeros(3)
z1 = np.dot(X, W1) + b1
a1 = relu(z1)  # Ativação OBRIGATÓRIA (sem ela, vira uma regressão linear)

# Camada Saída (3 entradas -> 1 saída)
W2 = np.random.randn(3, 1)
b2 = np.zeros(1)
z2 = np.dot(a1, W2) + b2
y_pred = z2  # Regressão (sem ativação no final)

print(f'Entrada (X): {X.shape}')
print(f'Ativação oculta (a1): {a1.shape} -> {a1}')
print(f'Previsão final (y_pred): {y_pred.shape} -> {y_pred}')
```

## 2️⃣ Função de Custo (Loss)

Para a rede aprender, ela precisa saber quão errada está. O MSE (Mean Squared Error) mede erros de regressão.

```python
y_real = np.array([[2.0]])

erro = y_pred - y_real
mse_loss = np.mean(erro ** 2)

print(f'Loss (MSE): {mse_loss:.4f}')
```
