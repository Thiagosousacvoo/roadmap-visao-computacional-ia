## 1️⃣ A Matemática do Neurônio

`Saída = Ativação(Soma(Pesos * Entradas) + Bias)`
O Perceptron é uma reta que divide o mundo em dois. Se o problema não for linearmente separável, ele falha miseravelmente.

```
import numpy as np

def perceptron(x, w, b):
    z = np.dot(x, w) + b
    return 1 if z >= 0 else 0

print('Teste Lógico AND')
print('1 AND 1 =', perceptron(np.array([1, 1]), np.array([1, 1]), -1.5))
```
