## 1️⃣ Classe Dataset Customizada

O PyTorch exige que você crie uma classe herdando de `Dataset` implementando 2 métodos vitais: `__len__` e `__getitem__`.

```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np

class MeuDataset(Dataset):
    def __init__(self, size=100):
        # Gerar dados falsos de idade(x) e preço(y)
        self.x = torch.randn(size, 5) * 10
        self.y = torch.randint(0, 2, (size,)) # 0 ou 1
        
    def __len__(self):
        return len(self.x)  # Retorna total de linhas
        
    def __getitem__(self, idx):
        # Pega APENAS 1 linha (1 exemplo)
        return self.x[idx], self.y[idx]

dataset = MeuDataset(100)
amostra_x, amostra_y = dataset[0]  # Chama __getitem__(0)
print(f'Uma amostra X: {amostra_x}')
print(f'Um rótulo Y: {amostra_y}')
```

## 2️⃣ O DataLoader (A Mágica)

Nunca passamos `dataset[0]`, `dataset[1]` num loop na mão. O DataLoader cuida de: Batching, Shuffle (embaralhar) e Multiprocessing.

```python
# DataLoader gera blocos (batches) de 16 elementos de uma vez
loader = DataLoader(dataset, batch_size=16, shuffle=True)

# Pega o PRIMEIRO batch do loop
batch_x, batch_y = next(iter(loader))

print(f'Shape do Batch X: {batch_x.shape} (16 amostras, 5 features)')
print(f'Shape do Batch Y: {batch_y.shape} (16 rótulos)')
```
