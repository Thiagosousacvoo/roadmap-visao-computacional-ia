## 1️⃣ Criando o Modelo

No PyTorch, definimos as peças em `__init__` e juntamos a arquitetura no `forward`.

```python
import torch
import torch.nn as nn

class MLP_Simples(nn.Module):
    def __init__(self):
        super().__init__()
        self.camada1 = nn.Linear(5, 10)
        self.relu = nn.ReLU()
        self.saida = nn.Linear(10, 2)
        
    def forward(self, x):
        x = self.camada1(x)
        x = self.relu(x)
        return self.saida(x)

modelo = MLP_Simples()
print(modelo)
```

## 3️⃣ Simulando o Treino Real

```python
import torch.optim as optim

x_dummy = torch.randn(32, 5)
y_dummy = torch.randint(0, 2, (32,))

criterio = nn.CrossEntropyLoss()
otimizador = optim.Adam(modelo.parameters(), lr=0.01)

for epoch in range(3):
    # Modo de Treino (Ativa Dropout/BatchNorm)
    modelo.train()
    
    otimizador.zero_grad()
    preds = modelo(x_dummy)
    loss = criterio(preds, y_dummy)
    loss.backward()
    otimizador.step()
    
    print(f'Epoch {epoch+1} | Loss: {loss.item():.4f}')
```
