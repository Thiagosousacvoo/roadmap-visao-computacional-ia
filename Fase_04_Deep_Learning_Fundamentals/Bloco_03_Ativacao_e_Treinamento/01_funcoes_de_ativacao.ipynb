## 1️⃣ Por que Ativação?

Se você encadear 100 camadas densas SEM ativação (`z = Wx + b`), matematicamente, elas se colapsam numa única matriz linear. A ativação traz as **curvas** (não-linearidade).

```python
import numpy as np
import matplotlib.pyplot as plt

z = np.linspace(-5, 5, 200)

def relu(z): return np.maximum(0, z)
def sigmoid(z): return 1 / (1 + np.exp(-z))
def tanh(z): return np.tanh(z)
def leaky_relu(z, alpha=0.1): return np.where(z > 0, z, z * alpha)

fig, axes = plt.subplots(1, 4, figsize=(18, 4))
axes[0].plot(z, sigmoid(z), 'b'); axes[0].set_title('Sigmoid (Satura nos cantos)')
axes[1].plot(z, tanh(z), 'g'); axes[1].set_title('Tanh (Centro no 0)')
axes[2].plot(z, relu(z), 'r'); axes[2].set_title('ReLU (Morre no negativo)')
axes[3].plot(z, leaky_relu(z), 'orange'); axes[3].set_title('Leaky ReLU (Não morre)')
for ax in axes: ax.grid(True)
plt.show()
```

## 2️⃣ O Problema do Gradiente Sumindo (Vanishing Gradient)

A Sigmoid amassa a saída entre 0 e 1. A derivada dela nas pontas (x=-5 ou x=5) é quase ZERO. 
Se o gradiente é zero, a rede **não aprende**. Por isso usamos ReLU escondida!

## 3️⃣ Softmax: Classificação Múltipla

Quando temos 10 classes (ex: dígitos de 0 a 9), não usamos Sigmoid (que dá probs independentes). 
Usamos Softmax, que garante que a SOMA de todas as probs seja exatamente 1 (100%).

```python
def softmax(z):
    e = np.exp(z - np.max(z)) # Estabilidade
    return e / e.sum()

logits = np.array([2.0, 1.0, 0.1, 5.0])
probs = softmax(logits)
print('Logits puros:', logits)
print('Probs Softmax:', np.round(probs, 3))
print('Soma das probs:', np.sum(probs))
```
