## 1️⃣ Overfitting (Decorando Respostas)

Uma rede com 10 milhões de pesos decora um dataset de 1.000 imagens perfeitamente. 
O sintoma é: **Loss de Treino zera**, mas a **Loss de Validação sobe**.

```python
import numpy as np
import matplotlib.pyplot as plt

epochs = np.arange(1, 50)
train_loss = np.exp(-0.1 * epochs)
val_loss = np.exp(-0.1 * epochs) + 0.01 * (epochs - 20)**2 * (epochs > 20)

plt.plot(epochs, train_loss, 'b', label='Treino Loss')
plt.plot(epochs, val_loss, 'r', label='Validação Loss (Overfit)')
plt.axvline(20, color='gray', linestyle='--', label='Momento de parar (Early Stop)')
plt.legend(); plt.title('Anatomia do Overfitting')
plt.show()
```

## 2️⃣ Dropout

Dropout desliga, por exemplo, 50% dos neurônios (aleatoriamente) em CADA passo do treino. 
Isso impede que um neurônio fique preguiçoso e deixe outro fazer todo o trabalho. A rede se torna um "comitê robusto".

```python
import torch
import torch.nn as nn

camada_dropout = nn.Dropout(p=0.5)
tensor_input = torch.ones(1, 10)

print('Entrada (Tudo 1):', tensor_input)
print('Saída com Dropout (Metade apagou, resto duplicou de peso):')
print(camada_dropout(tensor_input))
```

## 3️⃣ Batch Normalization

Assim como normalizamos as imagens na entrada (0 a 1), Batchnorm normaliza as ativações NO MEIO da rede. 
Ele puxa a média para 0 e desvio para 1 em cada batch, curando sumiços de gradiente.
