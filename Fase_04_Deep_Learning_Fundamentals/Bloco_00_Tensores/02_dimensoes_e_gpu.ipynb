## 1️⃣ Dimensões em Visão Computacional (B, C, H, W)

No PyTorch, uma imagem RGB (3 canais) de 224x224 num batch de 32 imagens tem shape `(32, 3, 224, 224)`. O canal vem antes da altura! (Diferente do NumPy).

```python
import torch

# Batch de 32 imagens RGB de 224x224
batch_imagens = torch.randn(32, 3, 224, 224)

print(f'Shape do batch: {batch_imagens.shape}')
print(f'Total de imagens: {batch_imagens.size(0)}')
print(f'Canais por img:   {batch_imagens.size(1)}')
```

## 2️⃣ Mudando Formatos (Reshape / View)

Para passar uma imagem por uma rede neural linear, você precisa "achatar" a imagem de 2D/3D para 1D.

```python
img_unica = torch.randn(1, 28, 28)  # Imagem grayscale 28x28 (tipo MNIST)

# Achatar para vetor 1D de tamanho 784
img_achatada = img_unica.view(1, -1)  # -1 faz o PyTorch calcular o tamanho restante (28*28 = 784)

print(f'Shape original: {img_unica.shape}')
print(f'Shape achatado: {img_achatada.shape}')
```

## 3️⃣ O Poder da GPU (CUDA)

Redes neurais precisam de milhões de multiplicações. A GPU faz isso muito mais rápido que a CPU.

```python
# Checar se tem GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Dispositivo disponível: {device}')

# Movendo tensor para GPU (se houver)
t = torch.tensor([1, 2, 3])
t_gpu = t.to(device)

print(f'Onde o tensor está? {t_gpu.device}')
```
