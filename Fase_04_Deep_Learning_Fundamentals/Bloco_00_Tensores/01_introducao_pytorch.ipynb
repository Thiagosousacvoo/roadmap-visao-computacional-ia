## 1️⃣ O que é um Tensor?

Tensores são matrizes multidimensionais. A diferença pro NumPy é que tensores:
1. Podem rodar em GPU (muito mais rápido).
2. Guardam o histórico de operações (autograd) para o Backpropagation.

```python
import torch
import numpy as np

# 0D: Escalar
escalar = torch.tensor(42)
# 1D: Vetor
vetor = torch.tensor([1.0, 2.0, 3.0])
# 2D: Matriz
matriz = torch.tensor([[1, 2], [3, 4]])

print(f'Escalar (0D) dim: {escalar.ndim}, shape: {escalar.shape}, dtype: {escalar.dtype}')
print(f'Vetor (1D) dim:   {vetor.ndim}, shape: {vetor.shape}, dtype: {vetor.dtype}')
print(f'Matriz (2D) dim:  {matriz.ndim}, shape: {matriz.shape}, dtype: {matriz.dtype}')
```

## 2️⃣ Funções Básicas de Criação

PyTorch tem a mesma API do NumPy para criar zeros, ones e randoms.

```python
zeros = torch.zeros((3, 3))
ones = torch.ones((2, 4))
rand = torch.randn((3, 3))  # Distribuição Normal

print('Zeros:\n', zeros)
print('\nNormal:\n', rand)
```

## 3️⃣ NumPy ↔ PyTorch

Eles são melhores amigos. Você vai transitar entre os dois o tempo todo.

```python
# Numpy para Tensor
array_np = np.array([1.5, 2.5, 3.5])
tensor_pt = torch.from_numpy(array_np)

# Tensor para Numpy
array_volta = tensor_pt.numpy()

print('NumPy:', type(array_np))
print('Tensor:', type(tensor_pt))
print('De Volta:', type(array_volta))
```
