## 1️⃣ O Git Não Aceita Imagens

Se você dar `git commit` num dataset de 20GB de imagens, o GitHub vai recusar e seu `.git` local vai travar seu PC. O Git foi feito para texto (código).

## 2️⃣ Como o DVC Funciona

Ele age igual ao Git, mas em vez de subir os dados pro GitHub, ele sobe pra um Storage (S3 da AWS, Google Drive, Azure). No GitHub fica salvo apenas um arquivo `.dvc` (Que é um ponteiro com o Hash MD5 dos dados).
Se alguém clonar seu repositório, dá um `dvc pull` e o DVC baixa as imagens exatas daquela versão do código.

```python
print('Comandos básicos conceituais:')
print('dvc init')
print('dvc add pasta_do_dataset/  (Isso gera um arquivo .dvc)')
print('git add pasta_do_dataset.dvc (O Git versiona o ponteiro, não as fotos)')
print('dvc push')
```
