## 1️⃣ O Segredo do Cache (Camadas)

O Docker constrói linha por linha. Se a linha falhar, ele cacheia as anteriores.
**Sempre** copie o `requirements.txt` e dê `pip install` ANTES de copiar o seu código Python. Porque você altera seu código toda hora, mas muda bibliotecas raramente. Assim, o build que levava 10 minutos passará a levar 2 segundos.

```python
dockerfile_ideal = """
# 1. Imagem Base (Use Slim para economizar espaço)
FROM python:3.10-slim

# 2. O pesadelo do OpenCV: Ele precisa de libs C++ no Linux que não vêm no 'slim'
RUN apt-get update && apt-get install -y libgl1-mesa-glx libglib2.0-0 && rm -rf /var/lib/apt/lists/*

WORKDIR /app

# 3. Aproveitando o Cache!
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt

# 4. Só agora copia o código fonte
COPY . .

EXPOSE 8000
CMD ["uvicorn", "app:app", "--host", "0.0.0.0", "--port", "8000"]
"""
print('Salve isso num arquivo chamado literalmente "Dockerfile" (Sem extensão).')
```
