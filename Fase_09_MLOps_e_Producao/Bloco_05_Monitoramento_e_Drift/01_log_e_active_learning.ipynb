## 1️⃣ Modelos Apodrecem

Você treina um modelo perfeito em 2024 pra achar defeitos em peças vermelhas. Em 2025, a fábrica troca as luzes do galpão pra lâmpadas de LED azuis e as peças passam a refletir azul. O modelo que tinha 99% de mAP passa a ter 20%. Isso é **Data Drift**.

## 2️⃣ Shadow Mode e Active Learning

- **Shadow Mode:** O modelo vai pra produção gerando predições, mas quem decide ainda é o humano. Medimos silenciosamente a métrica até termos confiança.
- **Active Learning:** Em produção, se o modelo fizer uma predição com Confiança Baixa (ex: `conf < 0.4`), a API não descarta, ela **salva a imagem no banco de dados**. Na sexta-feira, o engenheiro de IA revisa essas fotos 'difíceis', anota corretamente e re-treina o modelo na segunda-feira.
