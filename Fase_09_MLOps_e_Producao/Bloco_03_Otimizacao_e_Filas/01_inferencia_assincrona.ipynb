## 1️⃣ A Morte por Timeout

Se o seu modelo demora 3 segundos para processar 1 foto, e um cliente envia um lote de 50 fotos na API... a requisição vai demorar 150 segundos. O navegador do cliente vai dar 'Timeout' e fechar a conexão, mas seu servidor vai continuar processando a toa.

## 2️⃣ A Arquitetura Assíncrona (Event-Driven)

Em vez de o servidor processar e responder, ele faz assim:
1. Cliente envia imagem (POST).
2. A API salva a imagem numa Fila (ex: Redis) e devolve quase instantaneamente: `{ "status": "Em processamento", "job_id": 123 }`
3. Um 'Worker' (outro script Python rodando em background) pega imagens da fila e passa na GPU.
4. O Cliente fica perguntando (GET `/status/123`) até receber a resposta final.
