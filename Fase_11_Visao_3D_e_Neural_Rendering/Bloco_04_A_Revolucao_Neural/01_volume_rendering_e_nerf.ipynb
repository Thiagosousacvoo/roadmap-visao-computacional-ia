## 1️⃣ O Fim dos Polígonos (Malhas 3D)

Tradicionalmente, objetos 3D são formados por triângulos (Polígonos). É difícil renderizar fios de cabelo ou reflexos perfeitos.
O NeRF joga isso fora. O cenário 3D **passa a ser os Pesos de uma Rede Neural (MLP)**.

## 2️⃣ Como o NeRF funciona

A rede neural recebe 5 números de entrada: `(X, Y, Z)` da posição no espaço, e `(Theta, Phi)` do ângulo que a câmera está olhando.
A rede cospe 4 números: `(R, G, B, Densidade)`.

Ou seja, para renderizar uma foto de um ângulo novo, nós 'jogamos' milhões de raios virtuais pela cena, e perguntamos para a Rede Neural qual é a cor e a densidade da fumaça em cada milímetro do raio. Isso se chama **Volume Rendering**.
