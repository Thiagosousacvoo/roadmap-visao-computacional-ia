## 1️⃣ Como fazer 3D de um vídeo passeando numa sala?

Se você andar ao redor de uma estátua gravando um vídeo, o **Structure from Motion** faz magia:
1. Extrai features (SIFT) de todos os frames (Fase 02).
2. Encontra a correspondência: 'O nariz da estátua no Frame 1 é o mesmo pixel no Frame 20'.
3. Deduz, via geometria complexa (Bundle Adjustment), onde a câmera estava em cada frame (Extrínseca).
4. Gera uma nuvem de pontos esparsa (Sparse Cloud).
