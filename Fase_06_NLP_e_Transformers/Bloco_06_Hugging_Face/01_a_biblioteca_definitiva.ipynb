## 1️⃣ A Biblioteca Definitiva

Na vida real, não reimplementamos ViT e BERT do zero. Nós importamos da HuggingFace, que hospeda os pesos pré-treinados de bilhões de parâmetros.

```python
# pip install transformers
from transformers import pipeline

try:
    # Exemplo: Classificação Zero-Shot de Texto
    classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
    resultado = classifier(
        'A nova lente 50mm 1.8 é super nítida e barata!',
        candidate_labels=['comida', 'fotografia', 'esporte']
    )
    print(resultado)
except Exception as e:
    print('HuggingFace não está instalado ou falhou no download.')
```
