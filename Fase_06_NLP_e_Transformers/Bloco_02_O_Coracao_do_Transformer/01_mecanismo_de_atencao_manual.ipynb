## 🎯 Objetivos

- Implementar a equação da Atenção: `Softmax(Q*K.T / sqrt(dk)) * V`
- Entender o que é Query, Key e Value

---

## 1️⃣ O Exemplo do Arquivo de Aço

Pense na atenção como uma busca no banco de dados:
- **Query (Q):** O que você pesquisa (ex: 'Arquivo sobre Cachorro').
- **Key (K):** As etiquetas nas gavetas (ex: 'Felino', 'Canino', 'Ave').
- **Value (V):** O documento dentro da gaveta.

A Atenção mede a similaridade entre Query e Key. A Key que der 'match' com a Query libera o Value correspondente.

```python
import torch
import torch.nn.functional as F

def attention_simples(Q, K, V):
    d_k = Q.size(-1)  # Dimensão das keys
    
    # 1. Produto Escalar (Q * K.T) para ver similaridade
    scores = torch.matmul(Q, K.transpose(-2, -1))
    
    # 2. Divide pela raiz de d_k (estabilidade numérica)
    scores = scores / (d_k ** 0.5)
    
    # 3. Softmax (Transforma em porcentagens que somam 1)
    pesos = F.softmax(scores, dim=-1)
    
    # 4. Multiplica pelo Value
    saida = torch.matmul(pesos, V)
    return saida, pesos

print('Função de Atenção Padrão (Scaled Dot-Product) criada!')
```

## 2️⃣ Vendo a Atenção em Ação

Vamos criar vetores falsos para 3 palavras: 'Eu', 'Amo', 'Gatos'.

```python
# (Batch=1, Sequência=3 palavras, Dimensão=4)
Q = torch.randn(1, 3, 4)
K = torch.randn(1, 3, 4)
V = torch.randn(1, 3, 4)

saida, pesos = attention_simples(Q, K, V)

print('Matriz de Pesos de Atenção (3x3):')
print(torch.round(pesos[0] * 100) / 100)  # Arredondando para ficar legível
print('\nCada linha diz o quanto a palavra X prestou atenção nas palavras 1, 2 e 3.')
```
