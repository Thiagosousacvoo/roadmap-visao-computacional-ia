## 1️⃣ O Conceito de IoU

Se o humano desenhou a caixa perfeita (Ground Truth) e a rede previu uma caixa um pouco torta (Prediction), como damos uma 'nota' matemática pra isso?
Dividimos a **Área de Interseção** pela **Área de União** das duas caixas.

```python
import numpy as np

def calcular_iou(box1, box2):
    """
    Caixas no formato VOC: [x_min, y_min, x_max, y_max]
    """
    # 1. Achar as coordenadas da Interseção
    x1_inter = max(box1[0], box2[0])
    y1_inter = max(box1[1], box2[1])
    x2_inter = min(box1[2], box2[2])
    y2_inter = min(box1[3], box2[3])
    
    # 2. Calcular Área de Interseção (se não cruzar, é 0)
    largura_inter = max(0, x2_inter - x1_inter)
    altura_inter = max(0, y2_inter - y1_inter)
    area_inter = largura_inter * altura_inter
    
    # 3. Calcular Áreas das Caixas
    area_box1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area_box2 = (box2[2] - box2[0]) * (box2[3] - box2[1])
    
    # 4. Calcular União
    area_uniao = area_box1 + area_box2 - area_inter
    
    # 5. IoU
    return area_inter / (area_uniao + 1e-6)  # 1e-6 evita dividir por zero

print('Função IoU criada!')
```

## 2️⃣ Testando a Função

Se as caixas forem idênticas, IoU = 1.0. Se não se tocarem, IoU = 0.0. Em Detecção, convenciona-se que um **Acerto** tem IoU > 0.50.

```python
gt = [50, 50, 150, 150]
pred_boa = [55, 55, 145, 145]
pred_ruim = [100, 100, 200, 200]
pred_errada = [200, 200, 300, 300]

print(f'IoU Boa (Quase idêntica): {calcular_iou(gt, pred_boa):.2f}')
print(f'IoU Ruim (Meio sobreposta): {calcular_iou(gt, pred_ruim):.2f}')
print(f'IoU Errada (Não encosta): {calcular_iou(gt, pred_errada):.2f}')
```
