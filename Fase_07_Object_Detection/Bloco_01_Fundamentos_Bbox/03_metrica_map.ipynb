## 1️⃣ Precision, Recall e o Limiar de IoU

Se o IoU entre predição e real é maior que 0.5 (threshold), consideramos um 'Verdadeiro Positivo' (TP). Se for menor, é um 'Falso Positivo' (FP).
A Average Precision (AP) calcula a área abaixo da curva Precision-Recall para UMA classe. 
A **mAP** (Mean AP) tira a média da AP de TODAS as classes.

## 2️⃣ A Escala COCO (mAP@0.5:0.95)

Exigir IoU=0.5 é fácil. O desafio moderno (padrão COCO Dataset) é calcular o AP exigindo IoU de 0.50, 0.55, 0.60... até 0.95, e tirar a média disso. 
Se uma rede atinge `mAP@0.5:0.95 = 50%`, ela é considerada EXCELENTE.
