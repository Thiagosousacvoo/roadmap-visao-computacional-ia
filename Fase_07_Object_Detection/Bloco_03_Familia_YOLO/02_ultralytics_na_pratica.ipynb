## 1️⃣ Ultralytics (A API Definitiva)

A Ultralytics empacotou o YOLO numa API tão fácil de usar que virou padrão na indústria.

```python
# Rode no terminal: pip install ultralytics
import cv2
import numpy as np
import matplotlib.pyplot as plt

try:
    from ultralytics import YOLO
    
    # Baixa automaticamente o modelo treinado no COCO (80 classes)
    # Modelos: n(nano), s(small), m(medium), l(large), x(extra large)
    model = YOLO('yolov8n.pt')
    
    # Vamos criar uma imagem fake para passar
    img = np.random.randint(0, 255, (480, 640, 3), dtype=np.uint8)
    
    results = model(img) # Predição é simples assim!
    
    print('Caixas detectadas:\n', results[0].boxes)
except Exception as e:
    print('Por favor, instale a ultralytics para testar')
```

## 2️⃣ Navegando nos Resultados

O objeto `results[0]` guarda tensores na GPU. Podemos extrair Bboxes (xyxy), Classes (cls) e Confiança (conf).
