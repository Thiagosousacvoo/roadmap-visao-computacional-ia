## 1️⃣ O Paradigma 'You Only Look Once'

Detectores antigos (R-CNN) extraíam 2.000 pedaços da imagem e classificavam cada um. O YOLO passa a imagem inteira na CNN **apenas uma vez**.

## 2️⃣ O Grid de Predição

A rede transforma a imagem (ex: 640x640) num grid final (ex: 20x20). 
Cada célula desse grid é responsável por prever X caixas de objetos cujo centro caiam dentro da célula. 
Isso transforma a Detecção num problema bizarro de Regressão Massiva (A CNN devolve um Tensor colossal com predições de caixas para o grid todo).

## 3️⃣ O Problema e a Solução (NMS)

Se houver um cachorro, 4 células vizinhas podem detectá-lo ao mesmo tempo.
A etapa de pós-processamento, **NMS (Non-Maximum Suppression)**, resolve isso: ela pega as caixas que têm IoU muito alto entre si, e apaga as de menor confiança, sobrando só uma.
